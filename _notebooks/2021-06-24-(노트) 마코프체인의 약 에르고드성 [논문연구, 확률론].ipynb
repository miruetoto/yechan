### Basic setting and Definitions 


`(м •мқҳ)` Let $E$ be a countable space and let $\alpha$ and $\beta$ be probability distribution on $E$. The distance in variation $d_V(\alpha,\beta)$ betweem $\alpha$ and $\beta$ is defined by 

$$ d_V(\alpha,\beta)=\frac{1}{2}|\alpha-\beta|=\frac{1}{2}\sum_{i\in E} |\alpha(i)-\beta(i)|.$$

`-` мҳҲм ң: $\alpha=(0.1,0.1,0.8)$, $\beta=(0.2,0.1,0.7)$ $\to$ $d_V(\alpha,\beta)=\frac{1}{2}\big(0.1+0+0.1\big)=0.1$

`(м •мқҳ)` Let $E$, $F$ denote countable sets. The matrix ${\bf Q}$ indexed by $F \times E$ called stochastic matrix whose rows (index by $F$) are probability distributions on $E$.
