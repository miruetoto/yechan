### groupby 

`-` 데이터프레임을 여러개의 서브데이터프레임으로 나누는 기능 


`-` 단독으로 쓸 이유는 별로 없다. $\to$ 그룹을 나누고 어떠한 "연산"을 하기 위함임. 

```python
df.groupby('AIRLINE')
```

- 데이터프레임을 각 항공사별로 나눔

`-` 확인

```python
grouped_df = df.groupby('AIRLINE')
```

```python
grouped_df.groups
```

- 너무 보기힘들다 

`-` 보기좋은 형태로 확인 

```python
list(grouped_df.groups)
```

```python
grouped_df.get_group('AA')
```

```python
for g in grouped_df.groups: 
    print(g)
    display(grouped_df.get_group(g))
```

#### AIRLINE에 따라 데이터 프레임 나누고 $\to$ ARR_DELAY에 mean함수 적용 (AIRLINE $\to$ {ARR_DELAY: mean}) 

`-` 방법1 ($\star$)

```python
df.groupby('AIRLINE').agg({'ARR_DELAY':'mean'})
```

`-` 방법2 ($\star\star\star$)

```python
df.groupby('AIRLINE').agg({'ARR_DELAY':np.mean})
```

`-` 방법3

```python
df.groupby('AIRLINE')['ARR_DELAY'].agg('mean')
```

`-` 방법4  ($\star$)

```python
df.groupby('AIRLINE')['ARR_DELAY'].agg(np.mean)
```

`-` 방법5

```python
df.groupby('AIRLINE')['ARR_DELAY'].mean()
```

`-` 방법2,4는 사용자 정의함수를 쓸 수 있다는 장점이 있음 **(엄청난 장점임)**

```python
df.groupby('AIRLINE').agg({'ARR_DELAY':lambda x: -np.mean(x)})
```

```python
df.groupby('AIRLINE')['ARR_DELAY'].agg(lambda x: -np.mean(x))
```

`-` 입력이 여러개인 사용자 정의함수도 가능함 

```python
def f(x,y): return np.mean(x)**y
```

```python
df.groupby('AIRLINE').agg({'ARR_DELAY': lambda x: f(x,2)})
```

```python
df.groupby('AIRLINE')['ARR_DELAY'].agg(f,2)
```

#### AIRLINE,WEEKDAY $\to$ {CANCELLED: sum}

`-` 방법1~5

```python
df.groupby(['AIRLINE','WEEKDAY']).agg({'CANCELLED':'sum'})
```

```python
df.groupby(['AIRLINE','WEEKDAY']).agg({'CANCELLED':np.sum})
```

```python
df.groupby(['AIRLINE','WEEKDAY'])['CANCELLED'].agg('sum')
```

```python
df.groupby(['AIRLINE','WEEKDAY'])['CANCELLED'].agg(np.sum)
```

```python
df.groupby(['AIRLINE','WEEKDAY'])['CANCELLED'].sum()
```

#### AIRLINE,WEEKDAY $\to$ {CANCELLED: sum, mean}, {DIVERTED: sum, mean}

`-` 방법1~4 (방법5는 쓸 수 없다.)

```python
df.groupby(['AIRLINE','WEEKDAY']).agg({'CANCELLED':['sum','mean'],'DIVERTED':['sum','mean']})
```

```python
df.groupby(['AIRLINE','WEEKDAY']).agg({'CANCELLED':[np.sum,np.mean],'DIVERTED':[np.sum,np.mean]})
```

```python
df.groupby(['AIRLINE','WEEKDAY'])[['CANCELLED','DIVERTED']].agg(['sum','mean'])
```

```python
df.groupby(['AIRLINE','WEEKDAY'])[['CANCELLED','DIVERTED']].agg([np.sum,np.mean])
```

```python
# df.groupby(['AIRLINE','WEEKDAY'])[['CANCELLED','DIVERTED']].sum() <-- 확장불가능
```

#### AIRLINE,WEEKDAY  $\to$ {CANCELLED: sum,mean,size}, {AIR_TIME: mean,var}

`-` 방법1,2만 적용가능함 

```python
df.groupby(['AIRLINE','WEEKDAY']).agg({'CANCELLED':['sum','mean','size'],'AIR_TIME':['mean','var']})
```

```python
df.groupby(['AIRLINE','WEEKDAY']).agg({'CANCELLED':[np.sum,np.mean,len],'AIR_TIME':[np.mean,np.std]})
```

`-` 거리를 기준으로 적당히 데이터프레임을 나누어보자. 

```python
df.DIST.hist()
```

```python
df.DIST.describe()
```

`-` 구간은 아래와 같이 잡으면 될것 같다. 

```python
bins=[-np.inf, 400,700,1200, np.inf] 
```

`-` pd.cut()을 이용하여 각 obs를 구간에 매칭시키자. 

```python
cuts=pd.cut(df.DIST,bins=bins)
cuts
```

`-` 아래와 같이 각 구간별로 적당히 균등한 숫자로 나누어졌다. (당연히 퀀타일을 보고 나누엇으니) 

```python
cuts.value_counts()
```

`-` cuts, AIRLINE $\to$ {DIVERTED: sum} 

```python
df.groupby([cuts,'AIRLINE']).agg({'DIVERTED': np.sum})
```

`-` 그룹핑이 잘되었다. 이전에 했던 결과와 비교해보자. 

```python
df.groupby(['AIRLINE']).agg({'DIVERTED':np.sum})
```

`-` 조금 다른 특징들을 관찰가능함. 
- AA항공사와 DL항공사는 모두 비슷한 우회횟수를 가지고 잇음. 
- AA항공사는 700이상 우회를 많이하고 DL항공사는 400~700사이에서 우회를 많이 함. (패턴이 다름) 

`-` 구간이름이 조금 보기 싫은것 같음. 

```python
cuts2=pd.cut(df.DIST, bins=bins, labels=['Q1','Q2','Q3','Q4'])
```

```python
df.groupby([cuts2,'AIRLINE']).agg({'DIVERTED': np.sum})
```
