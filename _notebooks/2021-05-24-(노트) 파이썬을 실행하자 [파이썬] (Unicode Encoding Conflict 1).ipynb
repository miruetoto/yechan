### python 콘솔 

#### python 콘솔

`-` 시스템 명령라인에서 python을 실행한뒤 파이썬 콘솔에서 실행 

```
(py38r40) cgb@cgb-All-Series:~/Desktop$ python
Python 3.8.5 (default, Sep  4 2020, 07:30:14) 
[GCC 7.3.0] :: Anaconda, Inc. on linux
Type "help", "copyright", "credits" or "license" for more information.
>>> [1,2,3]+[4]
[1, 2, 3, 4]
```

`-` 장점? 
- print 하지 않아도 결과를 바로바로 확인가능 (디버깅, 테스트에 극도로 유리함)
- 남의 코드를 빨리 배울 수 있다. 

`-` 단점
- 코드를 실행하는 순간 사라진다. $\to$ ?? 그냥 `test.py` 쓸때가 좋았던것 같기도? 

#### python 콘솔 + import module

`-`  코드를 실행하는 순간 사라지는 단점은 모듈을 임포트하여 극복할 수 있다. 즉 파이썬콘솔에서 
```python
import test
```
를 치면 해결된다. 

#### python 콘솔 + exec

`-` `test.py`의 코드를 조금 수정하여 다시 실행하고 싶다.

`-` 메모장 켜서 수정을 할 수 있겠으나 
```python
import test 
```
로 실행할수는 없다. (임포트는 한번만 가능)

`-` 그냥 붙여넣기해서 넣을순없나? $\to$ open, read, exec 조합으로 해결가능 
```python
>>> exec(open('test.py').read())
[1, 2, -1, -2, -3]
```

`-` 심지어 그냥 test.txt로 저장한뒤 넣어도 가능

#### ipython 콘솔

`-` python에서 for문을 돌려보자. $\to$ 들여쓰기가 안될것임. 

`-` python에서 아래를 실행해보자. 
```python
a=[1,2,3]
a?
```

`-` 이제 ipython에서 동일한 작업을 해보자. 

`-` `a.`을 치고 tab을 눌러보자.
