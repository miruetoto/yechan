### map, map_* : list(a=1:5, b:1:30) $\to$ list(fun(a),fun(b))

`-` map, map_* 의 특징 

- C로 구현되어 있어 빠르다. 하지만 가독성이 희생되었다. (읽기 어렵다)
- 두번째 인수는 적용할 함수를 나타내며 `.f`로 되어있는데 이는 공식, 문자형벡터, 정수형벡터등을 지정할 수 있다. 
- map_* 은 `.f`가 호출이 될 때마다 추가인수를 `...` 사용하여 전달할 수 있다. 여기에서 `...` 은 길이가 정해지지 않은 입력. 

`-` data

```r
df <- tibble(
  a = rnorm(10),
  b = rnorm(10),
  c = rnorm(10),
  d = rnorm(10)
)
print(df)
```

#### ex1: map, map_* 차이

```r
map_dbl(df, mean) %>% print()
map_dbl(df, median) %>% print()
map_dbl(df, sd) %>% print()
```

`-` 결과는 벡터로 나온다. (벡터는 길이가 4이고 각 원소마다 a,b,c,d 라는 이름이 있음)

`-` 결과가 벡터이므로 아래와 같은 연산이 가능함 

```r
print(map_dbl(df, mean) + 1 )
```

`-` 아래와 같이 각 원소에 이름이 있는 벡터를 선언할 수 있음. 

```r
c(a=1.1,b=2.2,c=3.3,d=4.4) %>% print()
```

`-` map_dbl()이 아니라 map()을 사용해도 좋음. 단지 결과가 리스트로 나옴

```r
map(df,mean) %>% print()
```

```r
map(df,median) %>% print()
```

```r
map(df,sd) %>% print()
```

`-` map_chr을 쓰면 문자열벡터가 결과로 나옴

```r
map_chr(df,mean) %>% print()
```

`-` map_int을 쓰면 인트벡터가 결과로 나옴

```r
map_int(df,length) %>% print()
```

```r
map_int(df, function(x) max(x)>0) %>% print()
```

`-` 그런데 결과가 이미 double로 나온 경우라면 int로 만들 수 없음

```r
map_int(df, mean)
```

`-` map_lgl 역시 마찬가지임

```r
map_lgl(df, function(x) max(x)>0) %>% print()
```

```r
map_lgl(df, function(x) max(x)>1) %>% print()
```

```r
map_lgl(df, function(x) max(x)>1.5) %>% print()
```

```r
map_lgl(df, mean)
```

#### ex2: %>% 연산자사용 

`-` pip를 사용할수도 있음 

```r
df %>% map_dbl(mean) %>% print()
df %>% map_dbl(median) %>% print()
df %>% map_dbl(sd) %>% print()
```

#### ex3: additional arguments 

```r
mean(c(0,0,0,1,1,1,1,1)) %>% print()
mean(c(0,0,0,1,1,1,1,1),trim=0.5) %>% print()
```

```r
map_dbl(df, mean, trim = 0.5) %>% print()
```

#### ex4: 길이가 같지 않은 벡터들의 모임에서도 적용가능 

```r
z <- list(x = 1:3, y = 4:50) 
print(z)
```

```r
map(z, length) %>% print()
```

```r
map_dbl(z, mean) %>% print()
```

#### ex5: 응용예제

```r
mtcars %>% print()
```

```r
mtcars %>% split(.$cyl) %>% print()
```

```r
mtcars %>% split(.$cyl) %>% map(function(df) lm(mpg ~ wt, data = df))
```

```r
mtcars %>% split(.$cyl) %>% map(~lm(mpg ~ wt, data = .))
```

```r
mtcars %>% split(.$cyl) %>% map(~lm(mpg ~ wt, data = .)) %>% 
  map(summary) 
```

```r
mtcars %>% split(.$cyl) %>% map(~lm(mpg ~ wt, data = .)) %>% 
  map(summary) %>% map_dbl(~.$r.squared) %>% print()
```
