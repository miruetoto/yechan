### map_dbl, map_int, map_chr, map_lgl: list(a=1:5, b:1:30) $\to$ list(fun(a),fun(b))

`-` data

```r
df <- tibble(
  a = rnorm(10),
  b = rnorm(10),
  c = rnorm(10),
  d = rnorm(10)
)
print(df)
```

#### use1

```r
map_dbl(df, mean) %>% print()
map_dbl(df, median) %>% print()
map_dbl(df, sd) %>% print()
```

`-` 결과는 벡터로 나온다. (벡터는 길이가 4이고 각 원소마다 a,b,c,d 라는 이름이 있음)

`-` 결과가 벡터이므로 아래와 같은 연산이 가능함 

```r
print(map_dbl(df, mean) + 1 )
```

`-` 아래와 같이 각 원소에 이름이 있는 벡터를 선언할 수 있음. 

```r
c(a=1.1,b=2.2,c=3.3,d=4.4) %>% print()
```

`-` map_dbl()이 아니라 map()을 사용해도 좋음. 단지 결과가 리스트로 나옴

```r
map(df,mean) %>% print()
```

```r
map(df,median) %>% print()
```

```r
map(df,sd) %>% print()
```

#### use2 

```r
df %>% map_dbl(mean)
df %>% map_dbl(median)
df %>% map_dbl(sd)
```

#### use3: map_dbl은 additional arguments 를 전달할 수 있다. 

```r
mean(c(0,0,0,1,1,1,1,1))
mean(c(0,0,0,1,1,1,1,1),trim=0.5)
```

```r
map_dbl(df, mean, trim = 0.5)
```

```r
mtcars %>% split(.$cyl) %>% map(function(df) lm(mpg ~ wt, data = df))
```

```r
mtcars %>% split(.$cyl) %>% map(~lm(mpg ~ wt, data = .))
```

```r
mtcars %>% split(.$cyl) %>% map(~lm(mpg ~ wt, data = .)) %>% 
  map(summary) 
```

```r
mtcars %>% split(.$cyl) %>% map(~lm(mpg ~ wt, data = .)) %>% 
  map(summary) %>% map_dbl(~.$r.squared)
```
