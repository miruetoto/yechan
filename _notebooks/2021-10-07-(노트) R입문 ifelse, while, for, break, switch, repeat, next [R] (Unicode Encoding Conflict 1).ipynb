### while 

`-` 예제1

```r
tot=0 
i=1
while(i<=100){
    tot = tot+i 
    i = i+1 
}
```

```r
tot
```

`-` 예제1의 다른구현 (1)

```r
tot=0 
for (i in 1:100) tot = tot + i 
```

```r
tot
```

`-` 예제1의 다른구현 (2)

```r
tot=0
i=1
while(TRUE){
   tot = tot+i
    i=i+1
    if (i>100) break 
}
```

```r
tot
```

`-` 예제1의 다른구현 (3)

```r
tot=0
i=1
while(TRUE){
   tot = tot+i
    if (i>=100) break 
    i=i+1
}
```

```r
tot
```

### break

`-` 예제1

```r
for(i in 1:10){
    print(i)
    if(i==3) break 
}        
```

`-` 예제2

```r
for(i in 1:10){
    for(j in 1:10){
        print(c(i,j))
        if(j==3) break 
    }
    if(i==5) break 
}
    
```

### ifelse 

`-` 예제1

```r
x = -2:2 
x 
```

```r
ifelse(x>0,x,-x)
```

`-` 예제1의 다른구현 

```r
length(x) 
```

```r
y_ = x*0 
```

```r
for (i in 1:length(x)){
    if (x[i]>0) y_[i] = x[i] else y_[i] = -x[i]
}
```

```r
y_
```

`-` 예제2 

```r
x= -5:5
```

```r
x
```

```r
y= ifelse(x<0,0, ifelse(x>2,x,1))
y
```

```r
plot(x,x)
points(x,y,col=2,pch=16)
```

`-` 예제2의 다른구현 

```r
x = -5:5 
x
```

```r
y = x*0 
y
```

```r
n= length(x)
n
```

```r
for (i in 1:n){
    if (x[i]<0) y[i]=0 else 
        if (x[i]>2) y[i]=x[i] else
            y[i]=1
}
```

```r
y
```

`-` ifelse와 if는 성격이 조금 다름
- if: 조건에 따라 코드 실행 흐름을 제어하는 성격 
- ifelse: 새로운 벡터를 생성하는 목적

`-` 예제4

```r
x=1:10
x
```

```r
c(
    mean(x),
    mean(x**2),
    mean((x-mean(x))**2)
)
```

```r
38.50- 5.50**2 
```

```r
switch("f1",
       f1=mean(x), 
       f2=mean(x**2), 
       f3=mean((x-mean(x))**2)
)
```

```r
switch("f2",
       f1=mean(x), 
       f2=mean(x**2), 
       f3=mean((x-mean(x))**2)
)
```

```r
switch("f3",
       f1=mean(x), 
       f2=mean(x**2), 
       f3=mean((x-mean(x))**2)
)
```

`-` 예제5: 예제4의 변형 

```r
myfun<-function(x,f){
    switch(f,
       f1=mean(x), 
       f2=mean(x**2), 
       f3=mean((x-mean(x))**2))
}
```

```r
myfun(x,"f1")
```

```r
myfun(x,"f2")
```

```r
myfun(x,"f3")
```

### repeat 

`-` 예제1

```r
tot=0 
i=1
repeat{
tot=tot+i
i=i+1
if (i>100) break
}
```

```r
tot
```

`-` 잘못된 예제1: 무한루프

### next 

`-` 예제1

```r
for(i in 1:5){
    if(i==3) next else 
    print(i)    
}
```

break와 비교

```r
for(i in 1:5){
    if(i==3) break else 
    print(i)    
}
```

`-` 예제2

```r
i=0
repeat{
    i=i+1
    if (i>5) break 
    if (i==3) next
    print(i) 
}
```
