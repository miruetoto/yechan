- 이런방식이 사실 좀 이상해보여서 한번쓰기 시작하면 잘쓰는데 금방까먹고 다시 헷갈림 

`-` 아래의 사실을 기억하자. 
- 기본적으로 loc, iloc은 `[2], [2:]` 처럼 1차원으로 원소를 인덱싱할수도 있고, `[2,3], [:,2]`와 같이 2차원으로 인덱싱 할수도 있다. 
- 1차원으로 인덱싱하는 경우는 기본적으로 행을 인덱싱한다. $\to$ **loc, iloc은 행과 더 친하고 열과 친하지 않다.**
- 따라서 열을 선택하는 방법에 있어서 loc, iloc이 그렇게 친화적인 방법은 아니다. 
- 그렇지만 달리 열을 선택하는 쉬운 방법이 존재하지 않아서 열을 선택할때에도 loc, iloc이 선호된다.
