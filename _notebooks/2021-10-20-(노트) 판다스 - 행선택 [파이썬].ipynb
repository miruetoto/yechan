### 기본 마인드: 1차원 자료형에서 원소 선택하듯이!

```python
import numpy as np
import pandas as pd 
```

```python
dic={'X1':np.random.normal(0,1,5), 
     'X2':np.random.normal(0,1,5),
     'X3':np.random.normal(0,1,5),
     'X4':np.random.normal(0,1,5),
     'X5':np.random.normal(0,1,5),  
     'X6':np.random.normal(0,1,5)
    }
df=pd.DataFrame(dic)
df
```

### 첫번째 행을 선택하는 방법 

`-` 방법1

```python
df.iloc[0] # 이상해 
```

`-` 방법2

```python
df.iloc[0,:]
```

`-` 방법3

```python
df.iloc[[0]]
```

`-` 방법4

```python
df.iloc[[0],:]
```

`-` 방법5

```python
df.loc[0]
```

`-` 방법6

```python
df.loc[0,:]
```

`-` 방법7

```python
df.loc[[0]]
```

`-` 방법8

```python
df.loc[[0],:]
```

### loc이 좋지 않아 보이지만 특정한 경우에는 좋다. 

```python
df=pd.DataFrame(np.random.randn(10,4),columns=list('ABCD'),index=pd.date_range('20210101',periods=10))
df
```

```python
df.iloc[0:2]
```

```python
df.loc['20210103':'20210105']
```
