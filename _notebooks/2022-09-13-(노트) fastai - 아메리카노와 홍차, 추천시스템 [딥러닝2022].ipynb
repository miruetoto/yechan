```python
dls = CollabDataLoaders.from_df(df)
```

```python
dls.show()
```

```python
lrnr = collab_learner(dls,y_range=(0,5))
```

```python
lrnr.fit(30)
```

```python
lrnr.model
```

```python
X,y = dls.one_batch()
```
