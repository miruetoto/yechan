`-` R에서는 맞는 비유인데 파이썬에서는 비유이다. 왜? 

````python
id(a) ## 변수 a가 저장된 메모리주소(=방번호)를 출력하는 함수 
````

````python
id(b) ## 변수 b가 저장된 메모리주소(=방번호)를 출력하는 함수 
````

실제로는 변수 a,b의 방주소가 동일하기 때문이다. 

`-` 따라서 파이썬에서는 아래와 같은 비유가 더 적절하다. 

(1) 메모리는 방이 여러개 있는호텔이라고 생각하자. (방은 변수를 저장할 수 있는 공간이다)

(2) 아래를 실행했을 때 
```python 
a=[1,2,3]
```
- 메모리주소1(=140628239102272)에서 [1,2,3]을 만든다. 
- 메모리주소1(=140628239102272)의 방문에 a라는 포스트잇을 붙인다. 
- 따라서 앞으로 [1,2,3]을 찾으려면 a라고 적힌 포스트잇을 찾으면 된다.

(3) 아래를 실행했을 때 
```python
b=a
```
- b라는 포스트잇을 새로 만들고 그 포스트잇은 a와 같은 효과를 주도록 한다. 
- 쉽게말하면 b라는 포스트잇을 만들고 메모리주소1(=140628239102272)의 방문에 b라는 포스트잇을 붙인다는 의미이다. 
- 따라서 [1,2,3]을 찾으려면 a라는 포스트잇을 찾거나 b라는 포스트잇을 찾으면 된다. 

(4) 아래를 실행했을 때 
```python
a.append(4)
```
- a라는 포스트잇이 붙어있는 방으로 찾아가서 [1,2,3]을 찾고 거기에서 append함수를 써서 [1,2,3,4]로 바꾼다. 
- 같은방에 a,b라는 포스트잇이 붙어이므로 b라는 포스트잇을 찾아가서 내용을 열어보면 [1,2,3,4]가 나온다. 

#### 할당문(=)의 이해

`-` 파이썬에서 할당문을 이해하려면 언제나 오른쪽을 먼저 읽어야 한다. 
- 할당문의 오른쪽에서 객체를 생성하거나 가져온다. 
- 그 후에 레이블을 붙이듯이 할당문 왼쪽의 변수가 객체에 바인딩 된다. (이걸 참조라고 부름) 

`-` `b=a`는 

> 나는 이미 a가 의미하는걸 알고 있어, 그런데 a가 의미하는걸 b라고도 부르고싶어

라는 것과 같다. 즉 이미 a라고 부르고 있는것을 가져와서 b라고도 부르고 싶다는 의미인데, 이러한 관점에서 b는 a의 별칭(alias)이라고 볼 수 있다. 

`-` 반대로 생각하면 a역시 b의 별명이라고 볼 수 있다. 

`-` 하나의 메모리 주소에 변수의 이름을 바인드 하는 것을 참조라고 하고, 하나의 메모리 주소에 여러개의 변수이름을 바인드 하는 것을 aliasing이라고 한다.

    - $\to$ 할수없이 새로운공간 X2를 만들고 거기에 (66,55,44,33)을 넣고 그것을 l2에 맵핑함. 

결국 아래와 같은 상황이 되었음 

- 공간X1: (66,55,44) 
- 공간X2: (66,55,44,33)

`-` 파이썬은 메모리를 아끼기 위해서 shallowcopy 라는 이상한 행동을 한다.

`-` 오묘하게도 [1,2,3,4,5]와 같이 0차원 자료형으로만 구성된 리스트라면 문제가 되지 않는다. (메모리는 아끼면서 문제가 되지 않는다? 천재인데?)

`-` 그런데 [[1,2],[3,4]]와 같이 리스트에 리스트가 들어간 형태라면 문제가 생긴다. (이건 개선이 필요하겠는걸?)
- 개선1: 깊은복사
- 개선2: numpy 

### deep copy 

`-` 얕은복사가 아니라 내부객체까지 모두 새로 만드는 깊은복사를 하기 위해서는 아래와 같이 copy 모듈을 사용한다. 

````python
import copy 
````

`-` 깊은복사 예제1

````python
l1 = [3,[66,55,44]]
````

````python
l2 = copy.deepcopy(l1)
````

````python
l2[1].append(55)
````

````python
l1,l2
````

`-` 깊은복사 예제2

````python
l1 = [3,[66,[55,44]]]
````

````python
l2 = copy.deepcopy(l1)
````

````python
id(l1),id(l2) ## shallow copy = 1 level deep copy 
````

````python
id(l1[1]),id(l2[1]) ## deep copy = 2 level deep copy 
````

````python
id(l1[1][1]),id(l2[1][1]) ## deep copy = 3 level deep copy 
````

````python
l2[1][1].append(33)
````

````python
l1,l2
````
