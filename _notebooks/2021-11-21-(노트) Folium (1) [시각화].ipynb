### folium.plugins.TimestampedGeoJson

```python
a1=[35.847336682624395, 127.13066143667929]
a2=[35.847163718642705, 127.12985784659246]
a3=[35.8466227437543, 127.13004852898592]
a4=[35.845842555784095, 127.13032547246222]
a5=[35.84567326870173, 127.13185547166708]
a1.reverse()
a2.reverse()
a3.reverse()
a4.reverse()
a5.reverse()
t1='2021-11-21 12:00:00'
t2='2021-11-21 12:05:00'
t3='2021-11-21 12:10:00'
t4='2021-11-21 12:15:00'
t5='2021-11-21 12:20:00'


points = [
    {
        "time": t1,
        "coordinates": a1,
    },
    {
        "time": t2,
        "coordinates": a2,
    },
    {
        "time": t3,
        "coordinates": a3,
    },
    {
        "time": t4,
        "coordinates": a4,
    },
    {
        "time": t5, 
        "coordinates": a5
    },
]

features = [
    {
        "type": "Feature",
        "geometry": {
            "type": "LineString",
            "coordinates": [
                a1,
                a2,
                a3,
                a4,
                a5,
            ],
        },
        "properties": {
            "times": [
                t1,
                t2,
                t3,
                t4,
                t5,
            ],
            "icon": "circle",
            "iconstyle": {
                "fillColor": "green",
                "fillOpacity": 0.6,
                "stroke": "false",
                "radius": 13,
            },
            "style": {"weight": 0},
            "id": "man",
        },
    }
]

m = folium.Map([35.84694109945701, 127.12941011336521],
               zoom_start=17,
               scrollWheelZoom=False)

folium.plugins.TimestampedGeoJson(
    {"type": "FeatureCollection", "features": features},
    period="PT1M",
    add_last_point=True,
).add_to(m)

m
```

```python
m = folium.Map([35.84694109945701, 127.12941011336521],
               zoom_start=17,
               scrollWheelZoom=False)
a1=[35.847336682624395, 127.13066143667929]
a2=[35.847163718642705, 127.12985784659246]
a3=[35.8466227437543, 127.13004852898592]
a4=[35.845842555784095, 127.13032547246222]
a5=[35.84567326870173, 127.13185547166708]
a1.reverse()
a2.reverse()
a3.reverse()
a4.reverse()
a5.reverse()
t1='2021-11-21 12:00:00'
t2='2021-11-21 12:05:00'
t3='2021-11-21 12:10:00'
t4='2021-11-21 12:15:00'
t5='2021-11-21 12:20:00'
lines=[{'coordinates':[a1,a2], 'dates': [t1,t2],'color':'red','weight':2},
       {'coordinates':[a2,a3], 'dates': [t2,t3],'color':'blue','weight':3},
       {'coordinates':[a3,a4], 'dates': [t3,t4],'color':'red','weight':6},
       {'coordinates':[a4,a5], 'dates': [t4,t5],'color':'blue','weight':8}]
           
features=[{'type':'Feature',
           'geometry':{'type':'LineString',
                       'coordinates':line['coordinates'],
                      },
           'properties':{'times':line['dates'],
                         'style':{
                             'color':line['color'],
                             'weight':line['weight']
                         }
                        }
          } for line in lines]

folium.plugins.TimestampedGeoJson({
    'type':'FeatureCollection',
    'features':features,
}, period='PT30S', add_last_point=True).add_to(m)
m
```

### folium.plugins.HeatMapWithTime

`-` 입자의 위치 

$$
\begin{cases}
x_i = x_{i-1}+\epsilon_{i} \\
y_i = y_{i-1}+\eta_{i}
\end{cases}
$$

단 $(x_0,y_0)=(35.836603154469884, 127.11406532500682)$

```python
t0 = np.array([[35.836603154469884, 127.11406532500682]]*3000)
walk = np.random.normal(size=(3000, 2, 200)) * 0.0001 
```

```python
t0[:5] # 시점 0 
```

```python
(t0+walk[:,:,0])[:5] # 시점1 
```

```python
walk[:5,:,0]
```

```python
walk[:5,:,1]
```

```python
walk[:5,:,0]+walk[:5,:,1]
```

```python
walk[:5,:,:2].sum(axis=2)
```

```python
(t0+walk[:,:,:2].sum(axis=2))[:5] # 시점2
```

```python
data = [(t0+walk[:,:,:i].sum(axis=2)).tolist() for i in range(100)] # 랜덤한 확산.. 
```

```python
m = folium.Map([35.836603154469884, 127.11406532500682], 
                tiles='OpenStreetMap', 
                zoom_start=17,
                scrollWheelZoom=False)
hmwtPlugin = folium.plugins.HeatMapWithTime(data,radius=7)
hmwtPlugin.add_to(m)
m
```

### 예제: earthquakes 

#### Step1: Pandas

```python
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/earthquakes-23k.csv')
```

```python
df
```

```python
lst=[df.assign(Year=list(map(lambda x: x.split('/')[-1],df.Date))).\
     groupby('Year').\
     pipe(list)[i][1].\
     iloc[:,[1,2,3]].\
     to_numpy().tolist() for i in range(2017-1965)]
```

#### Step2: folium

```python
#Define the Folium Map
hm = folium.Map([0,0], tiles='OpenStreetMap', zoom_start=1,scrollWheelZoom=False)
#Add the Heatmap data to HeatMapWithTime plugin        
hmwtPlugin = folium.plugins.HeatMapWithTime(lst,index=list(range(1965,2017)),radius=6)
hmwtPlugin.add_to(hm)
#Render the Finished Map
hm
```
