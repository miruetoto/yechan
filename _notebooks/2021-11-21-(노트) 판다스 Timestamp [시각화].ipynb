```python
import pandas as pd 
import numpy as np
import datetime
```

### datetime

`-` 문자열을 판다스에서 지원하는 `Timestamp` object로 바꿔준다. 

`-` 어지간하면 바꿔줌 

```python
pd.to_datetime("1/1/2018")
```

```python
pd.to_datetime("1/1/2018")
```

```python
pd.to_datetime("2018.01.01")
```

```python
pd.to_datetime("2018-01-01")
```

```python
pd.to_datetime("2018/01/01")
```

```python
pd.to_datetime("2018/01/01 21:30")
```

```python
pd.to_datetime("2018/01/01 21:30:05")
```

`-` 아래는 못바꿔줌 

```python
pd.to_datetime("2018년 01월 01일 21시 30분 20초")
```
