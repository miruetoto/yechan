#### 2차원 배열과 연립1차방정식

`-` 아래의 연립방정식 고려 

$\begin{cases}
y + z + w= 3 \\
x + z + w= 3 \\
x + y + w= 3 \\ 
x + y + z= 3 \\ 
\end{cases}$ 

`-` 행렬표현은? 

$\begin{bmatrix}
0 & 1 & 1 & 1 \\ 
1 & 0 & 1 & 1 \\
1 & 1 & 0 & 1 \\ 
1 & 1 & 1 & 0 \\ 
\end{bmatrix} \begin{bmatrix} x \\ y \\ z \\ w \end{bmatrix} = \begin{bmatrix} 3 \\ 3 \\ 3 \\ 3  \end{bmatrix}$

`-` 풀이

```python
A = np.array([[0,1,1,1],[1,0,1,1],[1,1,0,1],[1,1,1,0]])
A
```

```python
b=np.array([3,3,3,3]).reshape(4,1)
b
```

```python
np.linalg.inv(A) @ b
```
