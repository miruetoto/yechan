### folium.plugins.TimestampedGeoJson

```python
m = folium.Map([35.84694109945701, 127.12941011336521],
               zoom_start=17,
               scrollWheelZoom=False)
a1=[35.847336682624395, 127.13066143667929]
a2=[35.847163718642705, 127.12985784659246]
a3=[35.8466227437543, 127.13004852898592]
a4=[35.845842555784095, 127.13032547246222]
a5=[35.84567326870173, 127.13185547166708]
a1.reverse()
a2.reverse()
a3.reverse()
a4.reverse()
a5.reverse()
t1='2021-11-21 12:00:00'
t2='2021-11-21 12:05:00'
t3='2021-11-21 12:10:00'
t4='2021-11-21 12:15:00'
t5='2021-11-21 12:20:00'
points=[{'coordinates':a1, 'time': t1},
       {'coordinates':a2, 'time': t2},
       {'coordinates':a3, 'time': t3},
       {'coordinates':a4, 'time': t4},
       {'coordinates':a4, 'time': t5}]
           
features=[{'type':'Feature',
           'geometry':{
               'type':'Point',
               'coordinates':point['coordinates'],
               },
           'properties':{
               'times':point['time'],
               'iconstyle':{
                   'iconSize':[20,20]}
               }
          } for point in points
         ]


features.append(
    {
        'type': 'Feature',
        'geometry': {
            'type': 'LineString',
            'coordinates': [a1,a2,a3,a4,a5],
        },
        'properties': {
            'times': [t1,t2,t3,t4,t5],
            'icon': 'circle',
            'iconstyle': {
                'fillColor': 'green',
                'fillOpacity': 0.6,
                'stroke': 'false',
                'radius': 13
            },
            'style': {'weight': 0},
            'id': 'man'
        }
    }
)

folium.plugins.TimestampedGeoJson({
    'type':'FeatureCollection',
    'features':features,
}, period='PT30S', add_last_point=True).add_to(m)
m
```

```python
import pandas as pd
import numpy as np
import folium
district_wise_census = pd.read_csv("https://raw.githubusercontent.com/kalilurrahman/datasets/main/district%20wise%20population%20and%20centroids.csv")
state_wise_centroid = pd.read_csv("https://raw.githubusercontent.com/kalilurrahman/datasets/main/state%20wise%20centroids_2011.csv")
state_wise_census = district_wise_census.groupby(by="State").sum()[["Population in 2001","Population in 2011"]].reset_index()
state_wise_census = state_wise_centroid.merge(state_wise_census, left_on="State", right_on="State")
#state_wise_census.head()
#district_wise_census.head()
tn_districts = district_wise_census[district_wise_census["State"]=="Tamil Nadu"]
center_lat = tn_districts.mean().Latitude
center_long = tn_districts.mean().Longitude
TNMap = folium.Map(location=[center_lat, center_long], zoom_start=6.5, tiles="stamenterrain",scrollWheelZoom=False)
tiles = ['stamenwatercolor', 'cartodbpositron', 'openstreetmap', 'stamenterrain','stamentoner']
for tile in tiles:
    folium.TileLayer(tile).add_to(TNMap)
folium.LayerControl().add_to(TNMap)
for district in tn_districts["District"].unique():
    district_census = tn_districts[tn_districts["District"]==district]
    str_2011_Pop = district_census["Population in 2011"].values[0]
    str_2001_Pop = district_census["Population in 2001"].values[0]     
    my_string = 'Population 2011 : :<b> {} </b>, <br> Population 2001 :<b> {} </b> <br>'.format(str_2011_Pop,str_2001_Pop)
    tootip_str = " District <b>" + district_census.District.values[0] + "</b><br>" + my_string
    folium.Marker(
        location=[district_census.Latitude.values[0], district_census.Longitude.values[0]],
        popup=my_string,
        tooltip = tootip_str,
        icon=folium.Icon(icon='info-sign', color="red"),
        draggable=False
    ).add_to(TNMap)
for district in tn_districts["District"].unique():
    district_census = tn_districts[tn_districts["District"]==district]
    str_2011_Pop = district_census["Population in 2011"].values[0]
    str_2001_Pop = district_census["Population in 2001"].values[0]     
    my_string = 'Population 2011 : :<b> {} </b>, <br> Population 2001 :<b> {} </b> <br>'.format(str_2011_Pop,str_2001_Pop)
    tootip_str = " District <b>" + district_census.District.values[0] + "</b><br>" + my_string
    folium.Marker(
        location=[district_census.Latitude.values[0], district_census.Longitude.values[0]],
        popup=my_string,
        tooltip = tootip_str,
        icon=folium.Icon(icon='cloud', color="blue")
    ).add_to(TNMap)
folium.Polygon(
    locations=list(zip(tn_districts.Latitude, tn_districts.Longitude))[1:22],
    popup="Path",
    tooltip = "Path",
    color="violet",
    fill_color="green"
).add_to(TNMap)
folium.Polygon(
    locations=list(zip(tn_districts.Latitude, tn_districts.Longitude))[1:12],
    popup="Path",
    tooltip = "Path",
    color="black",
    fill_color="yellow"
).add_to(TNMap)
TNMap
```
