## 클래스 활용하여 예제1,2,3 동시관리

```python
class STOOOP: 
    title = '학교폭력'
    url = 'https://github.com/guebin/IP2022/blob/master/_notebooks/2022-05-07-stop1.jpeg?raw=true'
    end = '멈춰~~~~' 
    def stop(self): ## 규칙1: class안에 정의된 함수의 첫번째 인자는 무조건 self 
        htmlstr = "<p> {title} </p> <img src='{url}'> <p style='color:black;'> {end} </p>" ## 규칙2: class안에서 정의된 변수를 쓰려면 self.변수이름 으로 써야함 
        display(HTML(htmlstr.format(title=self.title,url=self.url,end=self.end)))    
```

### 클래스 사용법 (암기)

`-` 선언하는 방법 (=양식문서를 만드는 방법) 

```python
class STOOOP: ## 양식 이름 설정 
    title = '학교폭력' ## 양식에 포함된 변수1 
    url = 'https://github.com/guebin/IP2022/blob/master/_notebooks/2022-05-07-stop1.jpeg?raw=true' ## 양식에 포함된 변수2 
    end = '멈춰~~~~' ## 양식에 포함된 변수3 
    def stop(self): ## 양식에 포함된 함수1 = 메소드1 
        htmlstr = "<p> {title} </p> <img src='{url}'> <p style='color:black;'> {end} </p>" 
        display(HTML(htmlstr.format(title=self.title,url=self.url,end=self.end)))    
```

- 규칙1: 메소드(=class안에 정의된 함수)의 첫번째 인자는 무조건 self 
- 규칙2: 메소드에서 class안에서 정의된 변수를 사용하려면 self.변수이름과 같은 꼴로 쓴다. 
    - 즉 self.start, self.img, self.end 와 같은 방식으로 써야한다. 
    - 가끔씩 self 자리에 KBSHWP.start, KBSHWP.img, KBSHWP.end 와 같이 클래스 이름을 쓰기도 한다. 

`-` 성능1 사용법: 클래스에서 인스턴스를 생성하는 방법

```python
stop1 = STOOOP() 
stop2 = STOOOP() 
```

- 함수사용법과 비슷
- 클래스 이름을 쓰고 콘텐츠를 구체화시키는 과정에서 필요한 입력1, 입력2를 ()에 넣는다. 
- 현재는 따로 입력이 없으므로, 그냥 KBSHWP 이후에 입력을 비워둔다. 즉 `KBSHWP()`로 생성

`-` 성능2 사용법: 클래스내에서 정의된 변수의 값을 확인하는 방법, 바꾸는 방법
