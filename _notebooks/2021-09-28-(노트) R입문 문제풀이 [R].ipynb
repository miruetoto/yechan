### 2004(인)/수능(홀) 19(고3)

자료 $x_1,x_2,\dots,x_{100}$에 대하여 다음 과정을 순서대로 시행하였다. 

(가) 처음 두 수 $x_1$과 $x_2$의 평균을 구한다. 

(나) $x_3$을 추가하여 $x_1,x_2,x_3$의 평균을 구한다. 

(다) $x_4$를 추가하여 $x_1,x_2,x_3,x_4$의 평균을 구한다. 

...

$x_{100}$을 추가하여 $x_1,x_2,\dots,x_{100}$의 평균을 구한다. 

위의 과정을 시행할 결과, $x_1$과 $x_5$의 평균이 5이고, 자료 하나가 추가될때 마다 평균이 1씩 증가하였다. 이때 $x_{100}$의 값은? 

(1) 194

(2) 196

(3) 198

(4) 200 

(5) 202 

(풀이)

```r
x<-c()
x[1]=5
x[2]=5
for(n in 3:100){
    temp_ = mean(x[1:(n-1)])+1 # 1~ (n-1)의 평균 + 1 = temp =1~n 의 평균 
    sum_ = temp_ *n # 1~n의 합 
    x[n] = sum_ -sum(x[1:(n-1)])
}
```

```r
x
```

```r
x[100]
```

### 2010(나)/수능(홀) 30(고3) -- G19

수열 $\{a_n\}$에 대하여 첫째항부터 제 $n$항까지의 합을 $S_n$이라 하자. 수열 $\{S_{2n-1}\}$은 공차가 -3인 등차수열이고, 수열 $\{S_{2n}\}$은 공차가 2인 등차수열이다. $a_2=1$일 때, $a_8$의 값을 구하시오. 

(풀이) 

```r
a<-c()
a[1]=0 # 그냥 0이라고 하자.. 틀리면 뭐.. 
a[2]=1 
```

```r
(1:100) %% 2 # (x %% 2) x를 2로 나누었을때 나머지를 구하는 함수이다. 
```

- 홀수이면 1, 짝수이면 0 이다. 

```r
S<-c()
S[1]=a[1]
S[2]=a[1]+a[2]
for(n in 3:100){
    if(n%%2 == 1){
        S[n]=S[n-2]-3
    }else{
        S[n]=S[n-2]+2
    }
}
```

```r
S[8]-S[7]
```

### 2007(나) 9월/평가원 16(고3) -- G178 

자연수 $n$에 대하여 점 $P_n$을 다음 규칙에 따라 정한다. 

(가) 점 $P_1$의 좌표는 $(1,1)$이다. 

(나) 점 $P_n$의 좌표가 $(a,b)$일 때, 
 - $b<2^a$이면 점 $P_{n+1}$의 좌표는 $(a,b+1)$이고 
 - $b=2^a$이면 점 $P_{n+1}$의 좌표는 $(a+b,1)$이다. 

점 $P_n$의 좌표가 $(10,2^{10})$일 때, $n$의 값은? 

(1) $2^{10}-2$ 

(2) $2^{10}+2$

(3) $2^{11}-2$

(4) $2^{11}$

(5) $2^{11}+2$

(풀이) for, if, else, break, ==, & 

```r
a=c()
b=c()
a[1]=1
b[1]=1
for(n in 1:2**12){
    if ((a[n]==10)&(b[n]==2^10)){
        print(n)
        break
    }
    if (b[n]<2^(a[n])){
        a[n+1]=a[n]
        b[n+1]=b[n]+1
    }else{
        if(b[n]==2^(a[n])){      
        a[n+1]=a[n]+1
        b[n+1]=1
        }
    }
}
```

```r
c(2^10-2, 2^10+2, 2^11-2, 2^11, 2^11+2)
```
