### numpy 

```python
def show(a):
    print('*id: %s' % id(a)) 
    print('*type: %s' % type(a))
    print('*base:')    
    print(a.base)
    print('*base id: %s' %id(a.base))    
    #print('\n')
    print('*value:')    
    print(a)
    #print('\n')
```

### 참조 

`-` a를 선언, b는 a의 참조 

```python
a=np.array([[1,2],[3,4]])
b=a ## 이거 shallow copy 아님. 인터넷에 자꾸 shallow copy라고 주장하는 사람들 있음. 
```

`-` shape을 조사하자 

```python
a.shape
```

```python
b.shape
```

`-` shape의 모양을 직접확인: (2,2) shape으로 잘 정리되어 있음 

```python
a
```

```python
b
```

`-` a의 shape을 바꾸면? $\to$ b도 따라서 바뀐다. 

```python
a.shape=(4,)
```

```python
a
```

```python
b
```

### view 

`-` a를 선언, b는 a의 view. (b는 a의 shallow copy 라고 부르기도 함) 

```python
a=np.array([[1,2],[3,4]])
b=a.view() ## 이게 shallow copy 
```

`-` a,b 의 shape확인 

```python
a
```

```python
b
```

`-` a의 shape을 바꿔도 b는 바뀌지 않음. 

```python
a.shape=(4,)
```

```python
a
```

```python
b
```

`-` 하지만 값을 바꾸면 같이 바뀐다. 

```python
a[0]=100
```

```python
a
```

```python
b
```

### copy 

`-` a를 선언, b는 a의 카피 

```python
a=np.array([[1,2],[3,4]])
b=a.copy()
```

`-` a의 shape을 바꿔도 b는 적용되지 않음 (이건 당연하죠) 

```python
a.shape = (4,)
```

```python
a
```

```python
b
```

`-` 그리고 a[0]의 값을 바꿔도 b에는 적용되지 않음 

```python
a[0]=100
```

```python
a
```

```python
b
```
