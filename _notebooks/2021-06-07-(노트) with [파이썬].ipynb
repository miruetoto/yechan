### with

#### motivation

`-` 예제2

생각해보니까 파일은 항상 열고 닫아야한다. 

이처럼 쌍으로 수행되는 처리는 반복적으로 해야하는데 그때마다 .close() 메소드따위를 쓰는것이 번거롭게 생각된다. 

파일을 열었으면 적당한 동작뒤에 알아서 닫아졌으면 좋겠다. 

이걸 구현한것이 with 문이다. 

````python
with open('2021-06-07-test.txt') as g:
    a=g.read()
    print(a)
````

파일이 닫아졌는지 확인해보자. 

````python
g.closed
````

잘 닫아졌다. 

#### 기본사용법

사실 코드는 직관적으로 이해가 된다. 

```python
with blabla [as variable]: 
    yadiyadi
    yadiyadi2
```

(1) `with blabla as` 의 blabla 가 실행된다.

(2) blabla 의 실행결과로 어떤 `특별한` 오브젝트가 만들어지는데 그 오브젝트를 g로 저장한다. 

(3) 탭으로 들여쓰기된 부분, 즉 yadiyadi, yadiyadi2가 순서대로 실행된다. 

(4) 탭으로 들여쓰기된 부분이 실행되고 난 뒤에 `g.closed()`따위의 미리 약속된 어떤 코드가 실행되는것 같다.  

#### 동작원리 

동작원리는 무엇일까? 

우선 g가 특별한 오브젝트라고 했는데 무엇이 특별한지 살펴보자. 

````python
dir(g)
````

비밀은 `__enter__` 와 `__exit__`메소드에 있다. 

아래 코드를 다시 관찰하자. 
```python 
with open('test.txt') as g:
    a=g.read()
    print(a)
```
- `for i in ...` 에서 ... 에 올 수 있는 오브젝트는 `__iter__` 메소드가 정의되어야 한다. 이러한 오브젝트를 iterable 오브젝트라고 한다. 
- 그와 유사하게 `with ...` 에서 ... 에 올 수 있는 오브젝트에는 `__enter__`, `__exit__` 메소드가 정의되어야 한다. 이러한 오브젝트를 context manager라고 한다.  
- with 문이 시작되면 `__enter__`가 실행된다. 
- 그 이후 들여쓰기 된 부분이 실행된다. 
- with 문이 끝나면 `__exit__`가 실행된다. 

`-` 예제3

````python
class MooYaHo: 
    def __enter__(self):
        print('무야호')
    def __exit__(self, exc_type, exc_value, traceback): # self 이외의 변수는 예외처리 관련 변수인데 여기서는 다루지 않음. 
        print('그만큼 신나시는 거지')
````

````python
with MooYaHo() as a:
    print(".")
````

`-` 예제3 변형

`__init__`과 `__enter__`중 무엇이 더 먼저 실행될까? 

````python
class MooYaHo: 
    def __init__(self):
        print('init')
    def __enter__(self):
        print('무야호')
    def __exit__(self, exc_type, exc_value, traceback): # self 이외의 변수는 예외처리 관련 변수인데 여기서는 다루지 않음. 
        print('그만큼 신나시는 거지')
````

````python
with MooYaHo() as a:
    print(".")
````

`-` 예제3 변형2

경우에 따라서 as 이하를 생략할수도 있다.

````python
with MooYaHo():
    print(".")
````
