### 예제

$X_1 \sim Ber(\theta)$ 

$T=X_1$은 CS이다. 

(pf) 

$\forall g \in {\cal M}$ 에 대하여 아래를 보이면 된다. 

$E_\theta\big[g(X_1)\big]\overset{\forall \theta}{=}0 \Longrightarrow P_{\theta}\big(\big\{g(X_1)=0\big\}\big)\overset{\forall \theta}{=}1$

$E_{\theta}\big[g(X_1)\big] \overset{\forall \theta}{=} 0$


$\Longleftrightarrow \begin{bmatrix} g(0) & g(1) \end{bmatrix} \begin{bmatrix} P_{\theta}(X_1=0) \\ P_{\theta}(X_1=1) \end{bmatrix} \overset{\forall\theta}{=} 0$
