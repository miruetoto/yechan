### 숫자 카드 게임 (2019 국가 교육기관 코딩테스트) 

숫자 카드 게임은 여러개의 숫자중에서 가장 높은 숫자가 쓰인 카드를 한장 뽑는 게임이다. 

단, 게임의 룰을 지키며 카드를 뽑아야 하고 룰은 다음과 같다. 

- 숫자가 쓰인 카드들이 N$\times $M 형태로 놓여 있다. 
- 먼저 뽑고자 하는 카드가 포함되어 있는 행을 선택한다. 
- 그 다음 선택된 행에 포함된 카드들 중에서 가장 숫자가 낮은 카드를 뽑아야 한다. 
- 따라서 처음에 카드를 골라낼 행을 선택할때, 이후에 해당 행에서 가장 숫자가 낮은 카드를 뽑을 것을 고려하여 최종적으로 가장 높은 숫자의 카드를 뽑을 수 있도록 전략을 세워야 한다. 

`-` 예시1

```
입력
3 1 2 
4 1 4 
2 2 2 

출력 
2
```

`-` 예시2

```
입력
7 3 1 8 
3 3 3 4 

출력
3
```

(풀이) 

(전략회의)

````r
a1=c(3,1,2) 
a2=c(4,1,4)
a3=c(2,2,2)
````

우리는 일단 아래와 같은 모양을 만들어야함 

```
3 1 2
4 1 4 
2 2 2 
```

a1,a2,a3을 각각 row-vector 로 해석하고 bind 하면된다. 

````r
A = rbind(a1,a2,a3)
````

````r
A
````

````r
A= rbind(c(3,1,2),c(4,1,4),c(2,2,2))
````

입력이 A일경우 출력이 2가 나와야한다. 

전략: 
1. A의 각 행(row)을 돌아가면서 가장 작은 숫자를 찾아 기록한다. 
2. 기록한 숫자들을 모아서 벡터로 만든다. 
3. 벡터의 최대값을 구한다. 

할수있어야 하는것 

- A의 1번째 행, 2번째행, 3번째행을 순서대로 뽑을수 있어야 한다. 

(실질적풀이)

````r
A= rbind(c(3,1,2),c(4,1,4),c(2,2,2))
````

````r
vec<-c()
for (i in 1:3) vec[i]<- min(A[i,])
````

````r
max(vec)
````

`-` 입력이 아래와 같은 경우 

```
7 3 1 8 
3 3 3 4
```

````r
A=rbind(c(7,3,1,8),c(3,3,3,4)) 
````

````r
vec<-c()
for (i in 1:2) vec[i]<- min(A[i,])
````

````r
max(vec)
````

### 카드뽑기 풀이에서 우리가 얻을 것

#### 매트릭스를 선언하는 방법 

`-` 방법1: 벡터를 만들고 그것을 row-vector로 해석 $\to$ bind 

````r
rbind(c(1,2,3,4),c(2,2,3,4))
````

`-` 방법2: 벡터를 만들고 그것을 col-vector로 해석 $\to$ bind 

````r
cbind(c(1,2,3,4),c(2,2,3,4))
````

`-` 일단 이정도만.. 

#### 매트릭스의 원소를 뽑는 방법 

`-` 아래와 같은 매트릭스가 있다고 하자. 

````r
A=rbind(c(1,2,3,4),c(5,6,7,8),c(9,10,11,12))
A
````

`-` 1열을 뽑고싶다. 

````r
A[,1]
````

`-` 1,4열을 뽑고싶다. 

````r
A[,c(1,4)]
````

`-` 2행을 뽑고싶다. 

````r
A[2,]
````

`-` 1,2행을 뽑고싶다. 

````r
A[c(1,2),]
````

`-` 2행3열의 원소 즉 (2,3)에 위치한 원소를 뽑고싶다. 

````r
A[2,3]
````

`-` 일단 이정도만.. 

#### 매트릭스를 왜 써야하는가? 

`-` 바둑판구조의 자료형의 표현하기 위해서 = 자료를 바둑판 구조로 저장하기 위해서 

`-` 자료를 바둑판(2d-array), 나아가 큐브형태(3d-array)로 저장하면 여러가지 이점이 있다.  

`-` 결국 벡터(1d-array)를 쓰는 이유와 같음. 자료를 일자로 배열하여 저장하면 자료를 관리하기 편하거나 계산상의 이점이 있는 경우가 많은데 매트릭스 역시 그러하다. 

(예시) 벡터의 계산상 이점: 초항이 1이고 공차가 2인 등차수열을 100개 만들어라. 

````r
a<-c()
for (n in 1:100) a[n]<- 2*n-1 
print(a)
````

위와 같은 간단한 코드도 결국 벡터형태로 자료를 저장하고 관리하기 때문에 for를 쓸 수 있는 것

혹은 아래와 같이 구할 수도 있는데 (브로드캐스팅) 이 역시 벡터라는 개념이 있기에 가능함 

````r
n=1:100
2*n-1
````

이는 벡터를 이용하여 2*n-1 이라는 단순한 수식을 벡터버전으로 확장해석한것이다. 계산상 이점을 극대화함. (수식적으로 간단 + 구현이 간단 + 빠름!!) 

#### apply 함수

`-` 우선 아래를 기억하자. 

```
apply(매트릭스,1,특정함수) 
```

를 쓰면 어떠한 특정함수가 각각의 row에 적용되어서 결과가 array로 저장된다. 

`-` 여기에서 특정함수는 "벡터입력-스칼라출력" 이라 생각하자. 즉 벡터입력-스칼라출력인 함수만 apply가 가능하다고 생각하자. (사실그렇지는 않음) 예를들면 아래와 같다. 

````r
sum(c(1,2,3,4))
````

````r
mean(c(1,2,3,4))
````

`-` 아래와 같은 함수는 벡터입력-벡터출력 이므로 조건에 해당하지 않는다. 

````r
sqrt(c(1,2,3,4))
````

`-` 이제 매트릭스 A에 다양한 함수를 적용해보자. 

````r
A
````

````r
apply(A,1,sum)
````

````r
apply(A,1,mean)
````

````r
apply(A,1,min)
````

````r
apply(A,1,max)
````

사용자가 정의한 함수도 가능하다. 

````r
f = function(a) (min(a)+1)**2-3
````

````r
apply(A,1,f)
````

`-` 숫자1을 2로 바꾸면 특정함수가 각각의 col에 적용되어서 결과가 array로 저장된다. 

````r
A
````

````r
apply(A,2,sum)
````

````r
apply(A,2,mean)
````

````r
apply(A,2,min)
````

````r
apply(A,2,max)
````

````r
apply(A,2,f)
````

`-` 아래도 같은 코드이다. 

````r
myfun = function(x,y) x**2+y
````

````r
myfun(2,3)
````

````r
2 %>% myfun(3)
````

````r
myfun(3,2)
````

````r
3 %>% myfun(2)
````

`-` 아래도 같은 코드이다. 

````r
myfun2 = function(x,y,z) x**2 + y*z
````

````r
myfun2(2,1,1)
````

````r
2 %>% myfun2(1,1)
````

`-` 왜 이러한 일을 해야하는가? (합성함수 처리에 유리)

````r
2:5 %>% log %>% sqrt %>% sum %>% myfun2(1,1) %>% sqrt
````

````r
sqrt(myfun2(sum(sqrt(log(2:5))),1,1))
````

````r
2:5 %>% log %>% myfun(3) %>% sqrt %>% sum %>% myfun2(1,1) %>% sqrt
````

````r
sqrt(myfun2(sum(sqrt(myfun(log(2:5),3))),1,1))
````

`-` 아래의 개념차이

- $h(g(f(x)))$
- $x \to f \to g \to h$

`-` 큰 매트릭스 만듬 

````r
A=sample(1:9,size=100*2000000,replace=TRUE)
````

````r
dim(A)<-c(2000000,100)
A 
````

````r
t1=Sys.time()
A %>% apply(1,min) %>% max 
t2=Sys.time()
t2-t1
````

````r
t1=Sys.time()
vec<-c()
for(i in 1:2000000) vec[i]<-min(A[i,])
max(vec)
t2=Sys.time()
t2-t1
````

`-` for가 좀 더 빠르지만 코드는 apply가 더 간결해보인다.
