(풀이1)

```r
a=c(1,3,2,3,2)
```

```r
count=0
for (i in 1:5){
    for (j in 1:5){
        if (a[i]!=a[j]) count=count+1
    }
}
count
```

- ? 중복이 계산되어서 이상한 값이 나왔다.. 

`-` 아래와 같이 수정하면 된다. 

```r
count=0
for (i in 1:5){
    for (j in 1:5){
        if ((a[i]!=a[j])&(i>j)) count=count+1
    }
}
count
```

`-` 아예 아래처럼 해도 괜찮다. 

```r
count=0
for (i in 1:5){
    for (j in i:5){
        if (a[i]!=a[j]) count=count+1
    }
}
count
```

```r
a=c(1,3,2,3,2)
A=rep(0,(length(a)**2)*2)
dim(A)=c(length(a)**2,2)
A
```

```r
count=0
k=1
for (i in 1:5){
    for (j in 1:5){
        A[k,]=c(a[i],a[j])
        k=k+1
    }
}
A
```

```r
vec1<-c()
vec2<-c()
for(i in 1:25){
    vec1[i]<- A[i,1]-A[i,2]
    vec2[i]<- A[i,1]>A[i,2]
}
```

```r
vec1
```

```r
(vec1!=0) & (vec2==TRUE)
```
