### 볼링공 문제 (2019 SW마에스트로 입학 테스트)

A,B 두 사람이 볼링을 치고 있습니다. 두 사람은 서로 무게가 다른 볼링공을 고르려고 합니다. 볼링공은 총 N개가 있으며 각 볼링공마다 무게가 적혀 있고, 공의 번호는 1번부터 순서대로 부여됩니다. 또한 같은 무게의 공이 여러개 있을 수 있지만, 서로 다른 공으로 간주합니다. 볼링공의 무게는 1부터 M까지의 자연수 형태로 존재합니다. 

예를들어 N이 5이고, M이 3이며 각각의 무게가 차례대로 1,3,2,3,2일 때 각 공의 번호가 차례대로 1번부터 5번까지 부여됩니다. 이때 두 사람이 고를 수 있는 볼링공 번호의 조합을 구하면 다음과 같습니다. 

(1번,2번), (1번,3번), (1번,4번), (1번,5번), (2번,3번), (2번,5번), (3번,4번), (4번,5번)

결과적으로 두 사람이 공을 고르는 경우의 수는 8가지입니다. N개의 공의 무게가 각각 주어질 때, 두 사람이 볼링공을 고르는 경우의 수를 구하는 프로그램을 작성하세요. 

#### 문제해설 

> $M$개의 볼링공의 무게를 각각 $x_1,\dots,x_M$ 이라고 하자. $M$개의 공중에서 무게가 서로 다른 2개 공을 뽑는 조합의 수를 구하는 알고리즘을 구현하라. 

> 예시: $(x_1,\dots,x_5)=(1,3,2,3,2)$ 일 경우 가능한 조합의 집합은 아래와 같다. 
$(x_1,x_2), (x_1,x_3),(x_1,x_4), (x_1,x_5),(x_2,x_3), (x_2,x_5), (x_3,x_4), (x_4,x_5)$

### 풀이1

```r
a=c(1,3,2,3,2)
```

```r
a
```

```r
count=0
for (i in 1:5){
    for (j in 1:5){
        if (a[i]!=a[j]) count=count+1
    }
}
count
```

- ? 중복이 계산되어서 이상한 값이 나왔다.. 

`-` 아래와 같이 수정하면 된다. 

```r
count=0
for (i in 1:5){
    for (j in 1:5){
        if ((a[i]!=a[j])&(j>i)) count=count+1
    }
}
count
```

### 풀이2

`-` `((a[i]!=a[j])&(j>i))` 에서 `j>i`는 for문의 선언부에서 처리할 수 있다. 

```r
count=0
for (i in 1:5){
    for (j in i:5){ ## index가 i부터 시작하도록 한다
        if (a[i]!=a[j]) count=count+1
    }
}
count
```

### 풀이3

```r
a=c(1,3,2,3,2)
A=rep(0,(length(a)**2)*2)
dim(A)=c(length(a)**2,2)
A
```

```r
k=1
for (i in 1:5){
    for (j in 1:5){
        A[k,]=c(a[i],a[j])
        k=k+1
    }
}
A
```

`-` 이중에서 (1) 첫번째열 = 두번째열 이면서 (2) 중복을 피하면 된다. 

```r
vec1<-c()
vec2<-c()
for(i in 1:25){
    vec1[i]<- A[i,1]-A[i,2] ## 첫번째열 = 두번째열 임을 파악하기 위해서 두 칼럼을 뺀다. (!=0이 되는 값을 체크하면됨) 
    vec2[i]<- A[i,1]>A[i,2] ## 중복을 피하기 위한 장치 
}
```

```r
vec1
```

```r
vec2
```

```r
(vec1!=0) & (vec2==TRUE)
```

```r
sum((vec1!=0) & (vec2==TRUE))
```

```r
A %>% mutate(C=Var1-Var2) # 
```

- mutate(): 데이터프레임에서 새로운 column 을 추가하는 기능 (위의 기존의 컬럼을 활용하여 새로운 파생컬럼을 만들 수 있음)
