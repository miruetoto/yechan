```r
count=0
for (i in 1:5){
    for (j in 1:5){
        if (a[i]!=a[j]) count=count+1
    }
}
count
```

- ? 중복이 계산되어서 이상한 값이 나왔다.. 

`-` 아래와 같이 수정하면 된다. 

```r
count=0
for (i in 1:5){
    for (j in 1:5){
        if ((a[i]!=a[j])&(i>j)) count=count+1
    }
}
count
```

### 풀이3

```r
a=c(1,3,2,3,2)
A=rep(0,(length(a)**2)*2)
dim(A)=c(length(a)**2,2)
A
```

```r
k=1
for (i in 1:5){
    for (j in 1:5){
        A[k,]=c(a[i],a[j])
        k=k+1
    }
}
A
```

`-` 이중에서 (1) 첫번째열 = 두번째열 이면서 (2) 중복을 피하면 된다. 

```r
vec1<-c()
vec2<-c()
for(i in 1:25){
    vec1[i]<- A[i,1]-A[i,2] ## 첫번째열 = 두번째열 임을 파악하기 위해서 두 칼럼을 뺀다. (!=0이 되는 값을 체크하면됨) 
    vec2[i]<- A[i,1]>A[i,2] ## 중복을 피하기 위한 장치 
}
```

```r
vec1
```

```r
vec2
```

```r
(vec1!=0) & (vec2==TRUE)
```

```r
sum((vec1!=0) & (vec2==TRUE))
```

### 풀이4 

`-` vec1,vec2를 꼭 for문을 사용해서 구할필요는 없음. 

```r
a=c(1,3,2,3,2)
A=rep(0,(length(a)**2)*2)
dim(A)=c(length(a)**2,2)
k=1
for (i in 1:5){
    for (j in 1:5){
        A[k,]=c(a[i],a[j])
        k=k+1
    }
}
A
```

```r
A[,1]-A[,2] # vec1 
```

```r
A[,1]>A[,2] # vec2
```

```r
(A[,1]-A[,2] != 0) & (A[,1]>A[,2])
```

```r
((A[,1]-A[,2] != 0) & (A[,1]>A[,2])) %>% sum
```
