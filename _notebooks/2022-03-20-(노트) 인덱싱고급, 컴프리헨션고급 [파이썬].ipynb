### 컴프리헨션 고급 (if문이 포함된 컴프리헨션)

`-` 예제: 제곱수중에서 12로 나누어 떨어지는 수만 만들고 싶다. 
- 제곱수: 1,4,9,16,25,...
- 12로 나누어 떨어지는 수

(풀이1)

```python
lst = [] 
for i in range(1,101): 
    if (i**2 % 12 ==0):
        lst.append(i)    
```

```python
lst
```

(풀이2)

```python
[i for i in range(1,101) if i**2 % 12 ==0] 
```

(풀이3)

```python
list(range(101))[6::6] # 위의 풀이에 비해서 가독성도 별로고 딱히 더 짧은것도 아님
```

### 함수고급 (조건부리턴)

```python
def test(a):
    if a % 2 == 0:
        return 'even'
    else:
        return 'odd'
```

```python
[test(a) for a in range(1,11)]
```

```python
def test(a): 
    return 'even' if a%2 ==0 else 'odd'
```

```python
[test(a) for a in range(1,11)]
```

### len함수 

`-` 0차원 자료형은 len함수가 동작하지 않음 

```python
a=1 
len(a)
```

```python
a=True
len(a)
```

```python
a=3.14
len(a)
```

> note: 이것이 어떠한 수학적인 의미를 가지거나 0차원의 본질적진리를 뜻하는건 아님. R에서는 1,3.14,TRUE등의 길이가 1로 존재함

`-` 1차원 자료형은 len함수가 동작

```python
a='guebin'
len(a)
```

```python
a=[1,2,3,4,5,6]
len(a)
```

```python
a=range(10)
len(a)
```

```python
a=1,2,3,4,5,6
len(a)
```

```python
a={1,2,3,4,5,6}
len(a)
```

```python
a={'a':1,'b':2}
len(a)
```

`-` 길이가 1인 1차원 자료형과 0차원 자료형은 다른것임 

```python
a='a'
len(a)
```

```python
a=[1]
len(a)
```
