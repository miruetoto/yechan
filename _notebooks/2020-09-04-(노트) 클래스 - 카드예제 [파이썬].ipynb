### 예제2 

`-` 나도 클래스생성자를 만들어보자. `FrenchDeck` 이라는 클래스생성자를 만들것이다. 

#### `FrenchDeck`이라는 클래스생성자를 정의하기전 사전코드이해 

```python
ranks=[str(n) for n in range(2,11)]
ranks
```

```python
suits='spades diamond clubs heaters'.split()
suits
```

```python
[Card(rank,suit) for suit in suits for rank in ranks]
```

#### `FrenchDeck` 클래스생성자 정의

```python
class FrenchDeck:
    ranks = [str(n) for n in range(2,11)]
    suits = 'spades diamond clubs heaters'.split()
    
    def __init__(self):
        self._cards = [Card(rank,suit) for suit in suits for rank in ranks]
    def __len__(self):
        return len(self._cards)
    def __getitem__(self,position):
        return self._cards[position]
```

`-` `FrenchDeck`라는 클래스생성자를 활용하여 인스턴스를 생성하고 그것을 `deck`에 저장하라. 

```python
deck = FrenchDeck()
```

`-` `__len__`을 정의한 덕분에 `deck`인스턴스에 아래와 같은 기능을 활용할 수 있다. 

```python
len(deck)
```

`-` `__getitem__`을 정의한 덕분에 `deck`인스턴스에 아래와 같은 기능을 활용할 수 있다. 

```python
deck[0]
```

```python
deck[-1]
```

```python
deck[:3]
```

```python
deck[12::13]
```

```python
from random import choice 
choice(deck)
```

```python
for i in deck: 
    print(i)
```

```python
for i in reversed(deck): 
    print(i)
```

`-`  `FrenchDeck` 클래스생성자를 조금 변형해서 새로운 클래스생성자를 정의하고 싶다. 
- 새로운 클래스 생성자의 이름은 `FrenchDeck_new`라고 하자. 
- 새로운 기능으로 `say`를 넣자. 

```python
class FrenchDeck_new(FrenchDeck):
    def say(self):
        print ("나에게는 {}개의 rank와 {}개의 suits로 조합된 {}개의 카드가 있어!".format(len(self.ranks),len(self.suits),len(self._cards))) 
```

```python
deck_new = FrenchDeck_new()
```
