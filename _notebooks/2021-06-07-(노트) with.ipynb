### with

`-` 예제2

생각해보니까 파일은 항상 열고 닫아야한다. 

이처럼 쌍으로 수행되는 처리는 반복적으로 해야하는데 그때마다 .close() 메소드를 쓰는것이 번거롭게 생각된다. 

파일을 열었으면 적당한 동작뒤에 알아서 닫아졌으면 좋겠다. 

이걸 구현한것이 with 문이다. 

````python
with open('ss.txt') as g:
    a=g.read()
    print(a)
````

````python
g.closed
````

사용법은 직관적으로 이해가 된다 .

(1) `with ... as` 의 `...` 이 실행된다.

(2) `...` 의 실행결과로 만들어지는 오브젝트를 g로 저장한다. 

(3) 탭으로 들여쓰기된 부분이 실행된다. 

(4) 탭으로 들여쓰기된 부분이 실행되고 난 뒤에 g.closed()가 실행된다. 

#### 동작원리 

동작원리는 무엇일까? 

````python
dir(g)
````

비밀은 `__enter__` 와 `__exit__`메소드에 있다. 

아래 코드를 다시 관찰하자. 
```python 
with open('ss.txt') as g:
    a=g.read()
    print(a)
```
- `for i in ...` 에서 ... 에 올 수 있는 오브젝트는 `__iter__` 메소드가 정의되어야 한다.  
- 그와 유사하게 `with ...` 에서 ... 에 올 수 있는 오브젝트에는 `__enter__`, `__exit__` 메소드가 정의되어야 한다.  
- with 문이 시작되면 `__enter__`가 실행된다. 
- 그 이후 들여쓰기 된 부분이 실행된다. 
- with 문이 끝나면 `__exit__`가 실행된다. 

`-` 예제3

````python
class MooYaHo: 
    def __enter__(self):
        print('무야호')
    def __exit__(self, exc_type, exc_value, traceback):
        print('그만큼 신나시는 거지')
````

````python
with MooYaHo() as a:
    print(".")
````
