### 유령버스 

```python
class Bus:
    def __init__(self,passengers=[]):
        self.passengers = passengers
    def pick(self, name):
        self.passengers.append(name)
    def drop(self, name):
        self.passengers.remove(name)
```

`-` 버스1

```python
bus1=Bus(['규빈','하영'])
```

```python
bus1.passengers
```

```python
bus1.pick("재인")
```

```python
bus1.passengers
```

```python
bus1.drop("규빈")
```

```python
bus1.passengers
```
