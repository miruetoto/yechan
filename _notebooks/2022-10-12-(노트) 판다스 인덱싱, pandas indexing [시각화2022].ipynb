### 데이터프레임 준비

`-` 데이터준비

### 컨셉1~4 정리 

`-` 표를 만들어 볼까..?

||`.`| `[]` | `.iloc` | `.loc`| 
|:-:|:-:|:-:|:-:|:-:|
|col선택가능?|O|O|O|O|
|row선택가능?|X|X|O|O|
|리턴값|ser|ser or df | ser or df | ser or df|
|단일레이블 입력가능?|O|O|O|O|
|레이블리스트 입력가능?|X|O|O|O|
|슬라이싱 입력가능?|X|X|O|O|

`-` col 이름을 알아야하는 경우 
- `.` : 앞글자만 대충 알아도 자동완성 가능 
- `[]`: 정확한 col 이름을 알아야 함
- `.loc`: 보통 정확한 col 이름을 알아야 하지만 슬라이싱 이용시 양 끝의 컬럼이름만 알면 무방 
- `.iloc`: 정확한 col 이름을 몰라도 번호로 인덱싱 가능
