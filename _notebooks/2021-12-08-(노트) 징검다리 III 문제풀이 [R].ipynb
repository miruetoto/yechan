```r
rslt <- c()
for(i in 1:1000000){
    x_=rbinom(5,size=1,0.8)
    x=sum(x_ * cumprod(x_))
    xx_ = 5-x-1  
    y=0 
    if(xx_ >0) y= rbinom(1, size=xx_, 0.5) 
    rslt[i]<-5+x+y
}
mean(rslt)
```
