### 정의

`-` stochastic matrix: row-sum이 1인 매트릭스
- 예: $\begin{bmatrix}1 & 0 & 0 \\ 1/3 & 1/3 & 1/3 \\ 1/3 & 1/3 & 1/3 \end{bmatrix}$

`-` matrix with identical rows (MwIR): 직전상태에 상관없이 다음상태에 있을 확률이 동일한 경우
- 예: $\begin{bmatrix}1/6 & 2/6 & 3/6 \\ 1/6 & 2/6 & 3/6 \\ 1/6 & 2/6 & 3/6 \end{bmatrix}$

`-` scrambling matrix: given any two rows, say $i$ and $i'$, there exists at least one column, $j$, such that both $p_{ij}>0$ and $p_{i'j}>0$. 
- 예: $\begin{bmatrix}0 & 1/2 & 1/2 \\ 1/2 & 0 & 1/2 \\ 1/2 & 1/2 & 0 \end{bmatrix}$

`-` $\{P\}=\min_{i,i'} \sum_j \min(p_{ij},p_{i'j})$
- $0 \leq \{P\} \leq 1$. 
- $P$가 MwIR. $\Longleftrightarrow \{P\}=1$. 
- $P$가 스크램블링 매트릭스 $\Longleftrightarrow \{P\}>0$.
- $P$가 스크램블링 매트릭스 아님 $\Longleftrightarrow \{P\}=0$.
- $P=\begin{bmatrix}1 & 0 & 0 \\ 1/3 & 1/3 & 1/3 \\ 1/3 & 1/3 & 1/3 \end{bmatrix}$ 이면 $\{P\}=1/3$이다. 
- $P=\begin{bmatrix}0 & 1 & 0 \\ 0 & 0 & 1 \\ 1 & 0 & 0 \end{bmatrix}$ 라면 $\{P\}=0$이다. 
- $P=\begin{bmatrix}0 & 1/2 & 1/2 \\ 1/2 & 0 & 1/2 \\ 1/2 & 1/2 & 0 \end{bmatrix}$ 라면 $\{P\}=1/2$이다. 

### 약속 

`-` 논문에 등장하는 모든 매트릭스는 stochastic matrix이다. 

### 레마1

$Q=LM \Longrightarrow \{Q\} \geq \{L\}$.

`proof`

$\{L\}=0$일 경우에는 증명이 자명하므로 생략한다. $\{L\}>0$이라고 가정하자. 

편의상 아래를 가정하자. (아래와 같이 가정하여도 일반성을 잃지 않는다.) 

$\{Q\}=\min_{i,i'}\sum_{j}\min(q_{ij},q_{i'j})=\min(q_{1j},q_{2j})$

매트릭스 곱셈의 정의에 의하여 

$q_{ij}=\sum_{k}l_{ik}m_{kj}$

이 성립한다. 따라서 

$q_{1j}=\sum_{k}l_{1k}m_{kj}$

$q_{2j}=\sum_{k}l_{2k}m_{kj}$

이다. 따라서 아래가 성립한다. 

$\min(q_{1j},q_{2j}) = \min \big(\sum_{k}l_{1k}m_{kj}, \sum_{k}l_{2k}m_{kj} \big) \geq \sum_{k}\min(l_{1k},l_{2k})m_{kj} $

양변에 $\Sigma_{j}$를 취하자. 

$LHS=\Sigma_j \min(q_{1j},q_{2j})=\{Q\}$

$RHS=\big(\Sigma_k \min(l_{1k},l_{2k})\big) \times \Sigma_{j} m_{kj}= \Sigma_k \min(l_{1k},l_{2k})\geq \{L\}$

$RHS$에서 두번째 등호는 $M$이 stochastic matrix이므로 성립한다. 마지막 부등호는 $\{L\}$의 정의에 의하여 성립한다. 

### 레마2

$L$과 $M$중 하나가 스크램블링 매트릭스이면 $Q=LM$역시 스크램블링 매트릭스이다. 

`proof` $L$이 스크램블링 매트릭스이면 $\{L\}>0$이다. 레마1을 쓰면 $\{Q\}\geq \{L\}$ 이므로 $\{Q\}>0$이다. 따라서 $Q$역시 스크램블링 매트릭스이다.
