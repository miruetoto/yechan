### 정의

### 이해 

`-` 당연한거 아닌가? 
- 그래디언트 기반의 학습(그래디언트 기반의 옵티마이저): 손실함수의 기울기를 통하여 업데이트 하는 방식
- 역전파: 손실함수의 기울기를 구하는 테크닉(체인룰 + $\alpha$). 구체적으로는 (1) 손실함수를 여러단계로 쪼개고 (2) 각 단계의 미분을 각각 구하고 (3) 그것들을 모두 곱하여 기울기를 계산한다. 
- 0근처의 숫자를 계속 곱하면 터지거나 0으로 간다. (사실 안정적인 기울기가 나올거라고 보는것이 이상한것임)

### data (시뮬레이션) 

```python
user = list(range(1,101))
item = list(range(1,21))
_lst = [[0,0,4.2]]
```

```python
def f(u,i): 
    if u<=50 and i<=10: 
        return np.random.randn(1)[0]*0.2+4 
    elif u<=50 and i>10:
        return np.random.randn(1)[0]*0.2+1 
    elif u>50 and i<=10:
        return np.random.randn(1)[0]*0.2+1 
    elif u>50 and i>10:
        return np.random.randn(1)[0]*0.2+4 
        
for u in user:
    _item = np.random.choice(list(range(1,21)),size=10,replace=False).tolist()
    for i in _item:
        _lst.append([u,i,f(u,i)])
```

### 예측? 

`-` None의 값을 추론하는 방법? 머리속으로는 이미 알고있음. 그런데 어떻게 수식화 할 수 있을 것인지? 

```python
df2.loc['user1']
```

`-` 유저의 취향은 예를들면 아래와 같이 표현할 수 있다. 
- user1의 취향 = (커피좋아함, 홍차싫어함) = (0.9,0.21)

`-` 숫자로 나타내면 아래와 같다. 

```python
user1_prprnc = (0.8, 0.21) 
user1_prprnc
```

`-` 여기에서 (0.8, 0.21)은 각각 (커피음료, 홍차음료)로 해석할 수 있다. 현재는 단순하게 원소가 2인 벡터이지만 다양한 아이템의 경우 아래와 같은 featrue도 가능하다. 
- 음료의특징: (커피음료, 홍차음료, 우유포함, 설탕포함정도, 아이스여부, 행사여부, 이벤트중인음료, ...) 

`-` 아이템1과 11의 특징은 예를들면 아래와 같이 표현할 수 있다. 

```python
item4_ftr = (0.71, 0.03) # 커피4
item11_ftr = (0.1, 0.88) # 홍차1
```

`-` 유저의 선호도와 아이템의 feature를 내적해보자. 

```python
np.array(user1_prprnc) @ np.array(item4_ftr)
```

```python
np.array(user1_prprnc) @ np.array(item11_ftr)
```

`-` 유저의 선호도와 item의 feature가 비슷할수록 값이 높게 나온다. 실제 user1이 커피4와 홍차1에 매긴 평점은 (3.935039,0.959677) 인데 이 값이 (0.574,0.256) 와 비슷하게 만들면 된다. 

`-` 적당히 값을 조정하여 다시 계산해보자.

```python
user1_prprnc = (0.8, 0.2) 
item4_ftr = (5, -0.5) # 커피4
item11_ftr = (-0.1, 5) # 홍차1
```

```python
np.array(user1_prprnc) @ np.array(item4_ftr)
```

```python
np.array(user1_prprnc) @ np.array(item11_ftr)
```

- 적당히 이런식으로?... 

`-` 따라서 `user1_prprnc`, `item4_ftr`, `item11_ftr`의 값들이 학습을 통하여 조정할 파라메터이다. 확장하면.. 
- `user1_prprnc`, ... , `user100_prprnc`
- `item1_ftr`, ... , `item20_ftr`

을 찾으면 된다. 

`-` 신기한것은 이 파라메터들이 있다면 `None`에 대한 추론도 가능하다는 것이다. 예를들어서 user1은 item1를 먹지 않았지만 (즉 커피1) item1_ftr의 `계수값`이 다른유저를 통하여 학습 되었고 그 값이 아래와 같다고 하자. 

```python
item1_ftr = (5.8, -0.17) 
```

그러면 커피1에 대하여 user1이 매길것이라 예상되는 평점은 아래와 같다. 

```python
np.array(user1_prprnc) @ np.array(item1_ftr)
```

`-` 생각해보니까 `user1_prprnc`, ... , `user100_prprnc`, `item1_ftr`, ... , `item20_ftr` 를 ${\bf W}$라고 생각하면 우리가 ${\bf W}$을 바꿈에 따라 예측값과 손실이 결정되므로 네트워크 구조가 만들어진다. 

`-` 정리하면 아래와 같은 적당한 매트릭스를 선택하여 

- ${\bf U}=\begin{bmatrix} {\tt user1~prprnc} \\ \dots \\ {\tt user100~prprnc} \end{bmatrix}$ $\quad 100 \times 2$ mat 

- ${\bf V}=\begin{bmatrix} {\tt item1~ftr} \\ \dots \\ {\tt item20~ftr} \end{bmatrix}$ $\quad 20 \times 2$ mat 

${\bf U} {\bf V}^\top \approx$`df2`가 만족하도록 하면 된다. 



### fastai를 이용한 학습: df $\to$ dls 

```python
dls=CollabDataLoaders.from_df(df,bs=100) # df의 형식: 첫번째열을 사용자인덱스, 두번째열은 아이템인덱스, 세번째열은 rating, 네번째열은 아이템이름으로 약속
```

```python
dls.items
```

```python
dls.show_batch()
```

### fastai를 이용한 학습: learn 

```python
lrnr = collab_learner(dls,n_factors=2,y_range=(0,5))
lrnr.fit(30,0.01)
```

```python
lrnr.show_results()
```

`-` 학습이 잘 되었다.
