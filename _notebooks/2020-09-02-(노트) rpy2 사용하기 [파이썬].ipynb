### R환경에 함수정의하기

```python
ro.r(''' my_fun<-function(x) x^2+1''')
ro.r['my_fun'](p2r([3,4,5]))
ro.r('my_fun')(p2r([3,4,5]))
ro.r.my_fun(p2r([3,4,5])) 
## 얼핏보면 이게 젤 쉬워보이지만 R함수이름에 .이 포함되면 사용할 수 없다는 단점이 있다. 
```

### R함수를 파이썬에서 사용 

```python
ro.r(''' my_fun<-function(x) x^2+1''')
my_fun_py_temp=ro.r.my_fun
def my_fun_py(a):
    rtn=r2p(my_fun_py_temp(p2r(a)))
    return rtn
my_fun_py([3,4,5])
```
