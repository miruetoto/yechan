### 자료

`-` $(x_i,y_i)$를 관측하였다. 

`-` $x_i$는 임의의 실수값을 가지는것 같은데 $y_i$는 0 혹은 1이다. 

`-` 자료를 대충 보니까 $x_i$가 커질수록 $y_i$는 $1$이 되는것 같다. 

### 모형1: $y_i=I(\beta_0+x_i\beta_1+\epsilon_i>\gamma)$

`-` 선형 회귀분석을 상상하면 아래와 같이 모형화 하면 좋을것 같다. 

$$y_i=I(\beta_0+x_i\beta_1+\epsilon_i>\gamma)$$



`-` 여기에서 $\epsilon_i$는 어떤분포라 하자. 그냥 정규분포라고 하자. 

`-` 이 모형은 주어진 자료를 설명할 수는 있다. 

`-` 이슈1: 하지만 데이터로부터 이 모형에 적합한 $\hat{\beta_0}$ $\hat{\beta_1}$ $\hat{\gamma}$를 구하는 것이 까다로워 보인다. 

`-` 이유는 (1) $I(x)$함수가 $x=0$에서 미분불가능한 점 (2) $I(x)$함수가 $x>0$ or $x<0$에서 기울기가 상수라는 점이다. (2)가 더 심각한 문제다. 

### 모형2: $y_i=\frac{e^{\beta_0+x_i\beta_1+\epsilon_i}}{e^{\beta_0+x_i\beta_1+\epsilon_i}+1}$

`-` 여기에서 $\epsilon_i$는 어떤 분포라고 하자. 잘 모르겠으니까 정규분포라고 치자. 

`-` 모형2는 모형1과 거의 모양이 비슷하는 장점이 있다. 또 $L_2$ 손실함수에 대하여 미분이 될 것 같은 기대감이 있다. 즉 이슈1이 해결된다. 

`-` 이슈2: 그런데 이슈1을 해결하면서 새로운 이슈가 생겼는데 모형자체가 좀 이상하다는 점이다. 

$$y_i=\frac{e^{\beta_0+\beta_1x_i+\epsilon_i}}{1+e^{\beta_0+\beta_1x_i+\epsilon_i}}$$

좌변은 항상 $0$또는 $1$이고 우변은 죽었다 깨어나도 $0,1$ 모두 될 수 없는데 $=$이 성립하나? 

`-` 이슈2는 일단 덮어두고 진행하자. (이걸따지면 아무것도 못할것 같으니까) 

#### 시도1 

`-` 이슈2는 무시하고 진행해보자. 아래를 최소화하는 $\beta_0$, $\beta_1$을 구하면 될 것 같다. 

$$\sum_{i=1}^{n}(y_i-\hat{y_i})^2=\sum_{i=1}^{n}\big(y_i -\frac{e^{\beta_0+x_i\beta_1}}{e^{\beta_0+x_i\beta_1}+1} \big)^2$$

`-` 이슈3: 하지만 위와 같은 손실함수는 이론적인 back ground가 부족하다는 단점이 있다. 

***이슈3-1: $L_2$ loss는 $\beta$에 대한 convex가 아니다.*** 

`-` 논의를 단순화 하기 위하여 

$$l_2=\sum_{i=1}^{n}(y_i-\hat{y_i})^2=\sum_{i=1}^{n}\big(y_i -\frac{e^{x_i\beta}}{e^{x_i\beta}+1} \big)^2$$

를 가정하자. 

`-` 손실함수를 $\frac{\partial}{\partial \beta}$로 두번 미분하여 항상 양수가 나와야 한다. 

`-` 예비계산: 

$$\frac{\partial \hat{y_i}}{\partial \beta}
=\frac{e^{x_i\beta}x_i(e^{x_i\beta}+1)-e^{x_i\beta}e^{x_i\beta}x_i} {\big(e^{x_i\beta}+1\big)^2}
=\frac{e^{x_i\beta}(e^{x_i\beta}+1)-e^{x_i\beta}e^{x_i\beta}} {\big(e^{x_i\beta}+1\big)^2}x_i
=\frac{e^{x_i\beta}} {\big(e^{x_i\beta}+1\big)^2}x_i=\hat{y_i}(1-\hat{y_i})x_i$$

`-` 한번미분: 

$$\frac{\partial l_2}{\partial \beta}=\sum_{i=1}^{n} -2(y_i-\hat{y_i})\hat{y_i}(1-\hat{y_i})x_i=\sum_{i=1}^{n} -2(y_i-\hat{y_i})\hat{y_i}(1-\hat{y_i})x_i=\sum_{i=1}^{n}-2\big(\hat{y_i}^3-\hat{y_i}^2(1+y_i)+\hat{y_i}y_i\big)x_i$$

`–` 두번미분: 

$$\big(\frac{\partial}{\partial \beta}\big)^2l_2=\sum_{i=1}^{n}-2\big(3\hat{y_i}^2-2\hat{y_i}(1+y_i)+y_i\big)x_i\hat{y_i}(1-\hat{y_i})x_i$$

`–` $\hat{y_i}(1-\hat{y_i})>0$ 이고 $x_i^2>0$이므로 

$$3\hat{y_i}^2-2\hat{y_i}(1+y_i)+y_i$$

임의 부호만 따지면 된다. 어떠한 조합 $(y_i,\hat{y_i})$에 대하여서도 위의 식이 항상 음수이면 된다.

`-` 그런데 $y_i=0$에서 $\hat{y_i}=0.99999$ 라면 위의 부호가 양수이므로 두번미분한것이 항상 양수인것은 아니다. 따라서 컨벡스가 아니다. 

***이슈3-2: 회귀모형에서의 MLE: 원래 이론적인 back ground는 로그우도함수에서 비롯된다***

`-` 잠시 아래와 같은 모형을 다시 복습하자. 

$$y_i=\beta_0+\beta_1 x_i +\epsilon_i$$

`-` $\epsilon_i \sim N(0,1)$ 이라면 $y_i \sim N(\beta_0+\beta_1x_i,1)$ 이다. 

`-` $y_i$에 대한 로그가능도를 구하면

$$-2l(\beta_0,\beta_1; y_i)=\sum_{i=1}^{n}(y_i-\beta_0-\beta_1 x_i)^2$$

이 되므로 로그가능도를 최대화하는 $\beta_0,\beta_1$의 값은 $\frac{1}{2} \big({\tt L_2~ loss}\big)$를 최소화하는 $\beta_0,\beta_1$와 동일하다. 

`-` 즉 우리가 일반적으로 배워왔단 회귀분석의 상황을 생각해보면 

$$-2l(\beta_0,\beta_1; y_i)=\sum_{i=1}^{n}(y_i-\beta_0-\beta_1 x_i)^2$$

이 성립하였기에 

$$\sum_{i=1}^{n}(y_i-\beta_0-\beta_1 x_i)^2$$

를 최소화한다는 논리전개가 힘을 받았다. 

#### 시도2

`-` 적당히 계산을 하면 아래와 같이 표현할 수 있다. 

$$\log\big(\frac{y_i}{1-y_i}\big)=\beta_0+x_i\beta_1 +\epsilon_i \quad \cdots (\star)$$

`–` 이런식의 논리전개라면 아래를 최소화하는 적당한 $\hat{\beta_0}$와 $\hat{\beta_1}$을 구한다면 될것 같다. 

$${\tt loss}=L(\beta_0,\beta_1)=\sum_{i=1}^{n}\big(\log \frac{y_i}{1-y_i}- \beta_0-x_i\beta_1 \big)^2 \quad \cdots (\star\star)$$

`–` 그리고 위와 같은 손실함수를 최소화하는 $\beta_0, \beta_1$의 조합을 구한다면 그것을 $\hat{\beta_0}, \hat{\beta_1}$이라 생각해도 무방할 것 같다. 즉 아래와 같이 볼 수 있다. 

$$\log\big(\frac{\hat{y_i}}{1-\hat{y}_i}\big) =  \hat{\beta_0}+\hat{\beta_1}x_i$$

`-` 이러한 사실에 근거한다면 $\hat{y}_i$을 아래와 같이 추론하면 될 것 같다. 

$$\hat{y_i}=\frac{\exp\big({\hat{\beta_0}+\hat{\beta_1}x_i}\big)}{\exp\big({\hat{\beta_0}+\hat{\beta_1}x_i}\big)+1}$$

`-` 좋아보이는데 무시했던 이슈2가 좀 더 심각한 문제가 되어 돌아왔다. 

`-` 이슈4: 그런데 $y_i$는 $0$ 또는 $1$의 값을 가지지 않았나? 그렇다면 수식 $(\star)$의 좌변은 어떻게 되는건지? 

`-` 이슈5: 수식 $(\star\star)$는 어떻게 계산하는가? 

`-` 오차항을 정규분포로 한정하니 조금 다루기가 어렵다. 

### 모형3: $\begin{cases}y_i \sim Ber(\mu_i) \\ \mu_i=\frac{\exp(\beta_0+x_i\beta_1)}{1+\exp(\beta_0+x_i\beta_1)}\end{cases} $

`-` 생각할수록 오묘한데 여러가지 이슈가 한번에 해결된다. 

`-` $y_i$를 고정하였을경우 로그우도함수는 아래와 같다. 

$$l(\pi_i; y_i)=\log\Big(\prod_{i=1}^{n}\pi_i^{y_i}(1-\pi_i)^{1-y_i} \Big)=\sum_{i=1}^{n}\big(y_i\log \pi_i +(1-y_i)\log(1-\pi_i) \big)$$

`–` 또한 $-l(\pi_i;y_i)$를 손실함수로 설정하면 이론적인 back-ground도 충분해진다.
