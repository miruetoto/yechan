### 함수도 객체다. 

`-` 아래를 살펴보자. 

```python
def myadd(a,b):
    return a+b
```

```python
myadd(1,2)
```

```python
?myadd
```

`-` Type이 function이다? 

`-` myadd 는 function class 의 instance이다. 

`-` 결국 myadd 역시 하나의 오브젝트에 불과하다. 

### higher-order function

`-` 예제1

아래의 예제를 살펴보자. 

```python
myadd(1,2)
```

myadd의 입력 1,2는 각각 int class 의 오브젝트였다. 

즉 문법의 논리로 보면 함수의 입력에 들어갈 수 있는 것은 오브젝트이면 된다.  

그런데 함수 자체도 오브젝트이다 $\to$ 함수도 함수의 입력처럼 쓸 수 있다? 

```python
def calc(fun,a,b):
    return fun(a,b)
```

```python
calc(myadd,2,3)
```

이처럼 함수를 입력으로 받거나 출력으로 보내는 함수를 higher-order function 이라고 한다.

`-` 예제2

미분: 아래의 함수

$$f(x)=3x^2-2x+5$$

에서 $x=2$에서의 접선의 기울기는 아래와 같이 대략적으로 구할 수 있다. 

$$\frac{f(2+h)-f(2)}{h}, \quad h=0.00001$$ 

$h$의 값을 0과 가깝게 만들수록 접선의 기울기의 정확도는 올라간다. 

미분에 익숙하다면 이론적으로는 아래와 같이 구할 수 있음을 알 수 있다. 

- $f'(x)=6x-2$ 
- $f'(2)=10$

미분을 계산해주는 코드를 구현하자. 

```python
def fx(x):
    return 3*x**2-2*x+5
```

```python
fx(2)
```

```python
def derivative(fun,x):
    h=0.000000001
    return (fun(x+h)-fun(x))/h
```

```python
derivative(fx,2)
```

`-` $g(x)=x^2$와 같은 함수도 미분할 수 있다. 

```python
def gx(x):
    return x**2
```

```python
derivative(gx,0)
```
