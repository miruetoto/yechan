**컴퓨터 상식**
- `.`: 현재폴더를 의미
- `..`: 상위폴더를 의미 
- `./myfuns.py`: 현재폴더안에 있는 myfuns 파일을 의미
- `./IPtest0403/myfuns2.py`: 현재폴더안의 IPtest0403폴더안의 myfuns2.py 파일을 의미
- `../myfuns.py`: 현재폴더보다 한단계 상위폴더에 있는 myfuns.py를 의미 
- `cd ./IPtest0403`: 현재폴더안에 있는 IPtest0403폴더로 이동하라. (`cd IPtest0403`으로 줄여서 쓸 수 있음)
- `cd ..`: 현재폴더보다 한단계 상위폴더로 이동하라. 

**따라서 `from . import myfuns2`는 현재폴더에서 myfuns2를 찾아서 임포트 하라는 의미로 해석가능** 

`-` 의미상으로는 아래가 사용가능할 듯 하지만 불가능

````python
from . import myfuns
````

`-` 이것도 `__init__.py`에 도움말을 만들 수 있다.
