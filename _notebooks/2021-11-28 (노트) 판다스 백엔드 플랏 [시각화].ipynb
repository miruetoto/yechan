```python
import random
fig, axes = plt.subplots(nrows=3, ncols=4)
#Let us build a random HexBin Chart using unique Color maps
def randomCmap():
    cmap=np.random.choice(plt.colormaps())
    return cmap

def retcombo():
    sampled_list = random.sample(symbols, 3)
    return sampled_list

for i, ax in enumerate(axes.reshape(-1)):
    sampled_list=retcombo()
    ax = returns_na.plot.hexbin(x=sampled_list[0], y=sampled_list[1],C=sampled_list[2],
                                ax=axes.flat[i],figsize=(20,12),
                                reduce_C_function=np.sum,gridsize=20,
                                cmap=randomCmap(),sharex=False)
fig.tight_layout()
```

```python
import pandas.plotting
import matplotlib as mplt
import matplotlib.pyplot as plt

#Let us get a colormap perspective by passing a parameter
def plot_colorMaps(cmap):
    fig, axes = plt.subplots(figsize=(4,0.4))
    color_map = plt.get_cmap(cmap)
    mplt.colorbar.ColorbarBase(axes, cmap=color_map, orientation = 'horizontal')
    plt.show()

    
# Get all Colormaps and set the index to zero
cmaps=plt.colormaps()
count=0

# Generate a bootstrap plot for the stock and returns for all symbols
for symbol in symbols:
    print('Bootstrap Plot for',symbol)
    pd.plotting.bootstrap_plot(data[symbol])
    pd.plotting.bootstrap_plot(returns_na[symbol])
    cmap_id = cmaps[count]
    count = count + 1
    plot_colorMaps(cmap_id)
```

```python
import pandas as pd
uscities = pd.read_csv('https://raw.githubusercontent.com/kalilurrahman/datasets/main/uscities.csv')
uscities.describe()
```

```python
#Filter for continental USA
uscities = uscities[ uscities['lng'] > -154.00]
uscities = uscities[ uscities['lat'] < 54.00]
latitude =  uscities['lat']
longitude =  uscities['lng']
population, density = uscities['population'], uscities['density']
density = density/50
plt.figure(figsize=(10,5))
plt.scatter(longitude, latitude, label=None,
            c=np.log2(population), cmap='YlGnBu',
            s=density, linewidth=0.1, alpha=0.5)
plt.axis('equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{2}$(population)')
plt.title('Continental US Cities:  Outline through Population')
```
