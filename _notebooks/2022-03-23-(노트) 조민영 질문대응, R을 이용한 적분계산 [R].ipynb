(문제) 아래를 해결하고 싶음 

$$\sum_{k=0}^{10}\left(\int_{0}^{2} \int_0^1 yx^{\alpha_1}(1-x)^{-\alpha_2}dxdy\right)k \Rightarrow argmin_{\alpha_1,\alpha_2}$$

(해결) $\alpha_1, \alpha_2$를 대입하면 적당한 값이 계산되는 함수를 구현 

```r
f = function(a1,a2,x,y){
    integral2(function(x,y) y*x^alpha1 * (1-x)^(-alpha2), xmin=0,xmax=1,ymin=0,ymax=2)$Q
}
```

```r
sumf = function(a1,a2){
    rtn = 0 
    for(i in 0:10) rtn= rtn+ f(a1,a2,i)
    rtn
}
```

```r
sumf(0.1,0.3) # alpha1=1, alpha2=3 
```
