`-` .mkdir()을 통하여 폴더를 만들 수 있다.

```python
path_dropbox.ls()
```

```python
(path_dropbox/'0000asdf').mkdir()
```

```python
path_dropbox.ls()
```

> 새로운 디렉토리를 만들때에는 `/`연산자를 사용해서 (path_dropbox/'0000asdf')와 같이 새로운 path를 만든 뒤에 .mkdir() 메소드를 사용한다. 

> Warninig: path_dropbox.mkdir('/0000asdf')는 동작하지 않음
