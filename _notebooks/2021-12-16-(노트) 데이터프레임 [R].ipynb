### 티블 만드는 방법 

`-` 방법1: 선언하여 만들기 (새롭게 정의)

```r
tibble(v1=1:3,v2=c(T,F,T), v3=c("a","b","c"))
```

- 각 열들의 자료형이 모두 달라도 가능 (티블의 매력) 
- 티블의 형태는 매트릭스 같지만 이처럼 다른 자료형이 들어가는 것은 마치 리스트를 연상시킨다. 

`-` 방법2: 선언하여 만들기2 (기존에 정의된 벡터등을 활용) 

```r
x=1:5
y=c("a","b","c","d","e")
tibble(x,y)
```

`-` 방법3: 벡터를 일단 만들고 티블로 변형 

```r
as_tibble(x)
```

`-` 방법4: 매트릭스를 일단 만들고 티블로 변형

```r
as_tibble(cbind(x,y))
```

- 모두 chr로 저장되는 단점이 있음! 

`-` 방법5: 리스트에서 생성 

```r
as_tibble(list(x=x,y=y))
```

`-` 데이터프레임의 경우도 티블과 비슷하게 만들 수 있다. 

```r
data.frame(v1=1:3,v2=c(T,F,T), v3=c("a","b","c")) # 방법1 
```

```r
data.frame(x,y) # 방법2
```

```r
data.frame(x) # 방법3
```

```r
data.frame(cbind(x,y)) # 방법4
```

```r
data.frame(list(x=x,y=y)) # 방법5
```

`-` tibble()과 data.frame()이 항상 같은 결과를 주는 것은 아님 

```r
lst= list(x=x,y=y,z=cbind(x,y))
lst 
```

```r
as_tibble(lst)
```

```r
data.frame(lst)
```

`-` 데이터프레임과 티블은 as_tibble()과 data.frame()을 이용해 서로의 형태로 변화가능하다. 

```r
data.frame(v1=1:3,v2=c(T,F,T), v3=c("a","b","c")) 
```

```r
data.frame(v1=1:3,v2=c(T,F,T), v3=c("a","b","c")) %>% as_tibble
```

```r
data.frame(v1=1:3,v2=c(T,F,T), v3=c("a","b","c")) %>% as_tibble %>% data.frame
```

즉 아래가 가능함 

```r
df=data.frame(v1=1:3,v2=c(T,F,T), v3=c("a","b","c")) 
df
```

```r
df=as_tibble(df)
df
```

```r
df=data.frame(df)
df
```

### 데이터프레임, 티블의 해석 

`-` 매트릭스: (1) 모든 열은 동일한 숫자의 row를 가져야하며 (2) 행렬의 모든 원소는 동일한 자료형을 가져야 한다. 

`-` 리스트: (1) 모든 원소는 동일한 길이의 벡터로 구성될 필요가 없다. (애초에 모든 원소가 벡터일 이유도 없어요!) (2) 그리고 각 원소가 동일한 자료형을 가질 필요도 없다.

`-` 티블 혹은 데이터프레임은 

- 자료형이 자유로운 매트릭스로 해석할 수 있고 
- 모든 원소가 동일한 길이의 벡터로 구성된 리스트

로 생각할 수 있다. 

### 데이터프레임과 티블 기본사용법 

`-` 데이터 셋팅 

```r
x= 1:3
y= c(T,F,T)
z= c("a","b","c")

lst = list(x=x,y=y,z=z)
mat = cbind(x,y,z)
df = tibble(x=x,y=y,z=z)
```

```r
print(lst)
print(mat)
print(df)
```

`-` df에서 열들의 이름을 알고 싶다? 

```r
print(colnames(lst))
print(colnames(mat))
print(colnames(df))
```

```r
print(names(lst))
print(names(mat))
print(names(df))
```

`-` df에서 첫번째 열을 추출하고 싶다? 

```r
print(mat[,1])
#print(lst[,1]) ## 이건 에러난다.. 
print(df[,1])
```

```r
# print(mat[[1]])
print(lst[[1]]) 
print(df[[1]])
```

`-` df를 매트릭스처럼 생각하는건 사실 직관적이다. 그런데 리스트라고 생각하는건 쉽지 않다. df를 리스트라고 해석하면 아래와 같은 문법도 쉽게 이해된다. 

```r
print(lst[1])
print(df[1])
```

```r
print(lst$x)
print(df$x)
```

```r
print(lst['x'])
print(df['x'])
```

```r
print(lst[1:2])
print(df[1:2])
```

`-` 매트릭스처럼 생각한다면 아래와 같은 형태도 가능하다. 

```r
print(mat[2:3,])
print(df[2:3,])
```

```r
print(mat[,1:2])
print(df[,1:2])
```

```r
print(mat[,c('x','y')])
print(df[,c('x','y')])
```

```r
print(mat[2:3,c('x','y')])
print(df[2:3,c('x','y')])
```

### dataframe과 tibble의 차이 

`-` 거의 똑같음 

`-` dataframe은 R에서 기본적으로 제공하는 자료형, tibble은 tidyverse제작자가 변형한 자료형 

`-` tibble이 dataframe의 개선형태라고 해들리위컴이 주장함 $\to$ 제가 사용할떄는 큰 차이없었고 아무거나 구분없이 쓰면 될것 같아요 

```r
df=tibble(
    f=c("runif","rnorm","rnorm"),
    params=list(list(n=3,min=-1,max=1),list(n=5,mean=0,sd=5),list(n=2,mean=5))
)
df
```

```r
df %>% data.frame %>% mutate(simrslt=invoke_map(f,params))
```

```r
df %>% data.frame
```
