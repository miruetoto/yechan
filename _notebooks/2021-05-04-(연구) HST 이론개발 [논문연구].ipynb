### Ref 
- Truquet, L. (2019). Local stationarity and time-inhomogeneous Markov chains. Annals of Statistics, 47(4), 2023-2050.

### 가정

Truquet2019의 p4에서 서술한 프레임워크를 가져오겠다. 

`-` $\Omega_1, \Omega_2$: sample space
 - $\Omega_1={\cal V}\times \mathbb{R}^n$
     - $\omega_1$의 예는 $(v_\ell,h_\ell)$. 
     - 여기에서 $h_\ell$은 $\ell$시점에서의 땅의 모양. 
 - $\Omega_2={\cal V}$
     - $\omega_2$는 ${\cal V}$에 있는 임의의 노드값. 

`-` $d_1,d_2$: $\Omega_1, \Omega_2$에서의 metric. 

`-` ${\cal B}_1,{\cal B}_2$: $\Omega_1,\Omega_2$에 대응하는 보렐 시그마필드 

`-` $\kappa: \Omega_1 \times {\cal B}_2 \to [0,1]$: 마코프커널
 - 모든 $B_2\in {\cal B}_2$에 대하여 맵핑 $\omega_1 \to \kappa(\omega_1,B_2)$는 ${\cal B}_1$-measurable 함. 
 - 모든 $\omega_1 \in \Omega_1$에 대하여 맵핑 $B_2 \to \kappa(\omega_1,{\cal B}_2)$는 probability measure임.

`-` 모든 $i \in \Omega_1$에 대하여 아래와 같이 표현할 수 있다. 
$
\kappa(B_2|i)=\sum_{j\in B_2}\kappa(\{j\}|i), \quad \forall i \in \Omega_1, \forall B_2 \in {\cal B}_2
$
