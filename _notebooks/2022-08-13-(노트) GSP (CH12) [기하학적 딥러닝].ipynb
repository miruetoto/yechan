### Graph Shift Operator

노드가 $N$개인 그래프 ${\cal G} = ({\cal N}, {\cal E})$가 있다고 가정하자. Graph Shift Operator (GSO) 는 아래와 같은 특징을 가지는 $N \times N$ matrix ${\bf S}$로 정의한다. 
- $i = j \quad \Longrightarrow \quad S_{ij}=0$ 
- $(i,j) \notin {\cal E} \quad \Longrightarrow \quad S_{ij}=0$ 

따라서 어제이슨시 매트릭스, 라플라시안 매트릭스 등은 GSO가 된다. 만약에 

$${\bf S}={\bf V} {\boldsymbol \Lambda} {\bf V}^H$$

를 만족하는 적당한 unitary matrix ${\bf V}=[V_1, V_2, \dots, V_N]$과 대각행렬 ${\boldsymbol \Lambda}$가 존재한다면 ${\bf S}$가 ***normal*** 하다고 부른다. 

### Graph Fourier Transform

${\boldsymbol x}=[x_1,x_2,\dots, x_N]^\top \in \mathbb{R}^N$ 을 그래프신호, $\tilde{\boldsymbol x}={\bf V}^{H}{\boldsymbol x}$ 를 ${\bf V}^H$에 의한 그래프퓨리에변환이라고 부른다. 

### Shift-invarinat graph filters

GSO ${\bf S}$의 polynomial 형태로 표현가능한 ${\bf H}: \mathbb{R}^N \to \mathbb{R}^N$를 고려하자. 즉 아래와 같은 ${\bf H}$를 고려하자. 

$${\bf H}:=\sum_{l=0}^{L-1}h_l{\bf S}^l$$

이처럼 GSO의 polynomial 형태로 표현가능한 operator를 ***shift-invarinat graph filters***라고 부른다. 우리가 이러한 필터에 관심을 가지는 이유는 (1) 이러한 필터는 time invariant filters에 관련된 classical한 이론이 있어서 그걸 퍼실러테이팅 할 수 있는 장점이 있고 (2) 이러한 필터가 그래프 신호의 local한 특징을 잘 잡아내기 때문이다. **쉬프트-인배리언트 그래프 필터**는 아래와 같이 표현가능하다. 

$${\bf H}:=\sum_{l=0}^{L-1}h_l{\bf S}^l=\sum_{l=0}^{L-1}h_l\big({\bf V}{\boldsymbol \Lambda}{\bf V}^H\big)^l={\bf V}\left(\sum_{l=0}^{L-1}h_l{\boldsymbol \Lambda}^l\right){\bf V}^H$$

여기에서 $\left(\sum_{l=0}^{L-1}h_l{\boldsymbol \Lambda}^l\right)$를 새로운 대각행렬 ${\bf \hat{H}}=\text{diag}({\bf \hat h})$ 로 정의하면 

$${\bf H}:={\bf V}{\bf \hat H}{\bf V}^H={\bf V}\text{diag}({\bf \hat h}){\bf V}^H$$

${\bf \hat{H}}$는 ${\bf H}$의 frequency response w.r.t ${\bf V}^H$ 라고 해석할 수 있다. 만약에 아래와 같이 $N \times L$ 방데르몽드(Vandermonde) 행렬을 정의하면 

$${\boldsymbol \Psi}=\begin{bmatrix} 1 & \lambda_1 & \dots & \lambda_1^{L-1} \\ \dots & \dots & \dots & \dots \\ 1 & \lambda_N & \dots & \lambda_N^{L-1}\end{bmatrix}$$

${\bf H}$를 다시 아래와 같이 표현가능하다. 

$${\bf H}:={\bf V}{\bf \hat H}{\bf V}^H={\bf V}\text{diag}({\bf \Psi h}){\bf V}^H={\bf V}\text{diag}({\bf \hat h}){\bf V}^H$$

### Weakly stationary

(정의) ${\boldsymbol x} = {\bf H}{\boldsymbol n}$ 으로 표현가능하면 ${\boldsymbol x}$를 ${\bf S}$에 대한 weakly stationary random process 라고 한다. 단 ${\boldsymbol x}$는 ${\cal G}=({\cal V},{\cal E})$ 에서 정의된 그래프신호, ${\boldsymbol n}$는 평균이 0인 white noise, ${\bf S}$는 noraml한 graph shift operator이며 ${\bf H}$는 ${\bf S}$에 대응하는 쉬프트-인배리언트 그래프 필터이다.
