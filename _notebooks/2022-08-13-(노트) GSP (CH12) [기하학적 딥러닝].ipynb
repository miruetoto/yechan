### Graph Shift Operator

`-` 노드가 $N$개인 그래프 ${\cal G} = ({\cal N}, {\cal E})$가 있다고 가정하자. Graph Shift Operator (GSO) 는 아래와 같은 특징을 가지는 $N \times N$ matrix ${\bf S}$로 정의한다. 
- $i = j \quad \Longrightarrow \quad S_{ij}=0$ 
- $(i,j) \notin {\cal E} \quad \Longrightarrow \quad S_{ij}=0$ 

따라서 어제이슨시 매트릭스, 라플라시안 매트릭스 등은 GSO가 된다. 

`-` 만약에 ${\bf S}={\bf V} {\boldsymbol \Lambda} {\bf V}^H$를 만족하는 적당한 unitary matrix ${\bf V}=[V_1, V_2, \dots, V_N]$과 대각행렬 ${\boldsymbol \Lambda}$가 존재한다면 ${\bf S}$가 ***normal*** 하다고 부른다. 

`-` ${\bf f}=[f_1,f_2,\dots,f_N]^\top \in \mathbb{R}^N$ 을 그래프신호, ${\bf \tilde{f}}={\bf V}^{H}{\bf f}$ 를 ${\bf V}^H$에 의한 그래프퓨리에변환이라고 부른다. 

`-` GSO ${\bf S}$의 polynomial 형태로 표현가능한 ${\bf H}: \mathbb{R}^N \to \mathbb{R}^N$를 고려하자. 즉 아래와 같은 ${\bf H}$를 고려하자. 

$${\bf H}:=\sum_{l=0}^{L-1}h_l{\bf S}^l$$
