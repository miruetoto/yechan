### R의 비밀 

***존재하는 모든것은 오브젝트이다. 그리고 발생하는 모든것은 함수호출이다.***

`-` 예제1

```r
f = function(x,y) (x+y)**2
```

```r
f(10,5)
```

발생한것은 225이고 슬로건에 따르면 225는 어떠한 함수의 결과이다. (f라는 이름을 가진 함수의 결과이다. 당연한걸 설명하나?) 

`-` 예제2

```r
10+20 
```

발생하는것은 30, 따라서 슬로건에 따르면 30은 어떠한 함수의 결과이다. 

```r
`+`(10,20)
```

- 사실은 `+` 라는 이름의 함수가 있어서 이를 적용한 결과였던것임 

`-` 예제3

```r
for (i in 1:2) print(i)
```

발생하는것은 1,2 인데 슬러건에 따르면 이것은 `for`라는 함수의 결과인가? 

```r
`for`(i, 1:2, print(i))
```

- `for`라는 함수가 존재했고, 그 함수는 i, 1:2, print(i))를 입력을 받고있었음 

`-` 예제4

```r
for (i in 1:2){
    i=i+1 
    print(i)
}
```

```r
`for`(i ,1:2, {i=i+1 
               print(i)})
```

```r
`for`(i ,1:2, {i=i+1;print(i)})
```
