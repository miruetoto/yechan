#### str 메소드 (str 특수기능)

`-` 파이썬의 변수는 단순히 정보를 담는 그릇이 아니다. 항상 기능이 있다. 

```python
a='ABCD' # a라는 변수는 단순히 정보를 담는 그릇이 아님
```

```python
a.lower() # a.lower()를 쓰면 a의 모든 문자열을 소문자로 바꾼연산결과를 제공
```

여기에서 lower()는 문자열에 특화된 기능임. 따라서 당연히 아래는 불가능 

```python
a=3.14
a.lower()
```

`-` 자료형에 특화된 기능을 확인하는 방법? `a.` + `tab` 으로 목록 확인가능

```python
a='guebin'
```

```python
a. # 이상태에서 탭을 누르면 된다. 
```

```python
a.capitalize()
```

`-` 문자열에 대한 다른내용들은 추후에..

`-` 마음의눈: `a.f()`형태를 읽는 팁 
- `a.f()`는 `f(a)`로 생각하면 편리함.
- `a.f(2)`는 `f(a,2)`로 생각하면 편리함. 
- 이런점에서 R의 `%>%`연산자와 비슷하다고 생각할 수 있다. (약간 다르긴함) 

`-` 사실 `.`은 좀 더 다양한 상황에서 쓰일수 있다. 위의 경우 `변수이름.함수이름` 의 형태인데 이것외에 
- `패키지이름.함수이름`
- `패키지이름.변수이름` 
- `패키지이름.패키지이름.함수이름` 
- ... 

와 같이 다양한 형태가 가능하다. 근본적인 공통점은 .을 기준으로 `상위개념.하위개념` 으로 이해하는것이 좋다.
