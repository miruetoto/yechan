### 데이터프레임을 왜 만들어야 하는가? 

`-` 결국 벡터를 왜 만들어야 하는가? 매트릭스를 왜만들어야 하는가? 와 같은 질문임 

`-` 벡터, 매트릭스가 자료의 관리성과 계산상의 이점을 챙기듯이 데이터프레임 역시 그러하다. 

`-` 의문: 매트릭스와 데이터프레임의 차이는? 

- 매트릭스는 수학과 좀 더 친하고 데이터프레임은 통계/데이터베이스와 좀 더 친하다.
- 지원하는 계산형태가 다르다. (매트릭스: 곱셈, 미분, 고유치분해, SVD ... // 데이터프레임: 쿼리, 필터, 조인, 자료시각화, R에서 지원하는 각종 통계분석패키지 ... )
- 매트릭스는 행과 열에 특별한 의미가 따로 없다. 
- 데이터프레임은 행=관측치, 열=변수 라는 의미가 있다. 

`-` **제 생각:** 결국 데이터프레임 역시 매트릭스의 한 종류라 볼 수 있는데 열과 행에 특별한 의미를 부여하여 통계/데이터베이스적인 반복작업을 더 잘 수행하도록 변형한 것

`-` (참고) 파이썬과 R에서 매트릭스와 데이터프레임 

||파이썬|R|
|:-:|:-:|:-:|
|매트릭스|넘파이,텐서|매트릭스| 
|데이터프레임|판다스|데이터프레임, 티블|

`-` 이 문제에서 데이터프레임을 만든이유: pmap_dbl 을 사용하기 위해서 

`-` 데이터프레임과 티블의 차이: 티블은 데이터프레임을 개량하여 만든 것. 

### 데이터프레임(티블)을 만드는 방법 

`-` 방법1: 일단 매트릭스 A를 만들고 as_tibble(A)와 같이 만든다. 

```r
A=rbind(c(1,2,3,4),c(5,6,7,8),c(9,10,11,12))
A 
```

```r
as_tibble(A) # 데이터프레임(=티블)
```

```r
A %>% as_tibble()
```

```r
A %>% as_tibble 
```

`-` 방법2

```r
library(tidyverse)
f=c("runif","rnorm","rpois") 
params=list(list(min=-1,max=1),list(sd=5),list(lambda=10))
tibble(f=f,params=params)
```

`-` 방법3

```r
tribble(
  ~f,      ~params,
  "runif", list(min = -1, max = 1),
  "rnorm", list(sd = 5),
  "rpois", list(lambda = 10)
)
```
