### 컬럼추가 

```python
df=pd.DataFrame({'X1':[1,2,3,4,5],'X2':[2,3,4,5,6]})
df
```

`-` 새로운 칼럼 [3,3,3,3,3]을 추가하자. 

```python
df.assign(X3=[3,3,3,3,3]) ## 계산결과 
```

```python
df ## 원래 자료는 변화하지 않음 
```

```python
df= df.assign(X3=[3,3,3,3,3]) ## 계산결과를 update 
```

- 엔터가 엄청 큰 역할을 한다. 

`-` 좀 더 복잡한 함수일 경우는? 

`-` 방법1

```python
f= lambda x: (x-np.mean(x))/np.std(x) 
```

```python
f(df.X1)
```

```python
df.assign(X4=f(df.X1))
```

`-` 방법2

```python
df.assign(X4= lambda df: (df.X1-np.mean(df.X1))/np.std(df.X1))
```

```python
def f(x):
    if x<2: 
        y=2
    elif x<4:
        y=4
    else:
        y=6
    return y
```

```python
df.assign(X4=list(map(f,df.X1)))
```

### where / mask (inplace transform)

`-` 기능1

```python
_df=pd.DataFrame(np.random.normal(size=(5,2)),columns=list('AB'))
_df
```

```python
_df.where(_df>0)
```

그다지 유용해 보이지도 않고 아래와 같은 코드이다.

```python
_df[_df>0]
```

`-` 기능2: NaN에 해당하는 값들을 치환할수 있다. 

```python
_df
```

```python
_df.where(_df>0,0)
```

```python
_df.where(_df>0, -_df)
```

`-` lambda와 함께 사용가능 

```python
_df.where(lambda x: x>0, lambda x: 0)
```

```python
_df.where(lambda x: x>0, lambda x: -x)
```

`-` mask: where의 반대

```python
_df
```

```python
_df.where(_df>0,0)
```

```python
_df.mask(_df>0,0)
```

```python
_df.mask(lambda x: x>0,0)
```

`-` np.where과 비교

```python
_df 
```

```python
np.where(_df<0,1,0) # R에서의 ifelse와 같음 
```
