```python
from fastai.text.all import *
```

```python
path = untar_data(URLs.IMDB)
```

```python
files = get_text_files(path, folders=['train','test','unsup'])
files
```

```python
dls=DataBlock(blocks=TextBlock.from_folder(path,is_lm=True),
              get_items=partial(get_text_files,folders=['train','test','unsup']),
              splitter=RandomSplitter(0.1)).dataloaders(path,path=path, bs=128, seq_len=80)
```

```python
dls.show_batch()
```

```python
lrnr = language_model_learner(dls,AWD_LSTM,drop_mult=0.3,metrics=[accuracy,Perplexity()]).to_fp16()
```

```python
lrnr.fit_one_cycle(5)
```

```python
TEXT = 'I liked this movie because' 
N_WORDS= 40 
N_SENTENCES = 2 
preds = [lrnr.predict(TEXT,N_WORDS,temperature=0.75) for _ in range(N_SENTENCES)]
print("\n".join(preds))
```
