## import 

```python
import numpy as np
```

```python
def test(a,b):
    if id(a)==id(b): 
        print('별명')
    elif id(a)==id(b.base) or id(a.base)==id(a):
        print('뷰')
    elif id(a.base)!=id(None) and id(b.base)!=id(None) and id(a.base)==id(b.base):
        print('공통의 base를 가짐')
    else:
        print('카피이거나 아무 관련 없는 오브젝트')
```

### np.swapaxes

```python
x=np.array([[1,2,3]])
np.swapaxes(x,0,1) # x.T 와 같음 
```
