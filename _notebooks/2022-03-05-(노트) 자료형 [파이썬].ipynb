**[예제4]**: 함수의 입력으로 튜플을 넣을때 

*[예제4의 예비학습]* 함수 벼락치기 

```python
def cal(a,b):
    print(str(a) + '+' + str(b) + '=' + str(a+b))
    print(str(a) + '-' + str(b) + '=' + str(a-b))
    print(str(a) + '*' + str(b) + '=' + str(a*b))
    print(str(a) + '/' + str(b) + '=' + str(a/b))
```

```python
cal(2,33)
```

벼락치기 끝 

```python
### 우리가 원하는 형태 
input=[3,4]
cal(input)
```

```python
### 실제 
cal(input[0],input[1])
```

```python
input=(3,4)
```

```python
cal(*input) 
```

**[예제5]** 함수의 입력을 튜플로 넣을때 (2) 

두점 사이의 거리를 구하는 함수를 만들어 보자. 

$x=(x_1,x_2,x_3)$

$y=(y_1,y_2,y_3)$

의 거리를 구하려면 

$\sqrt{(x_1-y_1)^2+(x_2-y_2)^2+(x_3-y_3)^2}$

```python
def distance(x1,x2,x3,y1,y2,y3):
    import math 
    d=(x1-y1)**2+(x2-y2)**2+(x3-y3)**2
    print(math.sqrt(d))
```

```python
x=(0,0,0)
y=(0,1,1)
distance(*x,*y)
```

```python
def distance(x,y):
    import math 
    x1,x2,x3=x 
    y1,y2,y3=y
    d=(x1-y1)**2+(x2-y2)**2+(x3-y3)**2
    print(math.sqrt(d))
```

```python
distance(x,y)
```
