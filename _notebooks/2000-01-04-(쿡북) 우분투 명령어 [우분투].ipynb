### conda 


`-` 설명보기
```
conda
conda env -h
conda install -h
conda remove -h 
conda update -h 
```


`-` 가상환경 리스트 출력 
```
conda env list
```

`-` 가상환경 생성
```
conda create -n py38r40 python=3.8
```

`-` 가상환경을 삭제 
```
conda env remove --name py38r40
```

`-` 패키지 설치
```
conda install -c conda-forge jupyterlab 
```

`-` 패키지 삭제 
```
conda remove jupyterlab 
conda remove -n py38r40 jupyterlab 
```

`-` 패키지 업데이트
```
conda update -n myenv scipy
```

`-` 패키지 리스트
```
conda list 
```

### pip

`-` 설명보기 
```
pip
```

`-` 설치된 패키지 목록
```
pip list
pip list > list.txt
pip freeze # 좀 더 자세히 나온다 
pip freeze > list.txt 
```

`-` **설치된 패키지의 정보**
```
pip show matplotlib
```
결과는 아래와 같이 나온다. 
```
Name: matplotlib
Version: 3.4.1
Summary: Python plotting package
Home-page: https://matplotlib.org
Author: John D. Hunter, Michael Droettboom
Author-email: matplotlib-users@python.org
License: PSF
Location: /home/cgb/anaconda3/envs/py38r40/lib/python3.8/site-packages
Requires: numpy, cycler, pillow, python-dateutil, pyparsing, kiwisolver
Required-by: seaborn
```

`-` 패키지 설치
```
pip install rpy2
pip install -r list.txt 
pip install dash==1.13.3
pip install jupyterlab "ipywidgets>=7.5"
pip install -U numpy
```

`-` 패키지 업그레이드
```
pip install --upgrade pip
pip install --upgrade tensorflow
```

`-` 패키지 삭제
```
pip uninstall matplotlib
```

### 나노에디터
`-` 파일열기
```
sudo nano /etc/apt/sources.list
```

`-` 읽기전용 파일만들기: 파일을 만드는것도 파일 여는방법과 동일함. 즉 아래와 같이 하면 `exam.txt`가 현재 폴더에 있다면 `열고`아니면 `(읽기전용으로) 만든다.`
```
sudo nano /home/cgb/Desktop/exam.txt
```

`-` 읽기전용이 아닌 일반파일 만들기 
```
nano /home/cgb/Desktop/exam.txt
```

`-` 나노에디터종료: `컨트롤+X`

`-` 파일저장: `컨트롤+O`

`-` 찾기: `컨트롤+W`

`-` 되돌리기: `알트+U`
