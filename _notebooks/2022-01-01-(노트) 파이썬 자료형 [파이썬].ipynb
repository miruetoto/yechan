```python
numpy.exp(1) # exp함수를 쓸것인데, 그 함수는 우리가 임포트한 numpy에 있음. 
```

```python
numpy.log(2.718281828459045) # log함수를 쓸것인데, 그 함수는 우리가 임포트한 numpy에 있음. 
```
