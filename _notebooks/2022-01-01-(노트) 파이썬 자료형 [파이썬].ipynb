```python
A=[[1,2,3],[4,5,6],[7,8,9]]
A
```

`-` 위와 같이 하면 A는 거의 아래와 같은 매트릭스처럼 생각하고 원소를 뽑을 수 있다. 

$$ {\bf A}=\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$$

`-` ${\bf A}$에서 (1,1)의 원소를 뽑자 $\to$ 1을 뽑자는 의미. 

```python
A[0,0] ## 파이썬의 인덱스가 0부터 시작하니까 (1,1)에 대응하는 인덱스는 (0,0)
```

- 실패: 이런문법은 배운적이 없어요 

```python
A[0][0]
```

- 이건 성공함: 이것도 배운적 없는 문법인것 같은데? 

`-` 성공의 이유분석? 

```python
A # A는 [1,2,3], [4,5,6], [7,8,9] 3개의 리스트를 원소로 가짐 
```

```python
A[0] # 첫번쨰 원소인 [1,2,3]이 나옴
```

```python
A[0][0] # A[0]의 첫번째 원소인 1이 나옴 
```

- 배운적은 없지만 배운걸 응용해서 코드를 차근차근 뜯어보면 알 수는 있음.

`-` ${\bf A}$에서 (2,3)위치의 원소를 뽑자 $\to$ 파이썬에서는 (1,2)인덱스임

$$ {\bf A}=\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$$

```python
A[1][2]
```
