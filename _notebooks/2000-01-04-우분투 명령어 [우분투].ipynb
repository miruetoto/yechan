### 프로그램 설치 
`-` 프로그램을 지우기 위해서는 아래와 같이 하면 된다. 
```
sudo apt-get remove r-base-core
```

### 나노에디터 사용방법  
`-` 파일열기
```
sudo nano /etc/apt/sources.list
```

`-` 읽기전용 파일만들기: 파일을 만드는것도 파일 여는방법과 동일함. 즉 아래와 같이 하면 `exam.txt`가 현재 폴더에 있다면 `열고`아니면 `(읽기전용으로) 만든다.`
```
sudo nano /home/cgb/Desktop/exam.txt
```

`-` 읽기전용이 아닌 일반파일 만들기 
```
nano /home/cgb/Desktop/exam.txt
```

`-` 나노에디터종료: `컨트롤+X`

`-` 파일저장: `컨트롤+O`

`-` 찾기: `컨트롤+W`

`-` 되돌리기: `알트+U` 

### ssh연결 
`-` 처음에 ssh를 연결하기위해서는 연결**당하는** 컴퓨터에 가서 아래를 실행해야 한다. 
```
sudo apt install openssh-server
```

### sublime

`-` 찾아바꾸기: 컨트롤+h

`-` 화면분할: 옵션+커맨드+2,3,4 ...


### conda 
`-` 가상환경 리스트 출력 
```
conda env list
```

`-` 가상환경 생성
```
conda create -n py38r40 python=3.8
```

`-` 가상환경을 삭제 
```
conda env remove --name py38r40
```

`-` 패키지 설치
```
conda install -c conda-forge jupyterlab 
```

`-` 패키지 삭제 
```
conda remove jupyterlab 
```



### 파일탐색기

#### 파일생성  
`-` 파일생성: 파일을 만들기 원하는 폴더로가서 아래를 입력한다.
```
cat > sample.txt
```
아무것이나 치고 (예를들어 `asdf`) `컨트롤+C`를 눌러서 빠져나온다. 

`-` 혹은 나노에디터를 사용해도 된다. 

#### 폴더삭제 

`-` 폴더삭제
```
rm -r folderName
```
