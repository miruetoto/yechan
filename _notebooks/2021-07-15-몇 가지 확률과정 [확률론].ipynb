### 1. 독립정상증분과정

#### 독립증분과정(random process with independent increment)

`def` 자연수 $k$와 실수 $t_0,t_1,\dots,t_k$를 어떻게 두더라도 $(Y_{t_1}-Y_{t_0},\dots,Y_{t_k}-Y_{t_{k-1}})$이 독립확률벡터이면 '확률과정 $\{Y_t, t \in I \}$의 증분은 독립이다' 라고 하거나, '확률과정 $\{Y_t,t\in I\}$는 독립증분 과정이다' 라 말한다. 

#### 정상증분과정(random process with stationary increment)

`def` 증분 $\{Y_t-Y_s\}$가 정상이면, 확률과정 $\{Y_t\}$를 정상증분 확률과정이라 부른다. 

#### 독립정상증분과정(random process with independent and stationary increment, i.s.i)

`def` 증분이 독립이고 또 정상인 확률과정을 독립정상증분 과정이라 부른다. 

#### 정리1

독립이고 분포가 같은 확률변수들을 더한 이산시간 확률과정은 독립정상증분 과정이고 그 역도 성립한다.
