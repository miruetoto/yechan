#### A. Basic CNN

**1) Sentence Modeling:** 

`-` ${\bf w}_i \in {\cal R}^d$ 
 - ${\bf w}_i$: 문장에 있는 i번째 단어 
 - $d$: dimension of the word embedding 

`-` ${\bf W} \in {\cal R}^{n \times d}$: embedding matrix 
- $n$은 문장에 포함된 단어의 수 

`-` ${\bf w}_{i:i+j}$: ${\bf w}_i, {\bf w}_{i+1}, \dots , {\bf w}_j$ 
- ${\bf w}_i, {\bf w}_{i+1}, \dots , {\bf w}_j$ 가 아니라 ${\bf w}_i, {\bf w}_{i+1}, \dots , {\bf w}_{i+j}$ 가 아닐까? 

`-` $k \in {\cal R}^{hd}$
- $h$는 윈도우 크기 

`-` $c_i$ 는 아래의 수식으로 계산한다.  

$$c_i = f\big({\bf w}_{i:i+h-1} {\bf k}^\top +b\big)$$ 

(해설) 

- ${\bf w}_{i:i+h-1}=[{\bf w}_i, {\bf w}_{i+1}, \dots, {\bf w}_{i+h-1}]$이다. 그런데 ${\bf w}_i$는 d차원이고 나머지도 d차원이므로 결국 ${\bf w}_{i:i+h-1}$는 $hd$차원이 된다. 
- $k$는 1차원 커널인데 $k$가 웨이트의 역할을 한다. 
- 따라서 ${\bf w}_{i:i+h-1}$는 ${\bf x}_i$로 ${\bf k}^\top$는 ${\boldsymbol \beta}$로 생각할 수 있다.
