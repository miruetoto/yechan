### 별명, 뷰, 카피 

`-` 별명, 뷰, 카피를 테스트하는 함수를 만들자. 

```python
def test(a,b):
    if id(a)==id(b): 
        print('별명')
    elif id(a)==id(b.base) or id(a.base)==id(a):
        print('뷰')
    elif id(a.base)!=id(None) and id(b.base)!=id(None) and id(a.base)==id(b.base):
        print('공통의 base를 가짐')
    else:
        print('카피이거나 아무 관련 없는 오브젝트')
```

`-` 혹시 잘 동작하나? 

(테스트1)

```python
a=np.array([1,2,3,4])
b=a
```

```python
test(a,b)
```
