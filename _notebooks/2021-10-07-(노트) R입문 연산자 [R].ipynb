`-` 예제2의 다른구현 (논리연산자와 TRUE, FALSE열을 다루는데 익숙치 않다면 고생스러울 때가 있다.) 

```r
x<-c() 
j=1 
for (i in 1:30){
    if (i%%3==0){
        x[j] = i 
        j=j+1
    }
}
```

```r
x
```

```r
x<-c() 
j=1 
for (i in 1:30){
    if (i%%3!=0){
        x[j] = i 
        j=j+1
    }
}
```

```r
x
```

아래와 같은 $y$를 생성하라. (abs함수와 중첩 ifelse를 이용하면 쉬움)

$y=\begin{cases} x & |x|>5 \\ 
5 & 2<|x|\leq 5 \\
0 & |x|\leq 2
\end{cases}$

(풀이)

```r
y=ifelse(abs(x)>5,x,ifelse(abs(x)>2,5,0)) 
```

```r
y
```
