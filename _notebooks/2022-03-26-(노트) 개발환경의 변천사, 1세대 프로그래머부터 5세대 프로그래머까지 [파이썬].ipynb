#### 개발환경 구축방법이 많은 이유? 

`-` 파이썬 개발환경 구축은 수많은 방법이 있다. (검색하면 진짜 무수히 많이 나옴) 

`-` 이는 마치 라면의 레시피를 검색했을때 수많은 방법이 나오는것과 유사함. 
- 방법1: 스프를 먼저 넣고 끓인다음에 라면을 넣어야 합니다. 
- 방법2: 양은냄비에 물넣고 물 끓으면 스프넣고 라면 넣고 마지막에 계란을 넣는다. 
- 방법3: 웤에 물을 조금 넣고 라면과 스프를 넣고 자글자글 끓이다가 마지막에 참기름을 넣어요  
- 방법4: 먹다남은 삼겹살을 후라이팬에 볶은뒤에 물을 붓고 라면을 먹는다. 
- 방법5: 용기에 라면과 스프와 뜨거운물을 넣고 랩을 쒸운뒤에 젓가락으로 구멍을 뚫는다. 그 다음에 전자렌지로 돌린다. 
- ... 

`-` 우리는 모든 방법을 나열할 순 없지만 모든방법을 이해할 수 있다. 왜냐하면 라면을 끓이는 어떤한 공통맥락을 우리는 파악하고 있기 때문. 

`-` 파이썬의 개발환경 구축 역시 사실 비슷하다. 공통맥락을 파악하면 쉽다. 

`-` 제 목적: 가장 맛있는 라면을 끓이는 하나의 레시피가 아니라 라면을 끓이는 레시피를 이해할 수 있는 공통맥락을 설명하고 싶어요 

> Note: 설치하는 방법이 다양한 이유? 파이썬이 인기있어서 + 다양한 방법으로 설치하면 각각의 장점이 뚜렷해서 (뻔한건 다양하지 않아요.. 감자삶는 레시피가 다양하진 않아요..)

`-` 주피터의 트릭 (실제로 주피터는 ipython에 기생할뿐 아무런 역할도 안해요) 

(1) 주피터가 실행 

(2) 새노트북을 생성 (파이썬으로 선택) 
- 컴퓨터는 내부적으로 ipython을 실행하고 그 ipython과 여러분들이 방금만든 노트북을 연결함. 

(3) 처음보이는 cell에 1+1 입력 -> 쉬프트엔터 -> 결과2가 출력
- 처음보이는 cell 하나 = 자동으로 열린 하나의 메모장 
- cell 1+1을 입력 = 메모장에 1+1을 적음 
- 쉬프트+엔터후 결과2출력 = cell의 내용을 복사 -> ipython에 붙여넣어요 -> ipython 계산된 결과를 복사 -> cell로 돌아와 붙여넣기 

(4) 새로운 cell을 추가하고 2+2입력 -> 쉬프트엔터 -> 결과4가 출력
- 새로운 cell 추가 = 새로운 메모장 추가 
- cell에 2+2입력 = 새로운 메모장에 2+2를 적음 
- 쉬프트+엔터후 결과4출력 = cell의 내용을 복사 -> ipython에 붙여넣기 -> ipython의 계산된 결과를 복사 -> cell로 돌아와 붙여넣기 

`-` 중요한 사실들 
- IDE는 내부적으로 연산수행능력이 없다. (생각해볼것: 왜 R을 꼭 설치하고 Rstudio를 설치해야 했을까?) 
- 주피터에서 커널을 재시작한다는 의미는 메모장은 모두 열려있는채로 ipython을 껐다가 다시 실행한다는 의미이다. 
- 주피터는 단순히 `메모장내용을 복붙하여 전달하는 기계`라고 생각할 수 있다. 이렇게 생각하면 주피터에 꼭 ipython을 연결할 이유는 없다. 실제로 주피터에 R을 연결하여 쓸 수 있다. 즉 하나의 IDE가 여러개의 언어와 연결될 수 있다. 
- jupyterlab이라는 프로그램은 크롬에 있는 내용과 ipython간의 통신을 제어하는 프로그램일 뿐이다.
