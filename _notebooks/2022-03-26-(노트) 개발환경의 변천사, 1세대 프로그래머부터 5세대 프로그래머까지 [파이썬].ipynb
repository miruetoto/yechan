#### 개발환경 구축방법이 많은 이유? 

`-` 파이썬 개발환경 구축은 수많은 방법이 있다. (검색하면 진짜 무수히 많이 나옴) 

`-` 이는 마치 라면의 레시피를 검색했을때 수많은 방법이 나오는것과 유사함. 
- 방법1: 스프를 먼저 넣고 끓인다음에 라면을 넣어야 합니다. 
- 방법2: 양은냄비에 물넣고 물 끓으면 스프넣고 라면 넣고 마지막에 계란을 넣는다. 
- 방법3: 웤에 물을 조금 넣고 라면과 스프를 넣고 자글자글 끓이다가 마지막에 참기름을 넣어요  
- 방법4: 먹다남은 삼겹살을 후라이팬에 볶은뒤에 물을 붓고 라면을 먹는다. 
- 방법5: 용기에 라면과 스프와 뜨거운물을 넣고 랩을 쒸운뒤에 젓가락으로 구멍을 뚫는다. 그 다음에 전자렌지로 돌린다. 
- ... 

`-` 우리는 모든 방법을 나열할 순 없지만 모든방법을 이해할 수 있다. 왜냐하면 라면을 끓이는 어떤한 공통맥락을 우리는 파악하고 있기 때문. 

`-` 파이썬의 개발환경 구축 역시 사실 비슷하다. 공통맥락을 파악하면 쉽다. 

`-` 제 목적: 가장 맛있는 라면을 끓이는 하나의 레시피가 아니라 라면을 끓이는 레시피를 이해할 수 있는 공통맥락을 설명하고 싶어요

### 3세대 프로그래머: 메모장 + anaconda prompt + python / 메모장 + ipython

#### 방법1: 메모장 + anaconda prompt + python 

`-` 전체적인 수행방법은 아래와 같다. 
- 메모장: 코드를 편집 수정 
- anaconda prompt: 메모장에 저장된 코드를 실행하고 결과를 확인 
- python: 코드의 결과를 그때그때 확인하며 짧은 단위의 코드를 만듬 

`-` 방법1의 아쉬운점 
- 아쉬움1: 메모장, anaconda prompt, python 3개를 동시에 사용하는 것이 난잡하다. 
- 아쉬움2: ananconda prompt에서 "python myadd.py"를 실행해도 실행내용이 python에 공유되지 않는다. (독립적인 두개의 python이 있는것임) 

`-` 상황극: 1~10까지의 수를 더함에 있어서 아래의 코드가 동작하지 않았다고 가정하자. 
```
### myadd.py 
total=0
for i in range(0,10):
    total = total +i 
    i=i+1 
print(total)
```
코드의 문제점을 살펴보다가 혹시 마지막순간에 i=10이 아니라 i=9가 들어가 있는것은 아닐까? 하는 의심이 들어서 i값을 확인해보고 싶어졌다. 

(시도1) python 콘솔에 i를 쳐본다. (이게 동작하지 않는다!) 

(시도2) myadd.py의 마지막줄을 print(i)로 수정하여 확인한다. 

#### 방법2: 메모장 + ipython

`-` 이 방법은 메모장 + anaconda prompt + python 조합보다 단순하면서 더 많은 기능을 제공한다.

(1) 주피터가 실행

(2) 새 노트북 생성 (파이썬으로 선택)

(3) 처음 보이는 cell에 1+1 입력 쉬프트+엔터를 누름 -> 결과 2 출력 
- 처음 보이는 cell 하나 = 자동으로 열린 메모장 하나
- cell에 1+1 입력 = 메모장에 1+1 적음 
- 쉬프트+엔터후 결과2출력 = cell의 내용을 복사 -> ipython에 붙어넣기 -> ipython 결과를 복사 -> cell로 돌아와 붙여넣기 

(4) 새로운 cell을 추가하고 2+2 입력 쉬프트+엔터 누름 -> 결과 4출력 
- 새로운 cell 추가 =  새로운 메모장 추가 
- cell 에 2+2 입력 = 새로운 메모장에 2+2 적음 
- 쉬프트+엔터후 결과4출력 = cell의 내용을 복사 -> ipython에 붙어넣기 -> ipython 결과를 복사 -> cell로 돌아와 붙여넣기

- 주피터는 단순히 `메모장내용을 복붙하여 전달하는 기계`라고 생각할 수 있다. 이렇게 생각하면 주피터에 꼭 ipython을 연결할 이유는 없다. 실제로 주피터에 R을 연결하여 쓸 수 있다. 즉 하나의 IDE가 여러개의 언어와 연결될 수 있다.

### 5세대 프로그래머: 가상컴퓨터(anaconda),  원격컴퓨터(server), 클라우드컴퓨터의 개념 등장 

`-` 지금까지는 ipython이 실행되는 컴퓨터와 크롬이 실행되는 컴퓨터가 동일하다는 전제였음. 

`-` 생각해보니까 어차피 ipython이 실행된 컴퓨터에 내가 크롬에서 입력한 명령이 "전달"되기만 하므로 꼭 같은 컴퓨터일 필요는 없음. 

`-` 모델1: 원격컴퓨터 

(준비상태) 
- 전북대컴퓨터: ipython 실행 + 이 컴퓨터는 인터넷 연결이 되어있음
- 우리집노트북: 크롬실행 + 이 컴퓨터는 인터넷 연결이 되어있음

(명령입력) 
- 우리집노트북에서 1+1을 크롬에 입력하고 쉬프트 엔터를 누름 

(우리집노트북->전북대컴퓨터) 
- 우리집노트북 내부의 어떤 프로그램은 1+1이라는 명령을 복사하여 카카오톡으로 전북대컴퓨터에 전달 
- 전북대컴퓨터 내부의 어떤 프로그램은 1+1이라는 명령을 카톡으로 받아서 그것을 ipython에게 전달 

(전북대컴퓨터->우리집노트북) 
- 전북대컴퓨터 내부의 ipython은 2라는 출력결과를 계산함

- 대학원생 노트북1개 (그냥 싸고 가벼운거) 

(아이디어)

- 대학원생은 (py38r40)에 접속하여 파이썬 3.8과 R 4.0을 설치한뒤 실습한다. 
- (base)는 예비용으로 아무것도 설치안한 깨끗한 상태유지

- 나는 (py39jl17)를 삭제하고 (base)로 부터 다시 새로운 컴퓨터 (py39jl17new)를 만들어 사용한다.
