### Problem setting 

`-` ${\cal G}=(V,E)$ or ${\cal G}=(V,E,W)$. 

`-` $x_i$: $i$번째 노드에 매겨진 $p$차원 입력

`-` $y_i$: $i$번째 노드에 매겨진 $q$차원 입력 

`-` layer: 

- 0층-1층: $H^{(1)}=f(H^{(0)},\bf{E})$ or $H^{(1)}=f({\bf X},\bf{E})$

`-` 입력벡터를 아래와 같이 설정한다. ${\bf E}{\bf X}$


`-` 예제 (1차원)

${\bf f}=\begin{bmatrix} 1 \\ 10 \\ 100 \\ 1000 \end{bmatrix},\quad {\bf E}=\begin{bmatrix} 0 & 1 & 0 & 0 \\ 1 & 0 & 1 & 0 \\0 & 1 & 0 & 1\\ 0 & 0 & 1 & 0  \end{bmatrix}$

```python
import numpy as np 
f=np.array([1,10,100,1000]); f.resize(4,1)
E=np.array([[0,1,0,0],[1,0,1,0],[0,1,0,1],[0,0,1,0]])
```

```python
E@f
```

- ${\bf E}{\bf f}$는 각 노드로 흘러들어온 유입양을 의미함
