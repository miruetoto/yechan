### 찰스미나드의 그림 (인류 역사상 가장 훌륭한 시각화)

![](https://upload.wikimedia.org/wikipedia/commons/2/29/Minard.png)

`-` 터프티의 평
- 지금까지 그려진 최고의 통계 그래픽일지도 모른다. 
- 여기에는 군대의 크기, 2차원 평면상의 위치, 군대의 이동방향, 모스크바에서 퇴극하는 동안의 여러날짜, 온도 -> 6차원의 변수 
- 백만번에 한번 이런 그림을 그릴수는 있겠으나 이러한 멋진 그래픽을 만드는 방법에 대한 원칙은 없다. $\to$ 미니멀리즘

`-` 왜 우수한 그래프일까? 
- 자료를 파악하는 기법은 최근까지도 산점도, 막대그래프, 라인플랏에 의존 
- 이러한 방법들은 고차원의 자료를 분석하기 쉽지 않음. 
- 미나드는 여러그림을 그리는 방법 대신에 한 그림애서 패널을 늘리는 방법을 선택함. 

#### 예제 

```python
x=[44,48,49,58,62,68,69,70,76,79] ## 몸무게 
y=[159,160,162,165,167,162,165,175,165,172] ## 키
g= 'f','f','f','f','m','f','m','m','m','m'
df=pd.DataFrame({'w':x,'h':y,'g':g})
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

`-` 미나드의 접근법

```python
sns.scatterplot(data=df,x='w',y='h',hue='g')
```

`-` 일반적인 사람들 (보통 색깔을 사용할 생각을 잘 못한다)

```python
figs = sns.FacetGrid(df, col='g') #figs = sns.FacetGrid(df, row='g')
figs.map(sns.scatterplot, 'w','h')
```

**바꾸는 방법2**

```python
df1=df.query(" g=='f' ")[['w','h']]
df2=df.query(" g=='m' ")[['w','h']]
```

```python
dflist=[df1,df2]
```

```python
pd.concat(dflist,keys=['f','m']).reset_index().iloc[:,[0,2,3]].rename(columns={'level_0':'g'})
```

`-` 어려운점: (1) 센스가 없어서 색깔을 넣어서 그룹을 구분할 생각을 못함 (2) 데이터를 변형할 생각을 못함 (3) 데이터를 변형할 생각을 한다고 해도 변형하는 방법을 모름. (그래서 엑셀을 킨다.. ) 
- (1)은 기획력 부족
- (2)는 데이터프레임에 대한 이해도 부족
- (3)은 프로그래밍 능력 부족
