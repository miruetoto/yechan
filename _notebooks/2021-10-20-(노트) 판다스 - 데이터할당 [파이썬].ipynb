### 컬럼추가 

```python
df=pd.DataFrame({'X1':[1,2,3,4,5],'X2':[2,3,4,5,6]})
df
```

`-` 새로운 칼럼 [3,3,3,3,3]을 추가하자. 

```python
df.assign(X3=[3,3,3,3,3]) ## 계산결과 
```

```python
df ## 원래 자료는 변화하지 않음 
```

```python
df= df.assign(X3=[3,3,3,3,3]) ## 계산결과를 update 
```

```python
df
```

`-` 길이가 다른것을 추가할수는 없다. 

```python
df.assign(X4=[3,3,3,3,3,5,5,5])
```

`-` 2개를 추가할 수 있다. 

```python
df.assign(X4=[4,4,4,4,4],X5=[5,5,5,5,5])
```

`-` 참고로 이것은 아래처럼 연달아 추가하는것과 같은 결과임 

```python
df.assign(X4=[4,4,4,4,4]).assign(X5=[5,5,5,5,5])
```

- 엔터가 엄청 큰 역할을 한다. 

`-` 좀 더 복잡한 함수일 경우는? 

`-` 방법1

```python
f= lambda x: (x-np.mean(x))/np.std(x) 
```

```python
f(df.X1)
```

```python
df.assign(X4=f(df.X1))
```

`-` 방법2

```python
df.assign(X4= lambda df: (df.X1-np.mean(df.X1))/np.std(df.X1))
```

```python
def f(x):
    if x<2: 
        y=2
    elif x<4:
        y=4
    else:
        y=6
    return y
```

```python
df.assign(X4=list(map(f,df.X1)))
```
