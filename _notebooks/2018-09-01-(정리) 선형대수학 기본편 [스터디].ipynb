### 벡터의 표현법

`-` 
${\bf X}=[{\bf x}_1,\dots,{\bf x}_p]$

`-`
${\bf X}=\begin{bmatrix} {\bf x}_{1\circ}^\top \\ ... \\ {\bf x}_{n\circ}^\top \end{bmatrix}$

   - 여기에서 ${\bf x}_{1\circ}^\top$는 ${\bf X}$의 첫번째 row-vector이다. 
   - 그러니까 ${\bf x}_{1\circ}$는 col-vector이다. 

`-` ${\bf X}=\left\{x_{ij}\right\}$
- 이인석교수님의 교재에서 나온  매트릭스 ${\bf X}$ 표현방법
- 이 역시 매우 유용한 표현이다. 처음에는 이러한 표현이 매우 헷갈렸지만 나중에는 이 표현을 안쓰는 것이 더 고통을 주므로 그냥 쓰기로 결정하였다. 


> Note:  앞으로는 혼란의 여지가 크지 않다면 벡터의 볼드체는 생략하겠다. 

### transpose

매트릭스 ${\bf X}$의 트랜스포즈는 보통 ${\bf X}^\top$로 표시한다. 혼란의 여지가 없을 경우 노테이션의 간략화를 위하여 ${\bf X}'$와 같이 표현할 수 있다. 매트릭스 ${\bf X}$가 복소행렬일 경우는 트랜스포즈를 사용하지 않고 **컨쥬게이트-트랜스포즈**(conjugate transpose)를 사용한다. 컨쥬게이트-트랜스포즈는 **에르미트-트랜스포즈**(Hermitian transpose), **에르미트-어드조인트**(Hermitian adjoint) 등으로 불리기도 한다. 기호로는 ${\bf X}^H$로 표시한다. 

 임의의 매트릭스 ${\bf X}_ {n \times p}$에 대하여 ${\bf X}'$는 아래와 같이 표현할 수 있다. 

`-` $
{\bf X}=[x_1, \dots, x_p], \quad 
{\bf X}'=\begin{bmatrix} x_1' \\ \dots \\ x_p' \end{bmatrix}.
$

`-` $
{\bf X}=\begin{bmatrix}x_{1\circ}' \\ \dots \\ x_{n\circ}' \end{bmatrix}, \quad 
{\bf X}'=[x_{1\circ}, \dots, x_{n\circ}].
$


`-` ${\bf X}=\left\{x_{ij}\right\}$, ${\bf X}^\top=\left\{x_{ji}\right\}$. 

> Tip: 트랜스포즈는 $L_2$-norm을 구할때 사용할 수 있다. $\|y\|_ 2^2=y^\top y$. 

### 행렬곱

`-` ${\bf A B} = [a_1 \dots a_n] 
	\begin{bmatrix}
		b_{1\circ}^\top \\ \dots \\ b_{n\circ}^\top 
	\end{bmatrix}
	=\sum_{i=1}^na_i b_{i\circ}^\top$

`-` ${\bf AB} =\begin{bmatrix}a_{1\circ}^\top \\ \dots \\ a_{n\circ}^\top \end{bmatrix}\Big[b_1\dots b_n\Big] = \begin{bmatrix}
		a_{1\circ}^\top b_1 & \dots & a_{1\circ}^\top b_n \\ \dots &\dots & \dots \\  a_{n\circ}^\top b_1 & \dots & a_{n\circ}^\top b_n\end{bmatrix}$


`-` ${\bf A B}=\begin{bmatrix}
		a_{1\circ}^\top \\ 
		\dots \\ 
		a_{n\circ}^\top
	\end{bmatrix}{\bf B}=\begin{bmatrix}
	a_{1\circ}^\top {\bf B} \\ 
	\dots \\ 
	a_{n\circ}^\top {\bf B}
\end{bmatrix}
$

`-` ${\bf AB}={\bf A} [b_1 \dots,b_n] = [{\bf A} b_1,\dots,{\bf A} b_n]$


### 행렬곱의 예제

`-` ${\bf X}{\beta}=\begin{bmatrix}x_{1\circ}^\top{\beta} \\ \dots \\ x_{n\circ}^\top {\beta}\end{bmatrix}
$

`-` ${\bf X}' {\bf X} = [x_{1\circ}^\top,\dots, x_{n\circ}^\top]\begin{bmatrix} x_{1\circ} \\ \dots \\ x_{n\circ} \end{bmatrix}=\sum_{i=1}^{n} x_{i\circ}^\top x_{i\circ}$


`-` ${\boldsymbol \Psi}{\boldsymbol\Lambda}{\boldsymbol\Psi}^\top= \sum_{i =1}^{p} \psi_i \lambda_i  \psi_i^\top$

`-` ${\bf U}{\bf D}{\bf V}^\top= \sum_{i =1}^{p} u_i d_i v_i^\top$

### 트레이스 

`-`  $tr({\bf A})=tr({\bf A}')$

`-`  $tr({\bf A}{\bf B}{\bf C})=tr({\bf B}{\bf C}{\bf A})=tr({\bf C}{\bf A}{\bf B})$

`-` 임의의 col-vector $y$에 대하여 아래식이 성립한다. 
$
y^\top y=tr(y y^\top)
$
