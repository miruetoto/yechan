### 리스트의 응용: 함수의 return으로 사용 

`-` R에서는 일반적으로 하나의 오브젝트(OOP에서 말하는 그 오브젝트 맞음)만을 리턴함 

`-` 하지만 벡터, 리스트등을 이용하면 다중리턴을 지원하것처럼 느껴지게 할 수 있음. 

````r
cal = function(x,y){
    return(list(add=x+y, sub=x-y, mul=x*y, div=x/y))
}
````

````r
cal(2,3)
````

`-` 이런일이 있으면 그냥 cal_add, cal_mul, cal_sub, cal_div처럼 함수를 4개 구현하면 되는것 아닌가? $\to$ 물론 그렇지만 코드가 지저분해 지니까.. 

`-` 보통 함수의 결과를 저장하여 아래와 같이 사용함 

````r
rslt= cal(4,3)
````

````r
rslt$add
````

````r
rslt$sub
````

`-` 왜 아래와 같이 쓰지 않을까? 

````r
cal(4,3)$add 
````

````r
cal(4,3)$sub
````

### 숙제 

`-` 숙제난이도를 좀 올릴게요. 

`-` 기본적인 자료형 vector, matrix, list에 대한 기능을 스스로 정리 및 요약해 볼 것 (제출의무없음) 
- 벡터의 길이를 재려면 어떻게 하나?

- 리스트의 원소에 접근하려면? 
- (리스트처럼) 벡터의 각 원소에도 이름을 넣고 이름을 통해 추출할 수 있을까? 할 수 있다면 왜 보통 그렇게 하지 않는가? 
- 그냥 $n\times 1$ 매트릭스 혹은 $1 \times n$ 매트릭스만 쓰면 벡터를 쓸 필요 없어보이는데 왜 굳이 벡터를 만들었을까? 

`-` 벡터 ${\bf x}$와 벡터 ${\bf y}$를 아래의 코드를 이용해 표준 정규분포에서 생성하라. (표준정규분포를 모른다면 질문하거나 네이버 구글에서 스스로 찾아볼것) 
```r
x=rnorm(100)
y=rnorm(100)
```
아래의 수식을 계산하는 코드를 작성하라. 

$$\frac{\sum_{i=1}^{100}(x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum_{i=1}^{100}(x_i-\bar{x})^2}\sqrt{\sum_{i=1}^{100}(y_i-\bar{y})^2}}$$

단 $\bar{x}=\frac{1}{100}\sum_{i=1}^{n}x_i$로 정의한다.
