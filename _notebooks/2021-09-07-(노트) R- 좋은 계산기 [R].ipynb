### R은 좋은 계산기이다. 

#### 2005 예비평가(나) 12월/평가원26(고3) 

어떤 호수에서 수면에서의 빛의 세기가 $I_0$일 때, 수심이 $d$m 인곳에서의 빛의 세기 $I_d$는 다음과 같이 나타내어진다고 한다. 

$$I_d = I_0 2^{-0.25d}$$

이 호수에서 빛의 세기가 수면에서의 빛의 세기의 25%인 곳의 수심은? (3점)

(1) 16m 

(2) 12m 

(3) 10m 

(4) 8m 

(5) 4m

(풀이1) 

```r
I0=1 
Id=I0*0.25
d=8
```

```r
I0*2^(-0.25*d) 
```

```r
Id
```

(풀이2) 

```r
y=I0*2^(-0.25*d)/Id #d=1,2,.. 
```

```r
print(d)
print(y)
```

(풀이3) 

함수라는걸 만들어보자.. 

d를 입력하면 알아서 y가 계산되면 좋겠다. 

```r
y_myfunc<-function(d){
    y=I0*2^(-0.25*d)/Id
    y
}
```

```r
y_myfunc(8)
```
