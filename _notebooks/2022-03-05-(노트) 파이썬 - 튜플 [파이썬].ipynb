### 튜플 

#### 리스트 vs 튜플

`-` 컨테이너형이라는 점에서 리스트와 비슷하다.
- 차이점1: [ ] 대신에 ( )를 사용한다. 
- 차이점2: 불변형이다. (값을 바꿀 수 없음)
- 차이점3: 하나의 원소를 선언할때는 (1,)와 같이 한다. 
- 차이점4: 의미가 명확할때는 튜플의 ( ) 를 생략가능하다. 

`-` 컨테이너형이라는게 무슨의미? 

````python
a=(4,6,"pencil",3.2+4.6j,[3,4]) # 원소의 타입이 통일될 필요가 없다.
````

````python
a[2]
````

````python
a[0:3]
````

````python
a[2]="Pencil"
````

`-` 예제: for문과 튜플

````python
idlist=[['guebin', '202112345','M'], 
        ['iu', '202154321','F'], 
        ['hodong', '201812321','M']]
````

````python
for i in idlist: 
    print(i)
````

````python
for name,studentid,sex in idlist: 
    print(name)
````

````python
for name,studentid,sex in idlist: 
    print(name,sex)
````

`-` 예제: for문과 튜플, dummy variable `_`

````python
for name, _, _ in idlist: # _는 자리를 지정해주는 역할을 함 
    print(name)
````

````python
for name, _, sex in idlist: 
    print(name,sex)
````

````python
for name, _ in idlist: # 이건 에러난다.. 
    print(name)
````

````python
for name, *args in idlist: # 하려면 이렇게 해야..
    print(name)
````

예제시작: 두점 사이의 거리를 구하는 함수를 만들어 보자. 

````python
import numpy as np
def distance(px,py,qx,qy):
    return np.sqrt((px-qx)**2+(py-qy)**2)
````

````python
p=(0,3)
q=(4,0)
distance(p,q)
````

(방법1) 

````python
px,py=(0,3)
qx,qy=(4,0)
distance(px,py,qx,qy)
````

(방법2)

````python
def distance2(p,q):
    px,py=p
    qx,qy=q 
    return np.sqrt((px-qx)**2+(py-qy)**2)

## 이건 진짜 나빴다..
# def distance2(p,q):
#     px=p[0]
#     py=p[1]
#     qx=q[0]
#     qy=q[1]
#     return np.sqrt((px-qx)**2+(py-qy)**2)
````

````python
distance2(p,q)
````

(방법3)

````python
distance(*p,*q) # 입력을 (px,py), (qx,qy) 형태로 넣기도 하고 
````

````python
distance(px,py,qx,qy) # 입력을 px,py,qx,qy 로 넣기도 한다. 
````
