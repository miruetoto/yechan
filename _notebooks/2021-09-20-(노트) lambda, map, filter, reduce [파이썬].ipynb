`-` 예제6

````python
def action(x): 
    return (lambda y: x+y)
````

````python
act = action(99)
````

````python
act(2)
````

이 예제를 응용하면 

````python
action = (lambda x: (lambda y: x+y))
````

````python
act = action(99)
````

````python
act(2)
````

괄호를 생략하여 더 간단히하면 

````python
action = lambda x: lambda y: x+y 
````

### map 

`-` 예제1

````python
def inc(x): return x+1 
````

````python
list(map(inc, [1,2,3,4]))
````

람다를 쓰면 

````python
list(map(lambda x: x+1, [1,2,3,4]))
````

`-` 예제2

````python
f = lambda x: 'X' in x 
````

````python
f('X1'),f('X2'),f('Y2'), f('Y3')
````

````python
list(map(f, ['X1','X2','Y3','Y4']))
````

아래와 비교해보자. 

````python
[f(x) for x in ['X1','X2','Y3','Y4']]
````

`-` 예제3

````python
pow(2,3)
````

````python
list(map(pow, [2,2,2,3,3,3],[0,1,2,0,1,2]))
````

아래와 비교해보자

````python
[pow(x,y) for x,y in zip([2,2,2,3,3,3],[0,1,2,0,1,2])]
````

`-` 예제4

````python
L=[lambda x: x+1, lambda x: x+2, lambda x: x+3]
````

````python
list(map(L,[100,200,300])) ## 에러난다..
````

리스트컴프리헨션은 가능함 

````python
[L[i](x) for i,x in zip([0,1,2],[100,200,300])]
````

`-` 종합: 리스트컴프리헨션과 비교하면, (1) 반복인덱스를 쓰지 않는 장점이 있는 반면 (2) 좀더 제약적으로 사용할수밖에 없다는 단점이 있음 

### filter

`-` 예제1

````python
list(filter(lambda x: x<3, [1,2,3,4]))
````

리스트컴프리헨션으로 하면

````python
[x for x in [1,2,3,4] if x<3]
````

위의 코드를 이해하기 어렵다면 아래를 관찰하라. 

````python
for x in [1,2,3,4]:
    if x<3: print(x)
````

**filter의 이해**

filter는 은근히 유용하다. (리스트 컴프리헨션이 너무 좋은것일 뿐..) 예제에서 배운 
```python
list(filter(lambda x: x<3, [1,2,3,4]))
```
는 아래와 같은 코드이다. 

````python
f=lambda x: x<3 
x=[1,2,3,4]
````

````python
fx=list(map(f,x))
fx
````

````python
import numpy as np
filtered=np.array(x)[fx]
filtered
````

````python
filtered.tolist()
````
