# 로지스틱 회귀분석 

### 자료

`-` $(x_i,y_i)$를 관측하였다. 

`-` $x_i$는 실수값을 가지는것 같은데, $y_i$가 $0$ 또는 $1$이다. 

`-` 자료를 대충 보니까 $x_i$가 커질수록 $y_i$가 $1$이 되는 것 같다. 

### 모형1: $y_i=I(\beta_0+x_i\beta_1+\epsilon_i >\gamma)$

`-` 선형회귀분석을 상상하면 아래와 같은 모형을 생각하면 좋을 것 같다. 

$y_i=I(\beta_0+x_i\beta_1+\epsilon_i >\gamma)$

`-` 여기에서 $\epsilon_i$ 는 어떠한 분포를 따르는 확률변수라고 하자. 그냥 정규분포라고 하자. 

`-` 이 모형은 주어진 자료를 설명할 수 있다. 

`-` 이슈1: 하지만 데이터로부터 이 모형에 적합한 $\hat{\beta_0},\hat{\beta_1},\hat{\gamma}$를 구하는 것이 까다로워 보인다. 

`-` 이유는 (1) $I(x)$ 함수가 $x=0$ 미분불가능하고 (2) $I(x)$ 함수가 $x<0$ or $x>0$ 에서의 기울기가 상수이다. (2)가 더 심각한 문제이다. 

### 모형2: $y_i=\frac{e^{\beta_0+x_i\beta_1+\epsilon_i}}{e^{\beta_0+x_i\beta_1+\epsilon_i}+1}$

`-` 모형1과 비슷한 모양을 가지는 아래의 모형을 고려하자. 

$y_i=\frac{\exp({\beta_0+x_i\beta_1+\epsilon_i})}{\exp({\beta_0+x_i\beta_1+\epsilon_i})+1}$

`-` 여기에서 $\epsilon_i$가 존재하는데, 이것도 어떠한 분포를 따르는 확률변수라고 하자. 그냥 정규분포라고 하자.

`-` 이슈1에서 자유로운 장점이 있다. 

`-` 이슈2: 이슈1을 해결하면서 새로운 이슈가 생겼는데 모형자체가 이상하다는 것이다. 

$y_i=\frac{\exp({\beta_0+x_i\beta_1+\epsilon_i})}{\exp({\beta_0+x_i\beta_1+\epsilon_i})+1}$

좌변은 항상 0 또는 1이고, 우변은 0 또는 1이 될 수 없다. 따라서 $=$ 자체가 성립하는 $i$가 없다. 

`-` 이슈2는 일단 덮어두고 진행하자.

#### 시도1

`-` 이슈2는 무시하고 진행하여 보자. 아래를 최소화하는 $\beta_0, \beta_1$을 구하면 될것 같다. 

$\sum_{i=1}^{n}(y_i-\hat{y_i})^2=\sum_{i=1}^{n}\big(y_i -\frac{e^{\beta_0+x_i\beta_1}}{e^{\beta_0+x_i\beta_1}+1} \big)^2$

`-` 이슈3: 하지만 위와 같은 손실함수는 이론적인 back ground가 부족하다는 단점이 있다. 

***이슈3-1: $L_2$ loss $\beta$에 대한 convex가 아니다.***

`-` 논의를 단순화 하기 위해서 

$l_2=\sum_{i=1}^{n}(y_i-\hat{y_i})^2=\sum_{i=1}^{n}\big(y_i -\frac{\exp(x_i\beta)}{\exp(x_i\beta)+1} \big)^2$

라고 가정하자. 

`-` $l_2$를 $\beta$로 두번 미분하여 항상 양수가 나오는지 따져보자. 

`-` 예비계산: 

$$\frac{\partial \hat{y_i}}{\partial \beta}
=\frac{e^{x_i\beta}x_i(e^{x_i\beta}+1)-e^{x_i\beta}e^{x_i\beta}x_i} {\big(e^{x_i\beta}+1\big)^2}
=\frac{e^{x_i\beta}(e^{x_i\beta}+1)-e^{x_i\beta}e^{x_i\beta}} {\big(e^{x_i\beta}+1\big)^2}x_i
=\frac{e^{x_i\beta}} {\big(e^{x_i\beta}+1\big)^2}x_i=\hat{y_i}(1-\hat{y_i})x_i$$

`-` 한번미분: 

$$
\frac{\partial l_2}{\partial \beta}=\sum_{i=1}^{n} -2(y_i-\hat{y_i})\hat{y_i}(1-\hat{y_i})x_i=\sum_{i=1}^{n}-2\big(\hat{y_i}^3-\hat{y_i}^2(1+y_i)+\hat{y_i}y_i\big)x_i
$$

`-` 두번미분: 

$$
\big(\frac{\partial}{\partial \beta}\big)^2l_2=\sum_{i=1}^{n}-2\big(3\hat{y_i}^2-2\hat{y_i}(1+y_i)+y_i\big)x_i\hat{y_i}(1-\hat{y_i})x_i
$$

`-` $\hat{y_i}(1-\hat{y_i})>0$ 이고 $x_i^2>0$ 이므로 아래의 부호만 따지면 된다. 

$$\sum_{i=1}^{n}\Big(3\hat{y_i}^2-2\hat{y_i}(1+y_i)+y_i\Big)$$

`-` 어떠한 조합에서도 위의 식이 항상 음수이어야 한다. 

`-` 그런데 $y_i=0$에서 $\hat{y_i}=0.999999$ 라면 위의 부호가 양수가 나올 수 있다. 따라서 위의식이 어떠한 상황에서도 항상 음수라고 주장할수 없다. 따라서 컨벡스가 아니다. 

***이슈3-2: 회귀모형에서의 MLE***

`-` 잠시 아래와 같은 모형을 다시 복습하자. 

$y_i=\beta_0+\beta_1x_i+\epsilon_i$

`-` $\epsilon_i \sim N(0,1)$ 이라면 $y_i \sim N(\beta_0+\beta_1x_i,1)$

`-` $y_i$에 대한 로그가능도함수를 구하면 

$-2l(\beta_0,\beta_1;y_i)=\sum_{i=1}^{n}(y_i-\beta_0-\beta_1x_i)^2$

와 같이 나온다. 그래서 

$\sum_{i=1}^{n}(y_i-\beta_0-\beta_1x_i)^2$

를 최소화한다는 논리전개가 힘을 받았다. 

#### 시도2 

`-` 적당히 계산을 하면 아래와 같이 표현할 수 있다. 

$$\log\big(\frac{y_i}{1-y_i}\big)=\beta_0+x_i\beta_1 +\epsilon_i \quad \cdots (\star)$$

`-` 이런식의 논리전개라면 아래를 최소화하는 적당한 $\hat{\beta_0},\hat{\beta_1}$을 구할것 같다. 

$${\tt loss}=L(\beta_0,\beta_1)=\sum_{i=1}^{n}\big(\log \frac{y_i}{1-y_i}- \beta_0-x_i\beta_1 \big)^2 \quad \cdots (\star\star)$$

`-` 묻어두었던 이슈2가 좀더 심각한 문제가 되어 되돌아온다. 

`-` 이슈4: 그런데 $y_i$는 $0$또는 $1$의 값을 가지지 않았나? 그렇다면 수식 $(\star)$의 좌변은 어떻게 되는지?

`-` 이슈5: 수식 $(\star\star)$는 어떻게 계산하는가?

$y_i$는 꼭정규분포일 필요는 없음.. 

### 모형3: $\begin{cases}y_i \sim Ber(\pi_i) \\ \pi_i=\frac{\exp(\beta_0+x_i\beta_1)}{1+\exp(\beta_0+x_i\beta_1)}\end{cases} $

`-` 생각할수록 오묘한데 여러가지 이슈가 한번에 해석되는 모형이다. 

`-` $y_i$를 고정하였을 경우 로그우도함수는 아래와 같다. 

$$l(\pi_i; y_i)=\log\Big(\prod_{i=1}^{n}\pi_i^{y_i}(1-\pi_i)^{1-y_i} \Big)=\sum_{i=1}^{n}\big(y_i\log \pi_i +(1-y_i)\log(1-\pi_i) \big)$$

`-` 손실함수를 $-l(\pi_i;y_i)$를 손실함수로 설정하면 이론적인 back-ground도 충분해진다.
