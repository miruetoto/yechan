### 0. 선행

`pmap`을 먼저 알아야 함

### 1. `pmap`복습 

```r
library(tidyverse)
mu<-c(10,10,10,5,5)
n<-c(1,4,6,3,5)
sigma<-c(1,1,1,5,5)
args<-tibble(sd=sigma,n=n,mean=mu)
print(args)
args %>% pmap(rnorm)
```

### 2. `mutate(...=invoke_map(...))`

위의 `pmap`예제의 경우 함수의 종류 `rnorm`은 고정되었음. 

대신에 그안의 `argument`만 바뀜 

하지만 `invokemap`의 경우 함수의 종류 `rnorm`까지도 자유자재로 바꿀 수 있다. 

```r
args2 <- tribble(
  ~f,      ~params,
  "runif", list(min = -1, max = 1),
  "rnorm", list(sd = 5),
  "rpois", list(lambda = 10)
)
print(args2)
```

```r
print(args2[[2]][[1]])
```

```r
print(args2[[2]][[2]])
```

```r
print(args2[[2]][[3]])
```

```r
args2 %>% mutate(simrslt=invoke_map(f,params,n=20)) 
```
