### np.unravel_index

`-` 사용법: `numpy.unravel_index(indices, shape, order='C')`

`-` 예제1

```python
X=np.array([[6,5,4],[3,2,1]])
X
```

```python
np.argmin(X)
```

```python
np.unravel_index(5,X.shape)
```

```python
X[1,2]
```

### np.ravel_multi_index

`-` 사용법: `numpy.ravel_multi_index(multi_index, dims, mode='raise', order='C')`

`-` 예제2

```python
np.ravel_multi_index((1,2),X.shape)
```
