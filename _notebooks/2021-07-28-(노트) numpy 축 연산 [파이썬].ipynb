### 2차원 

`-` 연습용배열 

````python
a2=np.array([[1,2,3],[4,5,6]])
a2
````

a2는 $2 \times 3$ 행렬이다. 

````python
a2.shape
````

`-` 예제1

`axis=0 `으로 연산하면 (2,3)중 첫번째 차원이 삭제된다. 
- 따라서 결과는 길이가 $3$ 인 벡터가 나옴
- 삭제된 차원값 = 2 $\to$ 결과의 각 원소는 2개의 숫자를 더한 것임 

````python
np.sum(a2,axis=0)
````

`-` 예제2

`axis=1` 으로 연산하면 (2,3)중 두번째 차원이 삭제된다. 
- 따라서 결과는 길이가 2인 벡터가 나옴 
- 삭제된 차원값 = 3 $\to$ 결과의 각 원소는 3개의 숫자를 더한 것임 

````python
np.sum(a2,axis=1)
````

### 3차원
