#### 예제: state_fruit2

(문제)

https://github.com/PacktPublishing/Pandas-Cookbook-Second-Edition/blob/master/data/state_fruit2.csv

`-` 문제의 깃헙주소로 들어가서 데이터를 관찰 $\to$ 좌측상단이 'State'로 채워져있음 $\to$ 데이터를 읽을때 `index_col=0` 옵션을 사용하지 않음. 

```python
url = 'https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook-Second-Edition/master/data/state_fruit2.csv'
state_fruit2 = pd.read_csv(url) # 문제의 주소로 들어가서 데이터를 살펴보자 -> 좌측상단이 채워짐 -> index_col=0 으로 불러오지 않음. 
state_fruit2
```

- index1: `{State: [Texas, Arizona, Florida]}` = row
- index2: `{??? :[Apple, Orange, Banana]}` = col

(풀이1)

```python
state_fruit2.melt()
```

```python
state_fruit2.melt(id_vars=['State'])
```

```python
- 
```

```python
state_fruit2.melt(id_vars=['State'],var_name='Fruit')
```

```python
state_fruit2.melt(id_vars=['State'],var_name='Fruit',value_name='Weight')
```

```python
state_fruit2.melt(id_vars=['State'],value_vars=['Orange','Apple','Banana'])
```

***melt를 이용하여 데이터를 깨끗하게 만들기 위한 작전 (변형하고 싶은 데이터의 형태가 직관적으로 떠오르지 않을 경우)***

`*` Step1: [State, Apple, Orange, Banana] 중 하나의 변수를 선택하여 index1로 설정한다. 
- 요령1: [State, Apple, Orange, Banana] 중에서 숫자를 값으로 가지는 변수명은 제외한다. $\to$ 여기에서 Apple, Orange, Banana가 제외됨. 
- 요령2: 여기서 선택되는 변수가 `G1`자리에 들어온다. `G1`자리에 어울릴만한 변수를 선택. 

`*` Step2: melt()사용. 필수옵션은 `id_vars`, `var_name`. 
- `id_vars`: Step1에서 선택한 변수이름, 즉 State. `G1`에 해당하는 변수명. 
- `var_name`: Step1에서 선택한 변수를 제외한 나머지 변수명(즉 [Apple, Orange, Banana])을 통합할 만한 변수이름을 '새롭게 정의함'. 이름이 잘 생각안나면 그냥 `G2`로 정하자. 
- `value_name`: 설정안해도 무방. `X1`자리의 이름임. `X1`에 적당한 이름을 주면 된다.
