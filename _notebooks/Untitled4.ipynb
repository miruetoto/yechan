```r
b=seq(10,5.5,-0.5)
alive=function(x){
    a=c()
    a[1]=x/100
    for(i in 2:10) a[i]=a[i-1]+1
    t=c()
    for(j in 1:10) t[j]=sum(a[1:j])
    print(rbind(a,b,t,a<=b))
    sum(a[a<=b])
    #if (sum(a[a<=b]) + b[!(a<=b)][1] >= 35) return(TRUE) else return(FALSE)
}
```

```r
alive(1)
```

```r
rslt=c()
for (i in 1:456) rslt[i]=alive(i)
```
