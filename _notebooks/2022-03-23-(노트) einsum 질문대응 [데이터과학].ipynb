```python
import numpy as np
```

### (2,3) @ (3,4) 행렬곱 

```python
A = np.array(range(2*3)).reshape(2,3)
B = np.array(range(3*4)).reshape(3,4) 
```

A는 (2,3) martix -> A매트릭스의 각 원소는 $a_{ij}$로 볼 수 있고 i=1,2 and j=1,2,3 

B는 (3,3) matrix -> B매트릭스의 각 원소는 $b_{jk}$로 볼 수 있고, j=1,2,3 and k=1,2,3,4 

```python
A @ B
```

(1,1)의 원소는 $a_{11}b_{11}+a_{12}b_{21}+a_{13}b_{31}$ 으로 볼 수 있음 

```python
A[0,0]*B[0,0] + A[0,1]*B[1,0] + A[0,2]*B[2,0]
```

유사한 논리로 (i,k)의 원소는 아래와 같이 볼 수 있음 

$\sum_{j=1}^{n} a_{ij}b_{jk}$

(재해석) 

가능한 모든 $a_{ij}$, $b_{jk}$에 대하여 $\sum_{j=1}^{n} a_{ij}b_{jk}$와 같은 연산을 정의한다면 결과는 $c_{ik}$로 나옴 = 최종차원 ik

마치 $a_{ij}, b_{jk} \to c_{ik}$ 로 가는 변환이 이루어진것 같으므로 `ij,jk -> ik`이라는 약어로 선형변환을 해석할 수 있음 

```python
np.einsum('ij, jk -> ik',A,B)
```

```python
np.einsum('ij, jk -> ik',B,A)
```

이건 수행불가능한데  $\sum_{j=1}^{n} b_{ij}a_{jk}$ 꼴로 연산을 정의할 수 없기 때문임. ($b$의 두번쨰축과 $a$의 첫번째축의 차원이 같지않음, b의 두번째축은 4, a의 첫번째 축은 2)

### (3,) @ (3,) -> None (무차원, 무축, 축이0개, scalar)

```python
a= np.array([1,2,3])
b= -a 
```

`-` 방법1

```python
a @ b 
```

```python
b @ a 
```

`-` 방법2

```python
np.einsum('i, i -> ',a,b)
```

### (3,) @ (3,) -> (3,3) 

`-` 외적

```python
a= np.array([1,2,3])
b= -a 
```

```python
a.reshape(3,1) @ b.reshape(1,3)
```

`-` 아인슈타인법

```python
np.einsum('i,j -> ij',a,b)
```

`-` a,b 길이 달라도 가능 

```python
a= np.array([1,2,3])
b= np.array([1,2,3,4])
```

```python
a.reshape(3,1) @ b.reshape(1,4)
```

```python
np.einsum('i,j -> ij',a,b)
```

(2,3,4,5) @ (4,) -> (2,3,5) 

```python
A = np.array(range(2*3*4*5)).reshape(2,3,4,5)
b = np.array(range(4)) 
```

```python
np.einsum('ijkl,k -> ijl',A,b)
```

A@b 는 불가능 

```python
A@b 
```

(2,3,4,5) @ (3,) -> (2,4,5) 

```python
A = np.array(range(2*3*4*5)).reshape(2,3,4,5)
b = np.array(range(3)) 
```

```python
np.einsum('ijkl,j -> ikl',A,b)
```

A @ b 혹은 b @ A 는 모두 불가능 

```python
A @ b 
```

```python
b @ A 
```

`-` 아래는 가능함 

```python
np.array(range(4)) @ A
```

```python
A @ np.array(range(5))
```

### np.einsum은 다양한 텐서연산으로 확장할 수 있음. 그런데 @는 2차원 매트릭스만 특화된 연산자임 

고차원일 경우에는 `@` 를 쓰는게 아님

2차원이하일 경우는 유연하게 해석하여 처리하는데 3차원 이상의 경우 맨 마지막 2개의 차원만 인정함 (저도 이건 몰랐어ㅛㅇ.. 저는 처음 2개의 차원으로 알고있었어요) 

헷갈리니까 2차원 넘어가면 np.einsum을 써야함 

그래도 @의 동작을 알아보자 

```python
A=np.array(range(2*3*4*5*6*7*9)).reshape(2,3,4,5,6,7,9) 
```

A는 (7,9)로 해석함 -> 앞에 7, 뒤에 9를 붙일 수 있음

```python
c= A @ np.array(range(9))
c.shape
```

```python
c= A @ np.array(range(7))
c.shape
```

```python
c= np.array(range(7))@ A 
c.shape
```

```python
c= A @ np.array(range(7))
c.shape
```
