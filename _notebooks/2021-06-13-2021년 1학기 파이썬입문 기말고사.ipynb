`-` 강화 성공시에 강화상태를 업그레이드 하는 코드예시 

```python
import pandas as pd
upgradestate=pd.DataFrame({'day0':[0]*100})
upgradestate
```

```python
i=1
attemptresult=np.random.binomial(n=1, p=0.7, size=100) 
upgradestate['day%s' % i]=upgradestate['day%s' % (i-1)]+attemptresult
upgradestate
```
