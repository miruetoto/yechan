### select column

#### `df`에서 하나의 열을 선택하는 방법

```python
dic={'X1':np.random.normal(0,1,5), 
     'X2':np.random.normal(0,1,5),
     'X3':np.random.normal(0,1,5)}
df=pd.DataFrame(dic)
df
```

`-` 방법1

```python
df.X1
```

`-` 방법2

```python
df['X1']
```

`-` 방법3

```python
df[['X1']]
```

- `df['X1']`는 series를 리턴하고 `df[['X1']]`는 dataframe을 리턴한다. 

`-` 방법4

```python
df.loc[:,'X1']
```

`-` 방법5

```python
df.loc[:,['X1']]
```

- `df.loc[:,'X1']`는 series를 리턴하고 `df[['X1']]`는 dataframe을 리턴한다. 

`-` 방법6

```python
df.iloc[:,0]
```

`-` 방법7

```python
df.iloc[:,[0]]
```

`-` 방법8 (`dfply`)

```python
df >> select(0)
```

```python
df >> select([0])
```

```python
df >> select('X1')
```

```python
df >> select(['X1'])
```

***요약***

- `df.X1` 이 가장 간단하고 편리함. 그런데 변수이름을 알고 있어야 한다는 단점이 있음. 
- `df.X1` 은 변수이름에 `.`이 포함된 경우는 사용불가. 그밖에도 변수이름에 공백이 있는등 이상한 형태의 변수이름에는 사용불가능.
- 변수이름을 모르면서 위치로 select하는 방법은 df.iloc을 이용하는 방법이 유일함. (dfply 패지키 사용하는 방법은 제외)

`-` 방법5

```python
df.iloc[:,0:3]
```

**(주의) loc에서의 슬라이싱은 마지막변수를 포함하지만 iloc에서는 포함하지 않음.**

`-` 아래를 비교하라. 

```python
df.iloc[:,0:3] ## 3은 포함되지 않는다. 
```

```python
df.loc[:,'X1':'X3'] ## 'X3'이 포함된다. 
```

`-` 그래서 column의 이름이 integer일 경우는 종종 매우 헷갈리는 결과가 나옴

```python
_df = pd.DataFrame(np.array([[1,2,3,4],[3,4,5,6],[5,6,7,8]]))
_df
```

```python
_df.loc[:,0:2]
```

```python
_df.iloc[:,0:2]
```

> note: 사실 이것은 일부러 헷갈리게 예제를 구성한것이다. 실제로는 헷갈리는 상황이 많이 발생하지 않는다. 왜냐하면 보통 위의 예제와 같은 경우에는 ndarray 로 처리하고 colname이 있는 경우만 데이터프레임으로 처리하기 때문
