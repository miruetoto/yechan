```python
_df.loc[:,0:2]
```

```python
_df.iloc[:,0:2]
```

> note: 사실 이것은 일부러 헷갈리게 예제를 구성한것이다. 실제로는 헷갈리는 상황이 많이 발생하지 않는다. 왜냐하면 보통 위의 예제와 같은 경우에는 ndarray로 처리하고 colname이 있는 경우만 데이터프레임으로 처리하기 때문
