#### 방법3: `df.loc[:,'X1']`

`-` 세번째는 `loc`을 이용한 방법이다. 

`-` `df.loc[:,'X1']`은 기본적으로 컬럼이름에 "."이 포함되어도 사용할 수 있다. 

````python
df.loc[:,'X1'] # series를 리턴
````

````python
df.loc[:,['X1']] # dataframe을 리턴
````

`-` 또한 여러개의 컬럼을 호출할 수 있다. 즉 아래가 가능하다. 

````python
df.loc[:,['X1','X2']]
````

`-` 슬라이싱을 통하여 여러개의 컬럼을 선택할수 있다. 

````python
df.loc[:,'X1':'X3']
````

>Note: (단점) 이 방법의 단점은 컬럼의 `index-set`을 정의하고 접근하는것은 불가능하다. 예를들어 `홀수번째 컬럼들만 출력하는 코드`를 짤 수 없다. 

#### 방법4: `df.iloc[:,0]` 

`–` 마지막 방법은 `iloc`을 사용하는 것이다. 

`–` `iloc`역시 `loc`과 마찬가지로 슬라이싱이 가능하다. 

`–` 하지만 `loc`의 경우와 그 결과가 다르다. 1,2,3번째 변수를 `loc`으로 추출하고 싶을 경우 

````python
df.loc[:,'X1':'X3']
````

와 같이 하지만 같은 결과를 `iloc`으로 구현하려면 

````python
df.iloc[:,0:3]
````

와 같이 해야한다. 즉 `loc`의 경우 양쪽인덱스가 모두 포함되지만 `iloc`의 경우 마지막 인덱스는 포함되지 않는다. 

***`iloc`의 장점***

`–` `iloc`의 장점은 아래와 같이 `index-set` 을 정의하고 접근하는것도 가능하다. 예를들어 홀수번째 번수들만 출력하는 코드를 다음과 같이 구현가능하다.

````python
index1=[0,2]
df.iloc[:,index1]
````
