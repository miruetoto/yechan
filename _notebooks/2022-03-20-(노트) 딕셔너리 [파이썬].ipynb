### list, tuple, str의 공통점 

`-` 모두 시퀀스형이다 $\to$ 원소의 위치번호로 인덱싱이 가능함 

`-` 물론 이들은 시퀀스형이라는 공통점 이외에 차이점이 있음. 잠깐정리해보자. 

***시퀀스형의 카테고리***
- 컨테이너형: list, tuple 
- 균일형: str 
- 가변형: list 
- 불변형: tuple, str 

***표로정리하면***

||컨데이너형|균일형|
|:-:|:-:|:-:|
|가변|list|-|
|불변|tuple|str|

`-` 그런데 시컨스형이 아닌 1차원 자료형도 있다. 즉 원소의 위치번호로 인덱싱이 불가능하다는 의미. 

`-` 왜 이런게 필요할까? 
- 벡터에서 원소를 뽑는건 정보의 모음에서 정보를 검색하는것과 같다. 
- 정보를 `순서`대로 나열한뒤 그 `순서`를 이용하여 검색하는 방법은 유용하다. 
- 하지만 정보의 `키워드`를 기억해서 그 `키워드`를 이용해 검색하는 방법 역시 유용하다.

***카카오톡 대화내용검색***

(상황1) 오늘 `아침`에 와이프가 `뭔가`를 카톡으로 부탁했었음. 그런데 그 내용이 기억나지 않는다?  

(상황2) `개강전에` 동료교수와 함께 `저녁약속`을 카톡으로 잡았음. 그런데 그 언제인지 기억나지 않는다? 

(상황3) `오늘아침` 동료교수와 함께 `점심약속`을 카톡으로 잡았음. 그런데 그 장소가 기억나지 않는다? 

`-` 순서대로 정리된 자료를 검색할때는 시컨스형이 유리하다. 그런데 키워드로 검색하고 싶을 경우는 딕셔너리 타입이 유리하다.

**(keys, values, items)**

`-` .keys()는 딕셔너리의 키를 리턴한다. 

```python
_keys=score.keys()
_keys
```

```python
type(_keys) # 리턴하는 자료형은 이상한것임 
```

```python
list(_keys) # 아무튼 리스트화 가능한 자료형임
```

`-` .values()는 딕셔너리의 값들을 리턴한다. 

```python
_values=score.values()
_values
```

```python
type(_values) # 리턴된 자료형은 이상한것임
```

```python
list(_values) # 아무튼 이것도 리스트화 가능
```

`-` .items()는 딕셔너리의 (키,값)을 리턴한다. 

```python
_pairs = score.items()
_pairs
```

```python
type(_pairs) # 이것도 자료형은 이상함 
```

```python
list(_pairs) # 리스트화 가능
```
