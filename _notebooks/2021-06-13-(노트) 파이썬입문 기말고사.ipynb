`-` p의 확률로 0 혹은 1을 100개 뽑는 함수예시

```python
import numpy as np 
np.random.binomial(n=1, p=0.3, size=100) # 0.3 의 확률로 0 혹은 1을 추출
```

```python
np.random.binomial(n=1, p=0.7, size=100) # 0.7 의 확률로 0 혹은 1을 추출
```

`-` 강화 성공시에 강화상태를 업그레이드 하는 코드예시 

```python
import pandas as pd
upgradestate=pd.DataFrame({'day0':[0]*100})
upgradestate
```

```python
i=1
upgradestate['day%s' % i]=upgradestate['day%s' % (i-1)]+np.random.binomial(n=1, p=0.7, size=100) 
upgradestate
```
