### \# 1. 
아래와 같은 특징을 가지는 아이템이 있다고 하자. 이 아이템은 총 5회에 걸쳐서 강화(upgrade)될 수 있는데 강화의 성공확률은 50%라고 하자. 강화를 실패하면 강화가 되지 않는다고 하고 강화가 5번성공하면 더 이상 강화가 진행되지 않는다고 하자. 아래는 강화를 진행하였을때 각 강화상태를 설명한 예시이다. 

|시도횟수|강화성공여부|강화상태|
|:-:|:-:|:-:|
|1|강화실패|+0|
|2|강화성공|+1|
|3|강화실패|+1|
|4|강화성공|+2|
|5|강화성공|+3|
|6|강화성공|+4|
|7|강화실패|+4|
|8|강화성공|+5|


이 아이템을 5번강화성공한 상태로 만들기 위하여 필요한 시도횟수의 평균을 파이썬을 통한 시뮬레이션을 활용하여 근사 추론하라. 

> Note: 위의 예시의 경우 8회의 시도만에 5번강화성공상태로 도달하였음.

이 경우 아이템을 5번강화성공한 상태로 만들기 위하여 필요한 시도횟수의 평균을 파이썬을 통한 시뮬레이션을 활용하여 근사추론 하라.
