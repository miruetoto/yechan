### 새로운 열 직접입력 

`-` 방법1: concat 이용 

```python
df= pd.DataFrame({'a':[1,2,3],'b':[2,3,4]})
df2 = pd.DataFrame({'c':[3,4,5]})
```

```python
pd.concat([df,df2],axis=1)
```

`-` 방법3: `.loc`이용 

```python
df= pd.DataFrame({'a':[1,2,3],'b':[2,3,4]})
df
```

```python
df.loc[:,'c'] = [3,4,5] 
df
```

`-` 방법4: `.assign` 이용 

```python
df= pd.DataFrame({'a':[1,2,3],'b':[2,3,4]})
df
```

```python
df.assign(c=[3,4,5])
```

`-` 방법5: `.eval`이용

```python
df.eval('c=[3,4,5]')
```

### 기존열을 변환하여 추가

```python
df = pd.DataFrame({'x':[1,2,3],'y':[-1,-3,-5]})
df
```

`-` 방법2,3: 

```python
df['r']=np.sqrt(df.x**2 + df.y**2)
df.loc[:,'r']= np.sqrt(df.x**2 + df.y**2)
```

`-` 방법4

```python
df.assign(r=np.sqrt(df.x**2 + df.y**2))
```

`-` 방법5

```python
df.eval('r=sqrt(x**2 + y**2)') # 왜 가능하지?
```
