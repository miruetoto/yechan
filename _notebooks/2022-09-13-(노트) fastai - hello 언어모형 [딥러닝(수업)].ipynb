```python
import pandas as pd
```

```python
from fastai.text.all import *
```

```python
df = pd.DataFrame({'text':['h e l l o . h e l l o ! h e l l o ? h e l l o ?? h e l l o !!']*20000})
df
```

```python
dls = TextDataLoaders.from_df(df,text_col='text',is_lm=True)
```

```python
dls.show_batch()
```

```python
lrnr = language_model_learner(dls,AWD_LSTM,metrics=accuracy)
```

```python
lrnr.fit(7)
```

```python
lrnr.predict("h e l", n_words=20)
```
