### imports

```python
from IPython.display import HTML
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
def show(fig): 
    return HTML(fig.to_html(include_plotlyjs='cdn',include_mathjax=False, config=dict({'scrollZoom':False})))
```

### fig_dict

`-` fig의 핵심은 어차피 data, layout 임 

```python
fig1 = px.line(x=["a","b","c"], y=[1,3,2], title="sample figure")
show(fig1)
```

```python
print(fig1)
```

`-` 이 핵심정보들만 dict로 재정의한다면? 

```python
fig2_dict = dict({
    "data": [{"type": "bar",
              "x": [1, 2, 3],
              "y": [1, 3, 2]}],
    "layout": {"title": {"text": "A Figure Specified By Python Dictionary"}}
})
```

```python
fig2_dict
```

```python
#
# pio.show(fig2_dict) 이게 그려짐..
```

`-` 원래는 이런느낌

```python
fig2=go.Figure(fig2_dict) # dict -> figure object로 변화!
type(fig1) == type(fig2) ## 같은타입
```

```python
show(fig2)
```

### go

`-` 아래도 가능 

```python
fig3 = go.Figure(
    data=[go.Bar(x=[1, 2, 3], y=[1, 3, 2])],
    layout=go.Layout(
        title=go.layout.Title(text="A Figure Specified By A Graph Object")
    )
)
show(fig3)
```

`-` 결국 figure는 data, layout, frame의 조합인데 이것은 다시 go에서 만들어지는 object들의 조합이다. 

***go.layout을 이용하여 그림의 height, width를 바꾸는 예시***

```python
fig4 = go.Figure(
    data=[go.Bar(x=[1, 2, 3], y=[1, 3, 2])],
    layout=go.Layout(height=300, width=800)
)
```

```python
show(fig4)
```
