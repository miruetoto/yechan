### imports

```python
from IPython.display import HTML
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
def show(fig): 
    return HTML(fig.to_html(include_plotlyjs='cdn',include_mathjax=False, config=dict({'scrollZoom':False})))
```

### fig

`-` fig의 핵심은 어차피 data, layout 임 

```python
fig1 = px.line(x=["a","b","c"], y=[1,3,2], title="sample figure")
show(fig1)
```

```python
print(fig1)
```

`-` 이 핵심정보들만 dict로 재정의한다면? 

```python
fig2_dict = dict({
    "data": [{"type": "bar",
              "x": [1, 2, 3],
              "y": [1, 3, 2]}],
    "layout": {"title": {"text": "A Figure Specified By Python Dictionary"}}
})
```

```python
fig2_dict
```

```python
#
# pio.show(fig2_dict) 이게 그려짐..
```

`-` 원래는 이런느낌

```python
fig2=go.Figure(fig2_dict) # dict -> figure object로 변화!
type(fig1) == type(fig2) ## 같은타입
```

```python
show(fig2)
```
