## 산후조리(삼칠일) 
### 자궁회복 
`-` 분만이후 8시간 이후에는 걸어라. 우리의 경우 저녁부터는 걸어야함

`-` 한쪽방향으로 눕지말고 엎드리고 오른쪽 왼쪽으로 번갈아가면서 누워야함. 

`-` 고양이자세를 하면 좋음.  

`-` 수술통증을참고 움직어야 자궁이 빠르게 회복한다. 

### 훗배앓이

`-` 자궁의 퇴축운동

`-` 젖을 먹일떄 아프다 (젖을 먹이면 옥시토신이 나와서 자궁의 수축을 촉진)

`-` 배를 따듯하게 하고 손으로 살살 만지면 통증이 준다. 

`-` 통증이 너무 심하면 진통제를 먹어라. 모유수유중이면 타이레롤 

`-` 분만직후는 아니다. 일주일이 지나고 훗배앓이가 생길수도. 

### 오로
`-` 자궁에서 피가 나온다. 자연스러운 현상. 회음 상처에서 나오는 분비물은 아님

`-` 1주만에 끝나기도 하고 8주가 지나도 계속 나오기도 한다. 

`-` 오로의 색은 개인에 따라 다르다. 적색, 갈색, 황색, 백색 

`-` 누웠다가 일어날때 많은 양의 응고된 오로가 한번에 나올수도 있다. 

### 복부가 쭈글함
`-` 고열량의 산후음식 -> 복부비만 

`-` 3~7일 동안 복부관리 매우 중요

`-` 제왕절개의 경우 복부마사지가 필요함. 
- 배꼽을 중심으로 오일을 바르고 시계도는 방향으로 둥글게 원을 그리면서 지그시 누르며 마사지 한다. 
- 옆구리 왼쪽에서 오른쪽으로 지그재그로 골반쪽으로 내려오며 마사지한다. 
- 분만 1주일 이후에는 따듯한 찜질 

`-` 산후음식은 굽고, 찌고, 삶고, 무침등의 조리법을 사용. 무조건 칼로리를 낮게해라. 

### 골반근육운동 (산후 1일차)

`-` P20~21 

### 튼살
`-` 로션발라라. 별다른 답은 없어보임

### 회음절개부위 통증 (해당없는듯)
`-` 10일까지 아픔

`-` 좌욕을 하면 좋음. 좌욕후 적외선치료기를 회음부위에 댄다. 공기에 노출시켜 상처를 건조시킨다. 통증이 줄어듬. 

`-` 얼음찜질식도 좋음 

`-` 수틀리면 병원가라. 회음부위가 욱신욱신 쑤시거나 열이나는듯 후끈거리거나 빨갛게 부어올랐거나 바늘로 콕콕 짜르듯이 아픈경우. 병원부터가라. 

### 젖몸살
`-` 젖이 탱탱하게 붓는다. 

`-` 젖을 자주 짜야한다. 특히 초유를 빨리 짜야 젖몸살이 없다. 그런데 제왕절개하면 초유분비가 늦을수 있음. 해결책은없음. 

`-` 적당히 아프면 온찜질 진짜 아프면 냉찜질을 한다. 냉찜질 전에 맨소레담을 바르면 좋다. 

`-` 마사지를 해주면 좋다. P26. 

`-` 젖은 하루에 8회, 시간은 4시간 간격. 

### 전유/후유

`-` 아이의 변이 묽다면 후유를 못먹어서 그렇다. 

`-` 유방을 싹싹 문질러서 마사지하고 젖을 먹여라.

`-` 젖을 20분쯤 먹이면 후유가 나오는데 여기에 영양이 많다. 아이도 이 젖을 먹으면 배불러한다. 

`-` 처음에 나오는 젖은 짜서 버리고 후유만 먹여도 좋다. 

### 어혈 

`-` 피가안통해서 냉/열/저림/마비/통증의 증상이 나오는것. 

`-` 실내온도는 24~26 

`-` 어혈의 원인은 추위 더위 습도등 다양함

`-` 땀으로 어혈을 제거할수 있다. 쌍화탕을 먹고 방바닥의 온도를 높이고 이불을 두껍게 하고 땀을내자. 

`-` 적당한 땀으로 어혈을 제거한 이후에 (삼칠일정도) 호박죽, 호박중탕, 호박전먹어라. 
- 호박의 효능: 산후변비 및 면역력회복 

### 산후부종

`-` 젖몸살과함께 시작함. 

`-` 3~8kg의 부종 

`-` 부종의 원인: 더위, 과식, 운동부족, 자궁염증

`-` 부종예방
- 충분한휴식 
- 음식을 싱겁게
- 과식하지말것 (소화하기 쉬운음식은 많이 먹어도 괜찮은듯, 소화기계의 에너지소비가 커질수록 몸이 피곤해져서 부종이 생김)
- 알맞게 움직이되 수틀리면 언제든지 쉬어라. 

`-` 그나마 제왕절개분만 산모가 부종이 적음. 제왕절개의 경우 수술이후 이튿날부터 실내를 걸으면 부종방지의 큰 역할을 함. 

`-` 부종도 방바닥을 따듯하게 하는게 좋다. 방바닥을 따듯하게 하면 너무 덥지 않은 상황에서 땀을 빼기에 유리하다. 긴옷을 입는다고 하여 꼭 내복을 입지말고 헐렁한 옷을 입어야 상쾌하게 몸조리를 할 수 있다. 

`-` 부종에는 잉어탕이 좋다. 기름이 적고 질적으로 좋은 단백질이기때문. 그밖에도 흰살생선, 두부, 한치, 기름기가 제거된 쇠고기, 돼지고기 좋음. 산모에게 가장좋은 단백질은 알부민으로 몸안의 노폐물을 빨리 제거하여 부종을 줄인다.
- 알부민: 달걀흰자 + 참기름. 소금을 넣지 않고 먹는다. 


## 산후 후유증 
### 산후풍 

`-` 증상: 손발이 차갑다. 창분쪽의 몸이 시리다. 

`-` 원인: 어혈, 혈액순환 방해 

`-` 땀을 많이 내라. 부항으로 제거하라. 적당한 운동으로 혈액순환을 도와라. 

`-` 몸이 시리면 그 부분만 따듯하게 하라. 

`-` 찜질팩, 통증부위, 원적외선

`-` 몸이 따듯해지는 음식을 먹어라. 고깃국, 사골국, 인삼차, 꿀차.

`-` 흑염소와 녹용은 피를 맑게 하는데 방해가된다. 

`-` 피가 맑아지는 국, 반찬, 야채를 먹어라. 

### 이가 시리다. 
`-` 분만이후에 칫솔질해라.

`-` 설탕이 있는음식은 피해라.

`-` 탄산음료 피해라. 

`-` 딱딱하고 질긴음식 피해라. 

`-` 등푸른생선, 녹차, 우유 ,치즈, 과일, 채소를 먹어라. 

### 뼈마디가 아프다. (류마티스)
`-` 특별한일이 없으면 3-6개월에 없어진다. 

`-` 그 이후에도 지속되면 정형외과로 가라. 

`-` 급성은 냉찜질, 만성은 온찜질. 

### 관절통증

`-` 6개월이후에도 다양하게 아플수 있음. 

`-` 발뒤꿈치, 발바닥, 손가락, 손목, 어꺠, 골반, 목 --> 진짜아프면 정형외과. 머 특별한방법은 예방법은 없음. 

`-` 대처방법: 아픈부분을 안쓰면된다. 발뒤꿈치, 발바닥이 아프면 발을 움직이지 말고 손이 아프면 손가락을 움직이지말고, 어꺠가 아프면 아이를 안지 않는다. 

### 허리가 아프다.
`-` 디스크(정형외과), 생리전요통(산부인과), 근육(그냥휴식) 

### 어지럽다.
`-` 먹어라. 미역국을 많이먹어라. 6끼 먹기도 한다고 함. 

`-` 뚱보가 될수가 있으니 적은양의 식사를 자주먹어라.

`-` 빈혈: 철분섭취를 늘려라. 달걀노른자, 소의간, 순대, 녹황색채소 
 - 철분은 과일의 비타민C와 함께 섭취해야 흡수가 잘된다. 오렌지주스랑 같이먹어라. 
 - 음식과 철분제제를 같이 먹는게 좋다. 
 - 철분제를 먹을때는 우유랑 먹지마라. (철부흡수방해). 

### 눈이 나빠졌다. 눈이 침침하다. 

`-` 비타민A: 소간, 녹화앵채소, 우유, 버터, 치즈, 마가린, 간유, 당근, 시금치, 겨자잎, 파슬리 

`-` 칼륨: 사과, 바나나, 꿀

`-` 칼슘: 우유, 치즈, 달걀, 멸치 

`-` 비타민B: 땅콩, 효모, 쌀껍질, 생굴, 우유, 돼지고기, 대부분의 채소 

### 눈출혈 

`-` 휴식 

`-` 소화쉬운음식 

`-` 눈에 찬물과 수건 

### 요실금

## 신생아 

### 태지 

`-` 하얀 기름기

`-` 목욕할때 억지로 벗기려하지마라. 

`-` 거즈에 베이비오일을 적셔 머릿속을 살살 닦아주어라. 

`-` 옛 어른들은 달걀노른자로 머리속 태지를 닦았음. 

`-` 태지를 닦아낸 이튿날부터 비누로 씼어라. 

`-` 로션을 발라라. 

### 걱정안해도 되는 증상들 (원래그런것) 

`-` 대부분 원래 그런증상들. 해결책은 시간이 지나가면 없어진다.

`-` 구체적으로는 아래와 같음. 
- 울면 얼굴이 파래진다. 
- 머리가 말랑거림
- 귀가 접히거나 찌그러져도 나중에 제모양을 찾음. 
- 팔은 W모양, 다리는 M모양 
- 배가 볼록하다. (원래그럼. 배부른것 X. 밥많이줘도괜찮음)
- 빠르게 숨을 쉰다. 

### 허준의 동의보감 (아이돌볼때 10가지 방법)

`-` 동의보감 10원칙
1. 등 따뜻
2. 배 따뜻
3. 발 따뜻
4. 머리 서늘
5. 가슴 시원
6. 이상하고 무서운것을 들려주지 말고 보이지 말아라
7. 따듯한 음식으로 비위를 따뜻하게 하라. 
8. 울음을 그치지 않을때 젖이나 음식을 주지 말아라. 
9. 가루 음식을 먹이지 말아라.
10. 목욕을 자주 씻기지 말아라. 


### 아이돌볼때 주의점 

`-` 찬 우유를 먹이면 장이 건강해진다는 속설이 있는데 사실 그렇지 않음. 신생아에게 위험할 수 있다. 

`-` 심하게 우는 아이에게는 젖을 물리지마라. 수유가 기도로 넘어갈 수 있다. 

`-` 실내온도 22이하는 안되고 28이상은 위험함. (24~26)

`-` 이불로 아이를 덮지마라. (외풍이 심한 옛날과 다름. 코막혀 질식함)

`-` 엄마가 어디에서 바라보든 아이의 얼굴이 보이도록 하라. 

`-` 이불은 어깨이상으로 올리지 마라. 

`-` 엄마 아빠 사이에 아이를 눕히지 마라. 

`-` 아기를 어른 침대에 눕히지마라. 아이는 푹신함 침대보다는 딱딱한 바닥에 눕히는 것이 좋다. 
- 어른침대에 눕히면 아기가 떨어질 위험이 있음
- 침대와 벽사이에 낄 수도 있음. 
- 아기전용 침대 아니면 바닥에 눕혀라. 

`-` 옆으로 재워라. 그래야 질식을 막을 수 있다. 
- 신생아는 항상 구토의 위험이 있다. (소화기계구조상 그럼)
- 엄마가 잠들때 애가 토를 해서 막힐수 있는데 이를 방지하려면 옆으로 눕혀야함. 

`-` 속싸개를 해라. 
- 자궁에서 아이는 좁은곳에 있었음 -> 갑자기 팔다리가 자유로워지면 불안해함 
- 속싸개로 싸주어야함. 
- 젖먹이거나 기저기갈떄는 속싸개를 벗기고 팔다리를 꾹꾺 눌러서 혈액순환및 운동해라. 
- 쭉쭉이를 하고 엄마손을 힘있게 잡는 아기가 건강한 아기

`-` 미숙아
- 부모의 철저한 손씻기를 생활화 (미숙아는 면역력과 저항력이 약함)
- 미숙아가 아니더라도 3키로 이전까지는 조심하는게좋다. 
- 외부출입도 자제. 특히 호흡기계 질환은 출입자제
- 온습도계. 온도계는 25~26. 가습기는 65% 
- 미숙아는 피부가 약하므로 이틀에 한번 목욕해야 한다. 목욕이후는 꼭 로션바를것
- 젖은 조금씩 자주. 젖먹이고 20분동안은 안고 등을 쓸어 트림을 시킴. 
- 다른아이보다 특별히 조용히할것 

`-` 손씻는법
- 흐르는물에 씼는다.

`-` 소독시기: 배꼽이 젖었을때나 목욕이후에 수시로 소독하여 잘 마르도록 한다. 

`-` 배꼽이 떨어져도 탈락부위가 아물때까지 약 15일이 걸리므로 소독을 계속한다.

`-` 배꼽이 마를때까지는 기저귀를 배꼽에서 접어서 배꼽이 잘 마르게한다.

- 무릎에 엎드리게 하고 등을 쓸어준다.

### 젖몸살 

`-` 유방마사지: 손에 힘을 뺴고 마사지하라. 
- 방법1: 유방을 크게잡고 둥글게 흔든다. (원래 찜질타월로 둘러싸면 좋다)
- 방법2: 지그재그로 둥글게 흔든다. 
- 방법3: 유선을 전체적으로 쓸어주며 유선을 따라 나선형으로! (3손가락)
- 방법4: 유관마사지. 산모를 앉히고 유방가장자리부터 모아주듯이 둥글게 마사지. 
- 방법5: 유관훑기. 오일이나 로션을 바르고 훑어준다. 
- 방법6: 유즙저장관 짜내기. 유룬주위를 눌러 저장되어있는 젖을 흘러내리게 한다. 

### 젖꼭지 트러블

#### 유두균열 
`-` 유두균열: 빠는힘이 강해 젖꼭지가 갈라짐

`-` 유두균열 해결을 위한 관리법 
- 젖을 먹인후 물로 유두를 닦아낸다. 비누나 바스사용 하지마라. 
- 닦아낸 유두를 물기없이 바싹 말린다. 
- 유두를 눌러 젖을 짜내어 유두에 묻힌다음 다시 바싹 말린다.  

`-` 유두보호캡 착용

`-` 유두균열 예방법

#### 유선염 
`-` 겨드랑이나 유방에 혹이 있음 (부분적으로 젖이 막히면 나옴)

`-` 해결책: 유방마사지 종합으로 수행하고 유축기로 젖을 짜거나 아이에게 젖을 물린다. 


#### 유두물집, 유두끝이 빨갛게 될때는
`-` 유두를 깊숙히 물려라. 

#### 모유에 피가나온다. 
`-` 출혈은 아이에게 영향 X

`-` 붉은 빛의 젖은 버리고 나중의 젖을 물리면 된다. 

`-` 15일 뒤면 없어진다. 


### 참젖과 물젖

`-` 물젖은 아이에게 물을 주는것. 갈증이 있으면 힘차게 물젖을 빨것이다.

`-` 어느정도 갈증이 풀리면 약하게 빤다. 참젖은 천천히먹음. 이때 천천히 먹는다고 하여 반대쪽으로 돌리지마라. 

`-` 참젖을 먹이려면 유방을 충분히 마사지하고 먹여라. 

### 모유보관

`-` 실온은 4~6시간, 냉장보관은 72시간, 냉동보관은 6개월 

`-` 냉장 냉동된 젖을 먹일때는 60도의 물에 중탕하여 먹여라. 

### 모유량

`-` 유방크기와 관련없음. 분만 1주일 안에 아기에게 젖을 얼마나 자주 먹였느냐에 따라 달라진다. 

`-` 3시간 간격으로먹여라. 

`-` 산모가 영양이 많이 있어야함. 단백질, 탄수화물, 무기질. 

`-` 칼슘을 많이 먹어야함. 

`-` 수용성비타민 B,C 를 많이먹어야함. (과일) 

`-` 불포화지방산을 많이먹어라. 우족, 돼지족. 하지만 장기간 기름진 음식을 먹으면 아기는 엄마젖을 느끼하게 생각한다. 

`-` 주스 많이먹을것

`-` 유선을 따듯하게 하여 모유나오는 길을 넓혀라. 따듯한 물로 샤워한후에 모유수유하면 잘나온다. 찜질팩이나 따듯한 수건으로 유방을 문질러라. 

`-` 전복죽 먹으면 좋다. 

`-` 아래의 음식은 모유량을 줄인다. 
- 엿기름, 인삼, 꿀차
