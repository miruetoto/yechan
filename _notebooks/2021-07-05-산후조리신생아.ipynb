## 삼칠일 
### 자궁회복 
`-` 분만이후 8시간 이후에는 걸어라. 우리의 경우 저녁부터는 걸어야함

`-` 한쪽방향으로 눕지말고 엎드리고 오른쪽 왼쪽으로 번갈아가면서 누워야함. 

`-` 고양이자세를 하면 좋음.  

`-` 수술통증을참고 움직어야 자궁이 빠르게 회복한다. 

### 훗배앓이

`-` 자궁의 퇴축운동

`-` 젖을 먹일떄 아프다 (젖을 먹이면 옥시토신이 나와서 자궁의 수축을 촉진)

`-` 배를 따듯하게 하고 손으로 살살 만지면 통증이 준다. 

`-` 통증이 너무 심하면 진통제를 먹어라. 모유수유중이면 타이레롤 

`-` 분만직후는 아니다. 일주일이 지나고 훗배앓이가 생길수도. 

### 오로
`-` 자궁에서 피가 나온다. 자연스러운 현상. 회음 상처에서 나오는 분비물은 아님

`-` 1주만에 끝나기도 하고 8주가 지나도 계속 나오기도 한다. 

`-` 오로의 색은 개인에 따라 다르다. 적색, 갈색, 황색, 백색 

`-` 누웠다가 일어날때 많은 양의 응고된 오로가 한번에 나올수도 있다. 

### 복부가 쭈글함
`-` 고열량의 산후음식 -> 복부비만 

`-` 3~7일 동안 복부관리 매우 중요

`-` 제왕절개의 경우 복부마사지가 필요함. 
- 배꼽을 중심으로 오일을 바르고 시계도는 방향으로 둥글게 원을 그리면서 지그시 누르며 마사지 한다. 
- 옆구리 왼쪽에서 오른쪽으로 지그재그로 골반쪽으로 내려오며 마사지한다. 
- 분만 1주일 이후에는 따듯한 찜질 

`-` 산후음식은 굽고, 찌고, 삶고, 무침등의 조리법을 사용. 무조건 칼로리를 낮게해라. 

### 골반근육운동 (산후 1일차)

`-` P20~21 

### 튼살
`-` 로션발라라. 별다른 답은 없어보임

### 회음절개부위 통증 (해당없는듯)
`-` 10일까지 아픔

`-` 좌욕을 하면 좋음. 좌욕후 적외선치료기를 회음부위에 댄다. 공기에 노출시켜 상처를 건조시킨다. 통증이 줄어듬. 

`-` 얼음찜질식도 좋음 

`-` 수틀리면 병원가라. 회음부위가 욱신욱신 쑤시거나 열이나는듯 후끈거리거나 빨갛게 부어올랐거나 바늘로 콕콕 짜르듯이 아픈경우. 병원부터가라. 

### 젖몸살
`-` 젖이 탱탱하게 붓는다. 

`-` 젖을 자주 짜야한다. 특히 초유를 빨리 짜야 젖몸살이 없다. 그런데 제왕절개하면 초유분비가 늦을수 있음. 해결책은없음. 

`-` 적당히 아프면 온찜질 진짜 아프면 냉찜질을 한다. 냉찜질 전에 맨소레담을 바르면 좋다. 

`-` 마사지를 해주면 좋다. P26. 

`-` 젖은 하루에 8회, 시간은 4시간 간격. 

### 전유/후유

`-` 아이의 변이 묽다면 후유를 못먹어서 그렇다. 

`-` 유방을 싹싹 문질러서 마사지하고 젖을 먹여라.

`-` 젖을 20분쯤 먹이면 후유가 나오는데 여기에 영양이 많다. 아이도 이 젖을 먹으면 배불러한다. 

`-` 처음에 나오는 젖은 짜서 버리고 후유만 먹여도 좋다. 

### 어혈 

`-` 피가안통해서 냉/열/저림/마비/통증의 증상이 나오는것. 

`-` 실내온도는 24~26 

`-` 어혈의 원인은 추위 더위 습도등 다양함

`-` 땀으로 어혈을 제거할수 있다. 쌍화탕을 먹고 방바닥의 온도를 높이고 이불을 두껍게 하고 땀을내자. 

`-` 적당한 땀으로 어혈을 제거한 이후에 (삼칠일정도) 호박죽, 호박중탕, 호박전먹어라. 
- 호박의 효능: 산후변비 및 면역력회복 

### 산후부종

`-` 젖몸살과함께 시작함. 

`-` 3~8kg의 부종 

`-` 부종의 원인: 더위, 과식, 운동부족, 자궁염증

`-` 부종예방
- 충분한휴식 
- 음식을 싱겁게
- 과식하지말것 (소화하기 쉬운음식은 많이 먹어도 괜찮은듯, 소화기계의 에너지소비가 커질수록 몸이 피곤해져서 부종이 생김)
- 알맞게 움직이되 수틀리면 언제든지 쉬어라. 

`-` 그나마 제왕절개분만 산모가 부종이 적음. 제왕절개의 경우 수술이후 이튿날부터 실내를 걸으면 부종방지의 큰 역할을 함. 

`-` 부종도 방바닥을 따듯하게 하는게 좋다. 방바닥을 따듯하게 하면 너무 덥지 않은 상황에서 땀을 빼기에 유리하다. 긴옷을 입는다고 하여 꼭 내복을 입지말고 헐렁한 옷을 입어야 상쾌하게 몸조리를 할 수 있다. 

`-` 부종에는 잉어탕이 좋다. 기름이 적고 질적으로 좋은 단백질이기때문. 그밖에도 흰살생선, 두부, 한치, 기름기가 제거된 쇠고기, 돼지고기 좋음. 산모에게 가장좋은 단백질은 알부민으로 몸안의 노폐물을 빨리 제거하여 부종을 줄인다.
- 알부민: 달걀흰자 + 참기름. 소금을 넣지 않고 먹는다. 


## 산후 후유증 
### 산후풍 

`-` 증상: 손발이 차갑다. 창분쪽의 몸이 시리다. 

`-` 원인: 어혈, 혈액순환 방해 

`-` 땀을 많이 내라. 부항으로 제거하라. 적당한 운동으로 혈액순환을 도와라. 

`-` 몸이 시리면 그 부분만 따듯하게 하라. 

`-` 찜질팩, 통증부위, 원적외선

`-` 몸이 따듯해지는 음식을 먹어라. 고깃국, 사골국, 인삼차, 꿀차.

`-` 흑염소와 녹용은 피를 맑게 하는데 방해가된다. 

`-` 피가 맑아지는 국, 반찬, 야채를 먹어라. 

### 이가 시리다. 
`-` 분만이후에 칫솔질해라.

`-` 설탕이 있는음식은 피해라.

`-` 탄산음료 피해라. 

`-` 딱딱하고 질긴음식 피해라. 

`-` 등푸른생선, 녹차, 우유 ,치즈, 과일, 채소를 먹어라.

`-` 급성은 냉찜질, 만성은 온찜질.

`-` 대처방법: 아픈부분을 안쓰면된다. 발뒤꿈치, 발바닥이 아프면 발을 움직이지 말고 손이 아프면 손가락을 움직이지말고, 어꺠가 아프면 아이를 안지 않는다.
