### About this doc 

`-` 더렛책의 마코프체인을 요약한다. 

### 절대공간 $(\Omega_o,{\cal F}_{\infty}, P_{\mu})$

`-` $X_i$는 $(\Omega,{\cal F}) \to (S, {\cal S})$인 랜덤변수이다. 
- $\Omega=\{node1, node2, \dots, node24\}$
- $S=\{1,2,\dots,24\}$

`-` $X_1,X_2$는 동일한 가측공간 $(\Omega,{\cal F})$를 공유하지만 $X_1$, $X_2$가 같은 분포를 가질 필요는 없다. 따라서 $X_i$에 대응하는 확률공간은 $(\Omega,{\cal F},P_i)$라고 볼 수 있다. 

`-` $X_i$에 대응하는 확률공간은 $(\Omega,{\cal F},P_i)$로 잘 정의되는데, 

$
{\bf X}=(X_0,X_1,X_2,\dots)
$

와 같은 무한차원 확률벡터를 잘 정의시키는 공간도 존재할까? $X_1,X_2,\dots$이 iid열이 아니므로 그렇게 쉽게 주장할 수 없어보인다. 하지만 일단 존재한다고 하자. 그리고 이러한 확률공간을 $(\Omega_o,{\cal F}_\infty, P_{\mu})$라고 정의하자. 구체적으로는 아래와 같다. 
- $\Omega_o=\Omega \times \Omega \times \Omega \times \dots$
- ${\cal F}_{\infty}={\cal F} \times {\cal F} \times {\cal F} \times \dots$ 
- $P_{\mu}=P_0 \times P_1\times P_2 \times \dots$ 

여기에서 $P_0$는 $X_0$의 distribution 이다. 즉 무한차원 확률변수열 ${\bf X}=(X_0,X_1,X_2,\dots)$ 에 대응하는 확률공간이 $(\Omega_o, {\cal F}_{\infty},P)$ 이다. 

`-` 보통은 $\Omega=S$라고 둔다. 따라서 $(\Omega_o,{\cal F}_{\infty},P_{\mu})$는 아래와 같이 정의한다. 
- $\Omega_o=\Omega \times \Omega \times \Omega \times \dots$
- ${\cal F}_{\infty}={\cal F} \times {\cal F} \times {\cal F} \times \dots$ 
- $P_{\mu}=P_0 \times P_1\times P_2 \times \dots$ 

여기에서 $P_0$는 $X_0$의 distribution 이다. 즉 무한차원 확률변수열 ${\bf X}=(X_0,X_1,X_2,\dots)$ 에 대응하는 확률공간이 $(\Omega_o, {\cal F}_{\infty},P)$ 이다. 

### 마코프체인의 정의

`(정의)` 아래를 만족하는 확률변수열 $X_1,X_2,X_3,\dots$를 마코프체인이라고 한다. 

$\forall B \in {\cal S}: \quad P(X_{n+1} \in B |{\cal F}_n )=P(X_{n+1}\in B | X_n)$

여기에서 ${\cal F}_n=\sigma(X_0,X_1,\dots,X_n)$. 

`-` 쉽게 말하면 $n+1$시점에 대한 확률분포(distribution)을 알기 위해서 시점1에서 시점n까지의 정보를 모두 기록한 ${\cal F}_n$이 필요한 것이 아니라 바로 직전시점의 정보인 $X_n$만 필요하다면 그것이 마코프체인이라는 것이다.
