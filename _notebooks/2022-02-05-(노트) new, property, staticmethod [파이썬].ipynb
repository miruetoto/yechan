### `__new__`

`-` `__new__`: 사용자가 `클래스()`를 입력하였을 경우 동작하는 명령을 정의한다. // `__new__`를 따로 정의하지 않아도 자동으로 어떤 동작이 정의되어있는데 그 동작은 '클래스->인스턴스화'. 

````python
class Report: 
    def __init__(self):
        self.a= '과목명: 파이썬입문'
        self.b= '학번: 2022-12345'
````

````python
Report.__new__?
````

- 따로 사용자가 정의하지 않아도 자동으로 `__new__`가 정의되어있음. 

````python
report=Report() ## Report() 는 Report.__new__() 와 같음
report
````

````python
report.a
````

````python
report.b
````

`-` `__new__` 의 디폴트 동작은 아마 아래와 같을 것임

````python
class Report: 
    def __new__(cls):
        return super().__new__(cls) ## 클래스->인스턴스화 하라는 명령 
    def __init__(self):
        self.a= '과목명: 파이썬입문'
        self.b= '학번: 2022-12345'
````

````python
report=Report()
report
````

````python
report.a
````

````python
report.b
````

`-` 엄청특별한 경우 아니면 `__new__` 로 구현가능한 기능은 모두 `__init__` 를 통해서도 구현가능 함 

### `__init__`으로는 구현불가능하지만 `__new__`로는 구현 가능한 것 

솔직히 예제가 너무 생각안나서 어거지로 만들었어요.. 아직도 쓰는이유가 `__new__` 확 와닿지는 않아요.. 실제로 쓰는것도 질문이전에는 제가 못봤고요.. 

`-` 예제: 경우에 따라서 인스턴스를 생성할수도 생성하지 않을수도 있는 클래스를 만들고싶은 경우

````python
class Report: 
    def __new__(cls):
        mychoice = np.random.choice(['레포트를 낼까','말까'])
        return super().__new__(cls) if mychoice == '레포트를 낼까' else None
    def __init__(self):
        self.a= '과목명: 파이썬입문'
        self.b= '학번: 2022-12345'
````

````python
report=Report()
report
````

- 위의셀을 반복실행해보자 $\to$ 경우에 따라서 report1이 만들어지기도 하고 아니기도 함 

이 기능은 `__init__`으로는 구현불가능한데 그 이유는 `__init__`함수는 순서상으로 인스턴스가 생성된 이후에만 실행되기 때문.

`-` 예제2: 모티브: (1) 괄호쓰기 귀찮음 + (2) 메소드를 마치 값처럼 쓰고싶음.. 

````python
class Report: 
    def __init__(self):
        self.a= '과목명: 파이썬입문'
        self.b= '학번: 2022-12345'
    @property
    def alen(self):
        return len(self.a)
````

````python
report=Report()
````

````python
report.a
````

````python
report.alen
````

````python
report.a='파이썬'
````

````python
report.alen
````

### static method

`-` 아래와 같이 상상하자. 

- Report = 클래스 = 레포트표지를 만드는 양식문서 
- report = 인스턴스 = 레포트표지 

`-` 예제1

````python
class Report: 
    a= '과목명: 파이썬입문'
    def __init__(self):
        self.a=Report.a
    def change(self,newname):
        self.a=newname
````

````python
report=Report() ## 양식문서를 컨트롤c+v해서 레포트 표지를 만듬 
report.a ## 양식문서에서는 기본제목이 파이썬입문으로 되어있음.. 
````

````python
report.change('과목명: 전산통계') ## 레포트표지변경 
report.a
````

````python
report2=Report() ## 양식문서를 변경한건 아니므로 새로운 인스턴스를 만들면 여전히 제목은 파이썬입문
report2.a
````

````python
Report.change('과목명: 전산통계') ## change는 메소드이므로 클래스에서는 사용불가능 
````

`-` 예제2

````python
class Report: 
    a= '과목명: 파이썬입문'
    def __init__(self):
        self.a=Report.a    
    def change(self,newname):
        Report.a=newname
````

````python
report=Report()
report.a
````

````python
report2=Report()
report2.a
````

note: 사실상 `report.change('과목명: 데이터시각화')` 의 기능은 `Report.change('과목명: 데이터시각화')` 로 생각해야 의미상맞을것 같은데 실제로 `report.change('과목명: 데이터시각화')`를 입력하면 에러난ㄷ.ㅏ

왜냐하면 change는 메소드이므로 클래스 자체에서는 쓸수없기때문 

`-` 예제3

예제2는 의미가 좀 이상한데 양식문서의 제목을 바꾸기 위해서는 인스턴스를 만든뒤에 메소드를 적용해야함. 

만약에

```python
report.change('과목명: 데이터시각화') 
```

이 아니라

```python
Report.change('과목명: 데이터시각화') 
```

처럼할수있다면 덜헷갈릴텐데.. 



````python
class Report: 
    a= '과목명: 파이썬입문'
    def __init__(self):
        self.a=Report.a    
    @staticmethod
    def change(newname):
        Report.a=newname
````

````python
report=Report()
report.a
````

````python
Report.change('과목명: 데이터시각화') ## 클래스의 제목변경 
report.a ## 인스턴스에는 영향없음. 
````

````python
report2=Report() ## report2부터는 바뀐제목이 적용
report2.a 
````
