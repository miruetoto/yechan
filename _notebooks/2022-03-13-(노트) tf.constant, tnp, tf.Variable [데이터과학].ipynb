#### tf.constant는 불편하다. 

`-` 불편한 제약 
- 텐서의 모든 원소는 같은 dtype을 가져야 한다. (이건 뭐 사실 그렇게 까지 제약은 아님)
- 텐서는 한번 생성되면 불변이다.  
- 텐서끼리의 연산에서 자료형이 다르면 형변환이 되지 않는다. 

`-` 텐서의 모든 원소는 같은 dtype를 가진다. 

```python
tf.constant([1,2,3])
```

`-` 텐서는 한번 생성되면 불변이다. (assingment 불가능) 

```python
a= tf.constant([1,2,3])
a
```

```python
a[0]
```

```python
a[0]=111
```

`-` 텐서의 연산에서 자료형이 다르면 연산이 불가능하다. 

```python
tf.constant([1,2,3])+tf.constant([1,2,3])
```

```python
tf.constant([1.0,2.0,3.0])+tf.constant([1.0,2.0,3.0])
```

```python
tf.constant([1,2,3])+tf.constant([1.0,2.0,3.0])
```

`-` 심지어 같은 float형도 dtype이 다른경우 연산이 불가능 

```python
tf.constant([1.0,2.0,3.0])
```

```python
tf.constant([1.0,2.0,3.0],dtype=tf.float64)
```

```python
tf.constant([1.0,2.0,3.0]) + tf.constant([1.0,2.0,3.0],dtype=tf.float64)
```

```python
tf.add(tf.constant([1,2,3]),tf.constant([1,2,3])) # 벡터+벡터 
```

```python
tf.add(tf.constant([1,2,3]),tf.constant(1)) # 벡터+스칼라
```

#### tf.stack 

`-` stack은 dim을 항상 증가시키는 방향으로 결합한다. 

`-` (3,) stack (3,)  

```python
a = tf.constant([1,2,3])
b = tf.constant([3,4,5])
```

```python
tf.stack([a,b])
```

- 오.. 
- a,b를 (1,3) 으로 해석했군 
- 그런데 누가 a,b rowvector라고 했어? 

```python
tf.stack([a,b],axis=1)
```

- a,b를 (3,1)로 해석하자 

`-` (2,3,4,5) stack (2,3,4,5) -> (?,?,?,?,?) 

```python
a=tf.constant(np.array(range(2*3*4*5)).reshape(2,3,4,5))
b=-a
```

**case1:** (2,3,4,5) 를 (1,2,3,4,5) 로 해석 
- 최종차원은 ( ,2,3,4,5) stack ( ,2,3,4,5) = (2,2,3,4,5) // 0번째 축이 변하네요 

```python
tf.stack([a,b],axis=0)
```

**case2:** (2,3,4,5) 를 (2,1,3,4,5) 로 해석 
- 최종차원은 (2, ,3,4,5) stack (2, ,3,4,5) = (2,2,3,4,5) // 1번째 축이 변하네요 

```python
tf.stack([a,b],axis=1) 
```

**case3:** (2,3,4,5) 를 (2,3,1,4,5) 로 해석 
- 최종차원은 (2,3, ,4,5) stack (2,3, ,4,5) = (2,3,2,4,5) // 3번째 축이 변하네요 

```python
tf.stack([a,b],axis=2) 
```

**case4:** (2,3,4,5) 를 (2,3,4,1,5) 로 해석 
- 최종차원은 (2,3,4, ,5) stack (2,3,4, ,5) = (2,3,2,2,5) // 4번째 축이 변하네요 // 끝에서 2번째 축이 변한다 

```python
tf.stack([a,b],axis=-2) 
```

**case5:** (2,3,4,5) 를 (2,3,4,5,1) 로 해석 
- 최종차원은 (2,3,4,5, ) stack (2,3,4,5, ) = (2,3,4,5,2) // 5번째 축이 변하네요 // 마지막 축이 변한다 

```python
tf.stack([a,b],axis=-1) 
```
