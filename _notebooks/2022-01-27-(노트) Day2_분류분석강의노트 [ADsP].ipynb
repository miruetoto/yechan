### 인공신경망

`-` 역사 
- 인간의 뇌를 모방하고 싶다 $\to$ 뉴론을 관찰 $\to$ 입력 + 활성화함수조합을 사용 $\to$ 이걸 많이 조합하면 인간과 비슷한 지능이 나온다. 
- 못한다 (할수있다면 입력층, 출력층 이외에 은닉층이라는걸 만들어야 하는데 은닉층을 만들면 학습할 방법이 없다) $\to$ 인공지능 암흑기 
- 1차암흑기에 아무것도 못함. (신경망 연구하면 거의 미친과학자 취급받았음, 연구를해도 발표안함) 
- 힌튼이 계속 연구 + 수학천재 르쿤이 힌튼의 제자로 합류 $\to$ 역전파기법이 등장하며 은닉층을 효율적으로 학습하는 방법론 등장 
- 역전파기법을 이용하여 학습하니 종종 학습이 너무 느렸는데 (기울기소멸, sigmoid의 한계) 이 문제를 해결하기 힘들었음 $\to$ 딜레마: 은닉층이 많을수록 (깊은신경망일수록) 성능이 올라가지만 은닉층이 많을수록 기울기소멸문제가 더 심해져 학습은 더 힘들어짐 
- SVM등장 $\to$ 신경망 2차암흑기 
- 이후에는 ReLU등의 개발 + GPU의 등장 + 등등으로 다시 연구가 활성화 $\to$ 딥러닝의 시대 

`-` 중요한 것들 
- 입력층, 은닉층, 출력층 / 활성화함수 
- 역전파기법 
- 기울기소멸, Sigmoid/Softmax, ReLU

`(참고)` 갑자기 왜 미분이 등장? 
- 인공신경망에서 학습이란 오차(오분류)를 최소화하는 어떠한 계수값을 찾는 일이라 할 수 있음. 
- 오차를 함수식으로 만들면 함수의 최소값을 찾는 문제랑 수학적으로 동일함. 
- 함수의 최소값은 미분을 통하여 찾을 수 있음. 
- 그런데 신경망의 오차함수는 너무 복잡해서 미분하기 어려움. 
- 그런데 체인룰(합성함수의곱)이라는 것을 이용하면 복잡한 하나의 미분을 여러개의 쉬운 미분으로 쪼개서 계산할 수 있음

### 분류모형 성과평가 

`-` 시험에 잘나오는것: 오분류포와 평가지표계산, ROC 커브, 향상도곡선 

`-` 내가 만든 모형이 얼마나 좋은지 평가받고 싶다? $\to$ 내가 만든 모형의 오분류표를 만들고 평가지표를 계산해보자, ROC 커브도 그려봐서 면적을 살펴보자, 향상도곡선을 그려봐서 랜덤모델과 비교해보자. 

#### ROC커브

`-` x축은 FPR(1-특이도), y축은 TPR. (x축과 y축을 설명하는 곡선은 ROC 커브.. 왜? 설명하기 좋음) 

`-` ROC커브아래의 면적(Area Under the Curve)이 클수록 좋은모형임 = AUC가 클수록 좋은 모형 

`-` p 321 그림에서는 왼쪽이 제일 좋음 

`-` 핵심어: 면적, x/y축 

#### 향상도곡선 

`-` **랜덤모델**과 비교하여 해당모델의 성과가 얼마나 향상되었는지 **구간별**로 파악하기 위한 그래프. 

`-` 좋은모델: 큰 향상도값에서 시작하여 급격히 향상도가 감소 

`-` 핵심어: 랜덤모델, 구간, 등급, 상위XX (상위등급, 상위데이터) 

#### 이익도표 

`-` **랜덤모델**과 비교하여 해당모델의 성과가 얼마나 향상되었는지 **구간별**로 파악하기 위한 도표. 

`-` 0.5에서 cut-off 

`-` 좋은모델: 상위구간에서 큰 값 향상도(lift value)를 가짐 

`-` 핵심어: 향상도곡선의 키워드 + cut-off

> Note: 이익도표와 향상도의 차이는 안물어봄

`-` 퇴사자예측 (회사나감=T, 안나감=F) 
- 10000명중에 100명 퇴사할까말까.. 
- 최규빈: 귀찮은데 그냥 다 퇴사안한다고 예측하자. (99퍼의 정분류율) --> 실제퇴사자 100명중 최규빈이 퇴사한다고 예측한사람은 0명 
- 인사팀: 실제 퇴사자를 퇴사자라고 맞춰야 의미있다.. 실제퇴사한사람이 100명인데 그중에서 최규빈이 퇴사한다고 말한사람이 100명있어야 좋은거 아니냐? 우리는 앞으로 리컬이라는 변수를 보겠다.. $\frac{TP}{TP+FN}$.. (니 모델을 돌렸을때 퇴사한다는 사람이 일단 나와야할거 아님, 재현되어야 할것 아님) 
- 최규빈: 구조를보니까 FN만 줄이면 되겠네? 귀찮은데 수틀리면 퇴사한다고 예측하자. 한 500명 퇴사할것 같다 해야하겠다. -->  실제퇴사자 100명중 최규빈이 퇴사한다고 예측한사람 100명, 따라서 이경우 리컬=1.
- 인사팀: 아니 멀쩡한사람까지 퇴사한다고 하면 어떡함? 니가 예측한 500명중에 실제로는 100명만 퇴사함. 이건 과녁중앙에 맞추라고 하니까 500발 기관총을 쏘아서 100발만 중앙에 들어간 꼴임. 정확도가 0.2인 셈인데 이건 너무 낮다. 우리는 앞으로 너의 예측에 대한 프리시즌도 평가하겠다. 기준은 $\frac{TP}{TP+FP}$임
- 최규빈: 리컬을 보겠다는것이야? 애큐러시를 보겠다는 것이야? 이랬다 저랬다함. 
- 인사팀: 둘다보겠다는 의미임, 그냥 F1스코어를 보겠음. F1스코어는 둘의 조화평균이고 두 값이 모두 골고루 커야지 큰값이 나옴. 

`-` 안중요한것들 
- 특이도 = 안나간사람을 안나갔다고 찾아낸 비율 = $\frac{500}{900}$ = $\frac{TN}{FP+TN}$
- 거짓긍정률 = 1-특이도 = 안나간사람인데 나갔다고 찾아낸 비율 (오해해서 미안해, 하지만 리컬을 올리려다보니 어쩔수 없었어) = $\frac{400}{900}$ = $\frac{FP}{FP+TN}$

`-` 오해해서미안해 vs 리컬을 그린곡선이 ROC커브: 오해한사람이 적은데 리컬도 우수하면 매우 좋은것, 그래서 초반부터 ROC커브가 높으면 좋은 모델임
