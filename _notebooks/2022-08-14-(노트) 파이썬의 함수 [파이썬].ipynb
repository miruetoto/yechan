(함수선언)

```python
def f(x,n):
    return x**n
```

(가능한 사용예시)

```python
f(2,3)
```

```python
f(x=2,n=3)
```

```python
f(2,n=3)
```

```python
f(n=3,x=2)
```

```python
f(3,x=2)
```

- x=3과 x=2이 동시에 선언된 모양임.
