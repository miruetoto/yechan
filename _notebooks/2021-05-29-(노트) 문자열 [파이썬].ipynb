% `연산자`는 왼쪽에 문자열 오브젝트, 그리고 오른쪽에는 명시적인 튜플이 있어야 연산이 진행된다. 

연산자라는 것에 포인트를 주면 당연히 아래와 같은 문법이 가능함을 알 수 있다. 

```python
s='addr: %s to %s' 
s % ('seoul','jeonju')
```

#### 딕셔너리 기반 포매팅

`-` 예제1

```python
'여기 %(food1)s 1개, %(food2)s 1개 주문이요' % {'food1':'짜장면', 'food2':'만두'}
```

```python
'여기 %(food1)s 1개, %(food2)s 1개 주문이요, 아.. 아니다.  %(food1)s은 취소하고 %(food2)s 2개 주세요..' % {'food1':'짜장면', 'food2':'만두'}
```

`–` 예제2

```python
mail='%(studentname)s 학생 안녕하세요\n저는 통계학과 최규빈 교수 입니다.\n전공진로설계 과목지침에 따라 %(studentname)s 학생과 2회 상담을 실시해야 합니다.\n저는 %(day)s에 시간이 괜찮은데 %(studentname)s 학생도 괜찮을까요?\n'
print(mail % {'studentname':'박혜원','day':'6월2일'})
```

`-` 예제3

```python
import pandas as pd
df=pd.DataFrame({'studentname':['박혜원','강호동'],'day':['6월2일','6월3일']})
df
```

```python
for i in [0,1]: 
    print(mail % dict(df.iloc[i]))
```

#### 메서드 

`-` 예제1

```python
mail='{studentname} 학생 안녕하세요\n저는 통계학과 최규빈 교수 입니다.\n전공진로설계 과목지침에 따라 {studentname} 학생과 2회 상담을 실시해야 합니다.\n저는 {day}에 시간이 괜찮은데 {studentname} 학생도 괜찮을까요?\n'
print(mail.format(studentname='박혜원',day='6월2일'))
```
