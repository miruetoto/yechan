`-` 예제3: cat()

```python
import torch
x1=torch.tensor([[1,2],[3,4]])
x2=torch.tensor([[10,20,30],[40,50,60]])
```

```python
x1
```

```python
x2
```

```python
torch.cat([x1,x2],dim=1) # cbind 
```

```python
torch.cat([x1,x2.t()],dim=0) # rbind
```

```python
torch.cat([x1,x2.t()]) # default is dim=0 
```

> Note: 차원이 안맞으면 에러가 난다. 

```python
torch.cat([x1,x2],dim=0)
```

```python
torch.cat([x1,x2.t()],dim=1) 
```

`-` 예제4: transpose()

```python
#collapse-output
x3=torch.rand(100,64,32,3)
x3
```

```python
x3.shape # 63*32*3 인 데이터가 100개 
```

```python
print('.transpose(0,1):', x3.shape, '-->', x3.transpose(0,1).shape)
print('.transpose(1,0):', x3.shape, '-->', x3.transpose(1,0).shape)
print('.transpose(0,2):', x3.shape, '-->', x3.transpose(0,2).shape)
print('.transpose(2,0):', x3.shape, '-->', x3.transpose(2,0).shape)
```
