```python
import torch
W = torch.Tensor([1,-2,5])
```

```python
X=torch.cat([torch.ones(100,1),torch.randn(100,2)],1)
X
```

```python
y=torch.mv(X,W)+torch.randn(100)*0.5
```

```python
import matplotlib.pyplot as plt
plt.plot(X[:,1],y,'o')
plt.plot(X[:,2],y,'o')
```

```python
What = torch.randn(3,requires_grad=True)
```

```python
What
```

```python
learning_rate=0.1
```

```python
losses=[]
```

```python
What.grad = None
yhat=torch.mv(X,What)
loss=torch.mean((y-yhat)**2)
loss.backward()
```

```python
What,What.data
```

```python
What.grad.data
```

```python
What.data=What.data - learning_rate * What.grad.data
```

```python
losses.append(loss.item())
```
