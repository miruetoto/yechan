### Import 

```python
import torch
import matplotlib.pyplot as plt 
from torch import nn, optim
```

### Data

```python
X=torch.randn(100,1)
W = torch.Tensor([1])
```

```python
y=torch.mv(X,W)+torch.randn(100)*0.5
ytrue=torch.mv(X,W)
```

```python
plt.plot(X,y,'o')
plt.plot(X,ytrue,'.')
```

### Learn

`-` (X,y)만 알고 있다고 가정하자. 

```python
(X,y)
```

```python
X.shape,y.shape
```

```python
net = nn.Linear(in_features=1,out_features=1,bias=False)
```

```python
optimizer=optim.SGD(net.parameters(),lr=0.1)
```

```python
loss_fn = nn.MSELoss()
```

```python
lesses = []
```

```python
for epoc in range(100):
    optimizer.zero_grad()
    yhat=net(X)
    loss=loss_fn(yhat.view_as(y),y)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())
```

```python
plt.plot(losses)
```
