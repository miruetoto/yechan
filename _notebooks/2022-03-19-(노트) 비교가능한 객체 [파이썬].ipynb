### 선언 

```python
a=1 
```

```python
dir(a)
```

### `__lt__`

```python
a.__lt__(0) # a=1 
```

```python
a.__lt__(4) # a=1 
```

### `__gt__`

```python
a.__gt__(4) # a=1 
```

### `__le__`

```python
a.__le__(2) # a=1 
```

```python
a.__le__(1) # a=1 
```

```python
a.__le__(0) # a=1 
```

### 비교값이 2개 이상이면? 

`-` 첫번째 값이 더 크면 이긴다. 

```python
['b',2] < ['a',3]
```

`-` 첫번째 값이 동점이면 두번째 값이 더 큰게 이긴다. 

```python
['a',2] < ['a',3]
```

```python
['a',2] < ['a',1]
```
