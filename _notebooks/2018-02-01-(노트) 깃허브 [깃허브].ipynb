### Auth

#### gh

`-` 아래를 친다. 
```
gh auth login 
```
그리고 아래에서 `GitHub.com` 을 선택
```
? What account do you want to log into?  [Use arrows to move, type to filter]
> GitHub.com
  GitHub Enterprise Server
```
아래에서 `y`를 선택 
```
? You're already logged into github.com. Do you want to re-authenticate? (y/N) 
```
아래에서 `HTTPS`를 선택
```
? What is your preferred protocol for Git operations?  [Use arrows to move, type to filter]
> HTTPS
  SSH
```
아래에서 `Paste an authentication token`를 선택 (웹브라우저를 확인할 수 없는 경우)
```
? How would you like to authenticate GitHub CLI?  [Use arrows to move, type to filter]
  Login with a web browser
> Paste an authentication token

```
그러면 아래의 메시지가 나옴. 
```
Tip: you can generate a Personal Access Token here https://github.com/settings/tokens
The minimum required scopes are 'repo', 'read:org', 'workflow'.
? Paste your authentication token: 
```
Tip 에 써있는대로 https://github.com/settings/tokens 에 접속한다. 그리고 
- `Generate new token` 버튼을 클릭
- 로그인 
- New personal access token 페이지가 나오면 title 에 fastpages 입력 
- 기간 무한대로 설정
- repo, workflow, read:org에 체크 (혹은 모두 체크) 
- Generate token 클릭 
- **토큰 복사**

다시 
```
Tip: you can generate a Personal Access Token here https://github.com/settings/tokens
The minimum required scopes are 'repo', 'read:org', 'workflow'.
? Paste your authentication token: 
```
로 돌아와서 복사한 토큰(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)을 붙여넣는다. 이제 아래와 같은 메시지가 나오면 성공임. 
```
- gh config set -h github.com git_protocol https
✓ Configured git protocol
✓ Logged in as guebin
```
