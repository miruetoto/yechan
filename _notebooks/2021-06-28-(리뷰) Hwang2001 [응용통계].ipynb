### Materials and methods 

#### Animals and experimental design 

`-` 38마리의 숫소+암소를 Hearnshaw1997의 연구에서 샘플링. 
 - 12마리: Brahman $\times$ Hereford (브라만은 미국의 어떤 품종임) 
 - 13마리: Brahman
 - 13마리: Piedmontese 

`-` 몇 가지 기본통계량 
- 38마리의 평균나이: 660day
- 평균몸무게(live): 444kg 
- 평균몸무게(carcass): 270kg 
- 12/13번째 갈비뼈의 평균 두께: 8.6mm

`-` 같은 종, 같은 성별을 모은 뒤에 (한 6~7마리 정도?) 각각의 소에게 아래와 같은 실험을 진행 

- 비교1: type of stimulation $\times$ time $\times$ ageing: $2\times 2 \times 4$ 
    - type stimulation: high / low
    - time: 3min / 40min 
    - ageing: 1day / 3day / 7day / 14day
- 비교2: time $\times$ ageing: $2\times 4$  
    - time: 40min / 60min 
    - ageing: 1day / 3day / 7day / 14day
- 비교3: existence of stimulation $\times$ time $\times$ ageing: $2\times 4$ 
    - existence of stimulation: stimulation / Non-stimulation
    - ageing: 1day / 3day / 7day / 14day

#### Slaughter and treatments


`-` 표1

![](2021-06-38-tb1.png)

`-` 표안의 숫자를 모두 더하면 38마리임. (단 1 left side + 1 right side = 1 carcass 로 계산)

`-` 6일동안 38마리의 소를 죽임 (6마리$\times$5일, 8마리$\times$1일)

`-` 38마리중에서 18마리는 각각 9마리씩 나눈뒤 3분뒤 고자극 (HV3) 3분뒤 저자 (LV3)을 각각 수행함. 

`-` 나머지 20마리는 모두 반을 잘라서 왼쪽, 오른쪽으로 나눔. (따라서 총 40개의 조각이 생김)

- 왼쪽 20조각은 각각 10조각씩 나누어 40분뒤에 고자극(HV40), 저자극(LV40)을 수행 
- 오른쪽 20조각중 10조각은 60분뒤에 고자극을 주고(HV60) 나머지 10조각은 자극을 주지 않음.
