### 예제1

`-` df1 데이터선언; 각 observation은 영화이름의 알파벳순서로 나열. 

```python
import pandas as pd 
df1=pd.DataFrame({'nodename':['Ant-man','Hulk','Iron-man'],'open':[2020,2012,2010]})
df1
```

`-` df2 데이터선언; observation은 영화개봉순으로 나열.

```python
df2=pd.DataFrame({'nodename':['Iron-man','Hulk','Ant-man'],'box-office':[100,300,50]})
df2
```

`-` df1,df2는 nodename이라는 공통된 열이 존재함 $\to$ inner join 으로 두 데이터를 합칠 수 있다. 

df1에서 나열된 obsevation 순으로 합치자. 

```python
pd.merge(df1,df2)
```

df2에서 나열된 observation 순으로 합치자. 

```python
pd.merge(df2,df1)
```
