`-` print(a)를 쳐서 나오는 결과와 주피터셀에 a만 쳐서 나오는 결과가 다르다? $\to$ 서로 다른 숨겨진 기능이 있다! $\to$ 결론: 그 기능은 `__repr__`에 저장됨

````python
class RPC: 
    def __init__(self,candidate=['가위','바위','보']): 
        self.candidate = candidate
        self.actions= list()
    def choose(self): 
        self.actions.append(np.random.choice(self.candidate))
    def show(self):
        print(self.actions[-1])
    def __repr__(self):
        return '낼 수 있는 패: {}\n기록: {}'.format(self.candidate,self.actions)
````

````python
a=RPC()
````

````python
a
````

`-` `__str__`, `__repr__` 을 건드리지 않고 싶다면? 

````python
class RPC: 
    def __init__(self,candidate=['가위','바위','보']): 
        self.candidate = candidate
        self.actions= list()
    def choose(self): 
        self.actions.append(np.random.choice(self.candidate))
    def show(self):
        print(self.actions[-1])
    def _repr_html_(self):
        html_str = """
        낼 수 있는 패: {} <br/>
        기록: {}
        """
        return html_str.format(self.candidate,self.actions)
````

````python
a=RPC()
````

````python
str(a)
````

````python
repr(a)
````
