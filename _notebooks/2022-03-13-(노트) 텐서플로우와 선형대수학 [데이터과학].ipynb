```python
tf.stack([a,b])
```

```python
tf.stack([a,b],axis=1)
```

- a,b를 (3,1)로 해석하자 

`-` (2,3,4,5) stack (2,3,4,5) -> (?,?,?,?,?) 

```python
a=tf.constant(np.array(range(2*3*4*5)).reshape(2,3,4,5))
b=-a
```

**case1:** (2,3,4,5) 를 (1,2,3,4,5) 로 해석 
- 최종차원은 ( ,2,3,4,5) stack ( ,2,3,4,5) = (2,2,3,4,5) // 0번째 축이 변하네요 

```python
tf.stack([a,b],axis=0)
```

**case2:** (2,3,4,5) 를 (2,1,3,4,5) 로 해석 
- 최종차원은 (2, ,3,4,5) stack (2, ,3,4,5) = (2,2,3,4,5) // 1번째 축이 변하네요 

```python
tf.stack([a,b],axis=1) 
```

**case3:** (2,3,4,5) 를 (2,3,1,4,5) 로 해석 
- 최종차원은 (2,3, ,4,5) stack (2,3, ,4,5) = (2,3,2,4,5) // 3번째 축이 변하네요 

```python
tf.stack([a,b],axis=2) 
```

**case4:** (2,3,4,5) 를 (2,3,4,1,5) 로 해석 
- 최종차원은 (2,3,4, ,5) stack (2,3,4, ,5) = (2,3,2,2,5) // 4번째 축이 변하네요 // 끝에서 2번째 축이 변한다 

```python
tf.stack([a,b],axis=-2) 
```

**case5:** (2,3,4,5) 를 (2,3,4,5,1) 로 해석 
- 최종차원은 (2,3,4,5, ) stack (2,3,4,5, ) = (2,3,4,5,2) // 5번째 축이 변하네요 // 마지막 축이 변한다 

```python
tf.stack([a,b],axis=-1) 
```
