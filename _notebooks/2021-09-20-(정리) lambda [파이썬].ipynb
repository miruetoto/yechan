`-` 예제5

(예비학습) 문자열의 대소비교

```python
'a'<'b'
```

```python
'c'<'b'
```

(예제시작) 

```python
lower = lambda x,y : x if x<y else y 
```

```python
lower('a','b')
```

```python
lower('c','b')
```

`-` 예제6

```python
def action(x): 
    return (lambda y: x+y)
```

```python
act = action(99)
```

```python
act(2)
```

이 예제를 응용하면 

```python
action = (lambda x: (lambda y: x+y))
```

```python
act = action(99)
```

```python
act(2)
```

괄호를 생략하여 더 간단히하면 

```python
action = lambda x: lambda y: x+y 
```

- 읽는법: 첫 입력 $x$를 받고 두번째 입력 $y$를 받고 결국 $x+y$를 출력하라.
