### 약속 

`-` 차원에 대한 약속: 차원을 명시하지 않아도 ${\bf X}, {\bf W}, {\boldsymbol \beta}$를 사용하면 아래와 같은 매트릭스를 혹은 벡터를 가정한다. 
- ${\bf X}$: (n,p) matrix 
- ${\bf W}$: (p,q) matrix 
- ${\boldsymbol \beta}$: (p,1) matrix (or length $p$ vector) 

`-` 매트릭스 혹은 벡터에 대한 곱셈에 대한 연산기호는 $@$로 정의한다. 이때 $@$는 생략가능하다. 즉 ${\bf X} @ {\bf W} = {\bf XW}$. 

### (n,p) matrix를 표현하는 방법 

`(ver1)` ${\bf X}=[X_1~ X_2~ \dots~ X_p]$

`(ver2)` ${\bf X}=\begin{bmatrix} x_1^\top \\ x_2^\top \\ \dots \\ x_n^\top  \end{bmatrix}$

`(ver3)` ${\bf X} = \big\{x_{ij} \big\}$

### transpose 

`(ver1)` ${\bf X}=[X_1~ X_2~ \dots~ X_p] \Longrightarrow {\bf X}^\top = \begin{bmatrix} X_1^\top \\ \dots \\ X_2^\top \\ X_p^\top \end{bmatrix}$

`(ver2)` ${\bf X}=\begin{bmatrix} x_1^\top \\ \dots \\ x_2^\top \\ x_n^\top \end{bmatrix} \Longrightarrow {\bf X}^\top=[x_1 ~ x_2 ~ \dots ~ x_n] $

### 행렬곱

`(ver1)` ${\bf X} {\bf W}= [X_1 \dots X_p] \begin{bmatrix} w_1^\top \\ \dots \\ w_p^\top \end{bmatrix} =  X_1 @ w_1^\top + \dots + X_p @ w_p^\top $

`(ver2)` ${\bf XW} =\begin{bmatrix}x_1^\top \\ \dots \\ x_n^\top \end{bmatrix}\Big[W_1\dots W_q\Big] = \begin{bmatrix}
x_{1}^\top  @ W_1 & \dots & x_{1}^\top @ W_q \\ 
\dots &\dots & \dots \\  
x_{n}^\top @ W_1 & \dots & n_{n\cdot}^\top @ W_q 
\end{bmatrix}$


`(ver3)` ${\bf X W}=\begin{bmatrix}
		x_1^\top \\ 
		\dots \\ 
		x_n^\top
	\end{bmatrix}{\bf W}=\begin{bmatrix}
	x_1^\top @ {\bf W} \\ 
	\dots \\ 
	x_n^\top @ {\bf W}
\end{bmatrix}
$

`(ver4)` ${\bf XW}={\bf X} [W_1 \dots W_q] = [{\bf X}@ W_1,\dots,{\bf X}@ W_q]$


#### 잘못적용하는 행렬곱

`-` 아래와 같은 표현들은 잘못되었다. (차원을 잘 생각해봐요~) 

- ${\bf XW} = {\bf X}\begin{bmatrix} w_1^\top \\ \dots \\ w_p^\top\end{bmatrix} = \begin{bmatrix}{\bf X}@ w_1^\top \\ \dots \\{\bf X}@ w_p^\top\end{bmatrix}$  

- ${\bf XW}= [X_1 ~ X_2 \dots ~ X_p] {\bf W} = [X_1 @ {\bf W},~ X_2 @ {\bf W}, \dots,  X_p @ {\bf W}]$ 

#### 행렬곱의 예제

`-` ${\bf X}{\boldsymbol \beta}=\begin{bmatrix}x_{1}^\top@ {\boldsymbol \beta} \\ \dots \\ x_{n}^\top @ {\boldsymbol \beta}\end{bmatrix}
$

`-` ${\bf X}^\top {\bf X} = [x_1,\dots, x_n]\begin{bmatrix} x_1^\top \\ \dots \\ x_n^\top \end{bmatrix}=\sum_{i=1}^{n} x_i@ x_i^\top$


`-` ${\boldsymbol \Psi}{\boldsymbol\Lambda}{\boldsymbol\Psi}^\top= \lambda_1 \Psi_1 @ \Psi_1^\top + \dots + \lambda_p \Psi_p@ \Psi_p^\top =  \sum_{i =1}^{p}  \lambda_i  \Psi_i @\Psi_i^\top$
- ${\boldsymbol \Psi}$: (p,p) matrix 

`-` ${\bf U}{\bf D}{\bf V}^\top= d_1 U_1@V_1^\top + \dots + d_p U_p@V_p^\top = \sum_{i =1}^{p} d_i U_i@V_i^\top$
- ${\bf U}$: (n,p) matrix 
- ${\bf D}$: (p,p) matrix 
- ${\bf V}$: (p,p) matrix

`-`  $tr({\bf A}{\bf B}{\bf C})=tr({\bf B}{\bf C}{\bf A})=tr({\bf C}{\bf A}{\bf B})$

`-` 임의의 col-vector $y$에 대하여 아래식이 성립한다.

y^\top y=tr(y y^\top)

### colspace, rowspace

`-` column space 
- **(def)** ${\cal C}({\bf X}):= \text{the linear space generated by columns of } {\bf X}:=\big\{ {\bf X}{\bf w}: {\bf w} \in \mathbb{R}^p \big\}$

### rank

`-` 매트릭스 ${\bf A}$의 column rank는 ${\bf A}$의 서로 독립인 열의 수로 정의한다. 그리고 ${\bf A}$의 row rank는 ${\bf A}$의 서로 독립인 행의 수로 정의한다. 즉 아래와 같다. 
- $\text{column rank of } {\bf A} = \text{maximal number of linearly independent columns of } {\bf A}$ 
- $\text{row rank of } {\bf A} = \text{maximal number of linearly independent row of } {\bf A}$ 

`-` 그런데 의해서 $\text{column rank of } {\bf A} = \text{row rank of } {\bf A}$ 가 성립하므로 ([위키](https://en.wikipedia.org/wiki/Rank_(linear_algebra))4절 참고)

- $rank({\bf A})=\text{column rank of } {\bf A} = \text{row rank of } {\bf A}$

와 같이 생각하여도 무방하다. 

`-` 요약하면 아래의 정리와 정의를 외우면 된다. 
- **(def)** $\text{column rank of } {\bf A} = \text{maximal number of linearly independent columns of } {\bf A}$ 
- **(def)** $\text{row rank of } {\bf A} = \text{maximal number of linearly independent row of } {\bf A}$ 
- **(thm)** $\text{column rank of } {\bf A} = \text{row rank of } {\bf A}$ 
- **(def)** $rank(A)=\text{column rank of } {\bf A} = \text{row rank of } {\bf A}$ 

`-` 아래의 정리들도 유용하다. 

- **(a)** $rank({\bf XW}) \leq \min\big(rank({\bf X}),rank({\bf W}) \big)$ 

`pf` (a) $rank({\bf XW}) \leq \min\big(rank({\bf X}),rank({\bf W}) \big)$를 증명하여 보자. 

증명전략: (1) $rank({\bf XW}) \leq rank({\bf X})$ 와 (2) $rank({\bf XW}) \leq rank({\bf W})$ 를 보이면 된다. 

먼저 (1)을 보이자. 

${\bf XW}=\begin{bmatrix}x_1^\top \\ x_2^\top \\ \dots \\ x_n^\top \end{bmatrix} {\bf W}= \begin{bmatrix}x_1^\top @ {\bf W} \\ x_2^\top @ {\bf W} \\ \dots \\ x_n^\top @ {\bf W} \end{bmatrix}= \begin{bmatrix}x_{11} w_1^\top + x_{12} w_2^\top + \dots + x_{1p} w_p^\top \\ x_{21} w_1^\top + x_{22} w_2^\top + \dots + x_{2p} w_p^\top  \\ \dots \\ x_{n1} w_1^\top + x_{n2} w_2^\top + \dots + x_{np} w_p^\top \end{bmatrix}$ 

이므로 ${\bf XW}$의 모든 row는 ${\bf W}$의 row들의 linear combination으로 표현가능하다. 

따라서 ${\cal R}({\bf XW}) \subset {\cal R}({\bf W})$이다. 따라서 $rank({\bf XW}) \leq rank({\bf X})$이다. 

유사한 논리전개로 (2)를 보일수 있다. 따라서 증명이 되었음.
