### 표현 

`-` ${\bf X}=[X_1, X_2, \dots, X_p]=\begin{bmatrix} x_1^\top \\ \dots \\ x_n^\top \end{bmatrix}$

`-` ${\bf W}=[W_1,W_2,\dots, W_q]=\begin{bmatrix} w_1^\top \\ w_2^\top \\ \dots \\ w_p^\top \end{bmatrix}$

### 행렬곱

`-` ${\bf X W}= [X_1 \dots X_p] \begin{bmatrix} w_1^\top \\ \dots \\ w_p^\top \end{bmatrix} =  X_1 @ w_1^\top + \dots + X_p @ w_p^\top $

`-` ${\bf XW} =\begin{bmatrix}x_1^\top \\ \dots \\ x_n^\top \end{bmatrix}\Big[W_1\dots W_q\Big] = \begin{bmatrix}
x_{1}^\top  @ W_1 & \dots & x_{1}^\top @ W_q \\ 
\dots &\dots & \dots \\  
x_{n}^\top @ W_1 & \dots & n_{n\cdot}^\top @ W_q 
\end{bmatrix}$


`-` ${\bf X W}=\begin{bmatrix}
		x_1^\top \\ 
		\dots \\ 
		x_n^\top
	\end{bmatrix}{\bf W}=\begin{bmatrix}
	x_1^\top @ {\bf W} \\ 
	\dots \\ 
	x_n^\top @ {\bf W}
\end{bmatrix}
$

`-` ${\bf XW}={\bf X} [W_1 \dots W_q] = [{\bf X}@ W_1,\dots,{\bf X}@ W_q]$


### 행렬곱의 예제

`-` ${\bf X}{\boldsymbol \beta}=\begin{bmatrix}x_{1}^\top{\boldsymbol \beta} \\ \dots \\ x_{n}^\top {\boldsymbol \beta}\end{bmatrix}
$

`-` ${\bf X}^\top {\bf X} = [x_1,\dots, x_n]\begin{bmatrix} x_1^\top \\ \dots \\ x_n^\top \end{bmatrix}=\sum_{i=1}^{n} x_i@ x_i^\top$


`-` ${\boldsymbol \Psi}{\boldsymbol\Lambda}{\boldsymbol\Psi}^\top= \lambda_1 \Psi_1 @ \Psi_1^\top + \dots + \lambda_p \Psi_p@ \Psi_p^\top =  \sum_{i =1}^{p}  \lambda_i  \Psi_i @\Psi_i^\top$

`-` ${\bf U}{\bf D}{\bf V}^\top= d_1 U_1@V_1^\top + \dots + d_p U_p@V_p^\top = \sum_{i =1}^{p} d_i U_i@V_i^\top$

### 트레이스 

`-`  $tr({\bf A})=tr({\bf A}')$

`-`  $tr({\bf A}{\bf B}{\bf C})=tr({\bf B}{\bf C}{\bf A})=tr({\bf C}{\bf A}{\bf B})$

`-` 임의의 col-vector $y$에 대하여 아래식이 성립한다. 
$
y^\top y=tr(y y^\top)
