### Long-Run Behavior


`(정의)` Let $E$ be a countable space and let $\alpha$ and $\beta$ be probability distribution on $E$. The distance in variation $d_V(\alpha,\beta)$ betweem $\alpha$ and $\beta$ is defined by 

$$ d_V(\alpha,\beta)=\frac{1}{2}|\alpha-\beta|=\frac{1}{2}\sum_{i\in E} |\alpha(i)-\beta(i)|.$$

`-` 예제: $\alpha=(0.1,0.1,0.8)$, $\beta=(0.2,0.1,0.7)$ $\to$ $d_V(\alpha,\beta)=\frac{1}{2}\big(0.1+0+0.1\big)=0.1$

`-` $X$,$Y$를 $E$에서 정의된 확률변수라고 하고 $X$, $Y$의 distribution을 각각 ${\cal L}(X)$ and ${\cal L}(Y)$ 라고 하자. 아래와 같은 기호를 약속하자. 

$$d_V({\cal L}(X),{\cal L}(Y))=d_V(X,Y)$$

`(레마)` Let $X$ and $Y$ be two random variable with values in the same coutable set $E$. Then 

$$\sup_{A \subset E}|P(X\in A)-P(Y \in A)|=\sup_{A \subset E}\{P(X\in A)-P(Y\in A) \}=d_V(X,Y)$$

`(증명)`

`>` 첫번째등호: $\sup_{A \subset E}|P(X\in A)-P(Y \in A)|=\sup_{A \subset E}\{P(X\in A)-P(Y\in A) \}$.

모든 $A$에 대하여 아래를 만족하는 $B$가 존재한다. (take $B=A$ or $B=\bar{A}$)

$$|P(X\in A)-P(Y \in A)|=P(X\in B)-P(Y \in B)$$

따라서 증명 끝. 

`>` 두번째등호: $\sup_{A \subset E}\{P(X\in A)-P(Y\in A) \}=d_V(X,Y)$.

`*` 관찰1: 아래가 성립함을 관찰하자. 

$$P(X\in A)-P(Y \in A)=\sum_{i \in E}1_{A}(i)\{P(X=i)-P(Y=i) \}$$

우변은 $A^*=\{i \in E; P(X=i)>P(Y=i)\}$에서 최대화된다. 

`*` 관찰2: 모든 $A \subset E$에 대하여 아래가 성립한다. 

$$\sum_{i\in E}1_A(i)\{P(X=i)-P(Y=i)\}+\sum_{i\in E}1_{\bar{A} }(i)\{P(X=i)-P(Y=i)\}=0$$

`-` 관찰1에서 살펴본 $A^*$에 대하여서도 관찰2의 등호가 성립하므로 

$$\sum_{i\in E}1_{A^*}(i)\{P(X=i)-P(Y=i)\}=\frac{1}{2}\sum_{i\in E}|P(X=i)-P(Y=i)|$$

`(정의)` Let $E$, $F$ denote countable sets. The matrix ${\bf Q}$ indexed by $F \times E$ called stochastic matrix whose rows (index by $F$) are probability distributions on $E$. 

> Note: 따라서 어떤 매트릭스의 rowsum이 모두 1이면 stochastic matrix 이다. 

`(정의)` Let ${\bf Q}$ be a stochastic matrix indexed by $F\times E$. Its ***ergodic coefficient*** is 

$$\delta({\bf Q}):=\frac{1}{2}\sup_{i,j \in F} \sum_{k \in E}|q_{ik}-q_{jk}|=\sup_{i,j \in F} d_V(q_{i\cdot},q_{j\cdot})=\sup_{i,j \in F}\sup_{A \subset E}(q_{iA}-q_{jA}).$$


`-` Exercise 4.4.12, 

$$\delta({\bf Q})=1- \inf_{i,j \in F}\sum_{k \in E}q_{ik} \wedge q_{jk}$$.

`-` 예제: ${\bf Q}=\begin{bmatrix}0.1 & 0.1 & 0.8 \\ 0.2 & 0.1 & 0.7  \\ 0.4 &0.5& 0.1 \end{bmatrix}$ 이라고 하자. 그러면 $\delta({\bf Q})$ 는 

- 1-2열의 원소값차이의 절대값합 : $|0.1-0.2|+|0.1-0.1|+|0.8-0.7|=0.2$
- 1-3열의 원소값차이의 절대값합 : $|0.1-0.4|+|0.1-0.5|+|0.8-0.1|=1.4$
- 2-3열의 원소값차이의 절대값합 : $|0.2-0.4|+|0.1-0.5|+|0.7-0.1|=1.2$

중 가장 큰 값을 고른뒤에 $\frac{1}{2}$을 곱하면 된다. 따라서 $\delta({\bf Q})=\frac{1.4}{2}=0.7$.

`-` 모든 row가 동일한 matrix의 경우 (MwIR) 에르고딕계수가 0이다. 

`-` 예제 (4.3.13) 2개의 마코프체인 $\{X_n\}$ and $\{Y_n\}$가 transition matrix ${\bf P}$를 가진다고 하자. $\{X_n\}$, $\{Y_n\}$은 각각 irreducible 하고 양의 

`(정리)` Let ${\bf Q}_1$ and ${\bf Q}_2$ be two stochastic matrices indexed by $F \times G$ and $G\times E$, respectively. Then 

$$\delta({\bf Q}_1{\bf Q}_2)\leq \delta({\bf Q}_1)\delta({\bf Q}_2)$$

### Weakly ergodic 

`(정리)` The family of transition matrices $\big\{ {\bf P}(n)\big\}_{n\geq 0}$ is weakly ergodic if and only if for all $m\geq 0$, 

$$\lim_{k\to \infty} \delta\big({\bf P}(m,k) \big)=0$$

where ${\bf P}(m,k)={\bf P}(m){\bf P}(m+1)\dots {\bf P}(k-1)$.
